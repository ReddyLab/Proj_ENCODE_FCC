## Merge and arrange

**Test**

```r
dat = bind_rows(
    dat_snps_OL13_rsid_query_hg38, 
    dat_snps_OL13_rsid_remain_query_hg38)
head(dat)
```

```r
### init
lst = list(
    "hg38" =  bind_rows(
        dat_snps_OL13_rsid_query_hg38, 
        dat_snps_OL13_rsid_remain_query_hg38
    ),
    "hg19" = bind_rows(
        dat_snps_OL13_rsid_query_hg19, 
        dat_snps_OL13_rsid_remain_query_hg19
    )
)

### process
lst = lapply(lst, function(dat){
    dat = dat %>%
        dplyr::mutate(
            RSID     = ID,
            Chrom    = paste0("chr", `#CHROM`),
            Loc = POS,
            Allele_Ref = REF,
            Allele_Alt = ALT
        ) %>%
        dplyr::filter(str_detect(RSID, "rs")) %>%
        dplyr::select(RSID, Chrom, Loc, Allele_Ref, Allele_Alt) 
    return(dat)
})

### combine and arrange
dat = bind_rows(lst, .id = "Genome")
dat = dat %>% 
    dplyr::mutate(Genome = paste("Loc", Genome, sep="_")) %>% 
    dplyr::distinct() %>%
    tidyr::spread(Genome, Loc) 

chroms = paste0("chr", c(1:22, "X", "Y"))
dat$Chrom = factor(dat$Chrom, levels=chroms)
dat = dat %>% dplyr::arrange(Chrom, Loc_hg19)

### assign and show
dat_snps_OL13_rsid_query = dat
head(dat)
```
