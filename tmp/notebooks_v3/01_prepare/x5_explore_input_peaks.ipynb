```r
###
fdiry = file.path(FD_RES, "region", "KS91_K562_ASTARRseq_peak_macs")
fname = "KS91_K562_hg38_ASTARRseq_Input.all_reps.masked.union_narrowPeak.q5.sorted.bed"
fpath = file.path(fdiry, fname)

###
cnames = c("Chrom", "Start", "End")
dat = read_tsv(fpath, col_names = cnames)

###
dat_peaks_sort = dat
print(dim(dat))
head(dat)
```

```r
fun_sortbed = function(dat){
    ### reset levels in chromosome
    CHROMS = paste0("chr", c(1:22, "X", "Y"))
    dat$Chrom = factor(dat$Chrom, levels=CHROMS)
    
    ### sort dataframe by location (Chrom, Start, End)
    dat = dat %>% arrange(Chrom, Start, End)
    return(dat)
}
```

```r
dat = dat_peaks
dat = fun_sortbed(dat)
dat_sorted = dat
```

```r
dat1 = dat_peaks
dat2 = dat_sorted
dat3 = dat_peaks_sort

print(unique(dat1$Chrom))
cat("\n===============================\n")
print(unique(dat2$Chrom))
cat("\n===============================\n")
print(unique(dat3$Chrom))
```

```r
dat1 = dat_peaks
dat2 = dat_sorted
dat3 = dat_peaks_sort
cnames = colnames(dat1)

for (cname in cnames){
    cat("======================\n")
    print(all(dat1[[cname]] == dat2[[cname]]))
    print(all(dat1[[cname]] == dat3[[cname]]))
    print(all(dat2[[cname]] == dat3[[cname]]))
}
```
