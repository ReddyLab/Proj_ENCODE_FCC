## Test loop

```r
lst = list(
    "ASTARR" = dat_ranknorm_astarr,
    "WSTARR" = dat_ranknorm_wstarr,
    "TMPRA"  = dat_ranknorm_tmpra
)

dat = bind_rows(lst, .id = "Assay")
dat = dat %>% 
    dplyr::mutate(RankNorm = RankNorm_Total) %>%
    dplyr::select(Assay, Peak, Dataset, Region, RankNorm)

mat = dat %>% 
    tidyr::spread(Assay, RankNorm) %>%
    dplyr::rowwise() %>%
    dplyr::mutate(
        Score_mean = mean(c(ASTARR, TMPRA, WSTARR)),
        Score_std  = sd(  c(ASTARR, TMPRA, WSTARR)))

dat_ranknorm = dat
mat_ranknorm = mat

cat("\n========================\n")
print(head(dat))
cat("\n========================\n")
print(head(mat))
```

```r
mat = mat_ranknorm
gpt = ggplot(mat, aes(x=Score_mean, Score_std)) + 
    geom_hline(yintercept = 0) +   
    geom_point() + 
    geom_smooth(method = "loess", formula = "y~x") +
    labs(x = "Mean (Scores)", y = "Std (Scores)") +
    theme_cowplot()

options(repr.plot.height=5, repr.plot.width=8)
print(gpt)
```

## Visualization

```r
dat   = mat_ranknorm
peaks = dat %>% 
    dplyr::arrange(Score_mean) %>%
    dplyr::pull(Peak)

head(peaks)
```

```r
assays = c("ASTARR", "WSTARR", "TMPRA")
```

```r
dat = dat_ranknorm
dat = dat %>% dplyr::mutate(Peak = factor(Peak, levels=peaks))
gpt = ggplot(dat)
```

```r
theme_fill = scale_fill_gradient2(
    low  = "#4575b4",
    mid  = "#ffffbf",
    high = "#d73027",
    midpoint = 0.5
)

theme_text = theme(
    axis.ticks.x = element_blank(),
    axis.text.x  = element_blank()
)

###
dat = dat_ranknorm
dat = dat %>%
    dplyr::mutate(
        Peak  = factor(Peak,  levels=peaks),
        Assay = factor(Assay, levels=assays)
    ) %>% 
    dplyr::arrange(Peak)
res_ranknorm = dat

gpt = ggplot(dat, aes(x=Peak, y=Assay, fill=RankNorm)) +
    geom_tile() +
    theme_cowplot() + 
    theme_text +
    theme_fill
gpt_ranknorm = gpt

options(repr.plot.height=5, repr.plot.width=15)
print(gpt)
```

```r
###
dat = dat_crispr
dat = left_join(dat_peak_info, dat_crispr, by = c("Peak", "Chrom", "Start", "End"))
dat = dat %>% dplyr::mutate(Score = ifelse(is.na(Score_mean), 0, Score_mean))
head(dat, 30)
```

```r
###
dat = dat_crispr
dat = left_join(dat_peak_info, dat_crispr, by = c("Peak", "Chrom", "Start", "End"))
dat = dat %>% dplyr::mutate(Score = ifelse(is.na(Score_mean), 0, Score_mean))
#dat = dat %>% dplyr::mutate(Score = ifelse(is.na(Score_mean_abs), 0, Score_mean_abs))

dat = dat %>% dplyr::mutate(Peak  = factor(Peak,  levels=peaks)) %>% arrange(Peak)
res_crispr = dat

gpt = ggplot(dat, aes(x=Peak, y=Score)) +
    geom_col() +
    labs(x="", y="CRISPR Mean Log2FC") +
    theme_cowplot() +
    theme_text
gpt_crispr = gpt

options(repr.plot.height=5, repr.plot.width=15)
print(gpt)
```

## Plot CA

```r
###
dat = res_ca
gpt = ggplot(dat, aes(x=Peak, y=Name, fill=Score)) +
        geom_tile() +
        scale_fill_gradient(low = "white", high = "red") +
        labs(x="", y="") +
        theme_cowplot() +
        theme(axis.text.x = element_blank(),
              axis.text.y = element_text(size=30),
              axis.ticks.x = element_blank(),
              legend.position = "bottom",
              legend.key.size = unit(1, 'cm'),
              legend.title = element_text(size=20),
              legend.text = element_text(size=20))

###
gpt_ca = gpt

###
options(repr.plot.height=2, repr.plot.width=15)
print(gpt)
```

```r
lst = list(
    gpt_crispr,
    gpt_ranknorm,
    gpt_ca
)

lst_gpt_legend = lapply(lst, function(gpt){
    gpt = gpt + theme(legend.position = "right")
    legend = cowplot::get_legend(gpt)
    return(legend)
})

lst_gpt = lapply(lst, function(gpt){
    gpt = gpt + 
        labs(x="", y="") +
        theme(
            axis.text.y  = element_text(size=15), # element_blank(),
            #axis.ticks.y = element_blank(),
            legend.position = "none")
    return(gpt)
})

lst_gpt[[1]] = lst_gpt[[1]] + theme(axis.text.y = element_text(size=20))
lst_gpt[[2]] = lst_gpt[[2]] + theme(axis.text.y = element_blank())
lst_gpt[[3]] = lst_gpt[[3]] + theme(axis.text.y = element_blank())
```
