## Compare ASTARR / WSTARR / TMPRA

```r
head(lst_dat[[1]])
```

```r
lst = lst_dat
dat = bind_rows(lst, .id = "Assay")

dat = dat %>% 
    dplyr::select(Assay, Peak, Region, RankNorm_Region) %>%
    tidyr::spread(Assay, RankNorm_Region)

dat_ranknorm = dat
print(dim(dat))
head(dat)
```

```r
dat = dat_ranknorm
lst = split(dat, dat$Region)
lst = lapply(lst, function(dat){
    dat = dat %>% dplyr::arrange(TMPRA, WSTARR, ASTARR)
    return(dat)
})

dat = bind_rows(lst)
dat_annot = dat %>% dplyr::select(Peak, Region)                %>% column_to_rownames("Peak") 
dat_value = dat %>% dplyr::select(Peak, TMPRA, WSTARR, ASTARR) %>% column_to_rownames("Peak")
```

```r
dat = dat_color
vec = dat$Color
names(vec) = dat$Region

num_palette = 50
vec_color_value = colorRampPalette(c("white", "red"))(num_palette)
vec_color_annot = list(Region = vec)

options(repr.plot.height=8, repr.plot.width=6)
pheatmap(dat_value,
         cluster_row   = FALSE,
         cluster_cols  = FALSE,
         show_rownames = FALSE,
         annotation_names_row = FALSE,
         annotation_row    = dat_annot,
         annotation_colors = vec_color_annot,
         color=vec_color_value,
         fontsize = 15)
```

```r
dat = dat_ranknorm
dat = dat %>%
    dplyr::rowwise() %>%
    dplyr::mutate(
        Score_mean = mean(c(ASTARR, TMPRA, WSTARR)),
        Score_std  = sd(  c(ASTARR, TMPRA, WSTARR)))

gpt = ggplot(dat, aes(x=Score_mean, y=Score_std)) + 
    geom_point(alpha=0.5) + 
    geom_smooth(method = "loess", ) +
    theme_cowplot()

options(repr.plot.height=5, repr.plot.width=5)
print(gpt)
```
