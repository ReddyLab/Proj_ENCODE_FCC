## PLot

```r
lst = lst_peak_Log2FC
lst_gpt_Log2FC = lapply(lst, function(dat){
    dat$Peak = factor(dat$Peak, levels=peaks)
    txt = unique(dat$Assay)
    gpt = ggplot(dat, aes(x=Peak, y=log2FoldChange, fill=Region)) + 
        geom_bar(stat = "identity") +
        labs(title=txt) +
        theme_cowplot() + 
        theme(axis.text.x = element_blank(),
              axis.text.y = element_text(size=20),
              axis.title.y = element_text(size=20),
              title =element_text(size=30, face='bold'),
             legend.position = "none")
    return(gpt)
})

options(repr.plot.height=15, repr.plot.width=20)
plot_grid(plotlist = lst_gpt_Log2FC, ncol=1, align = "v")
```

```r
dat = dat_rank_diff 
dat = dat %>% 
    dplyr::select(Peak, Region, z_diff_ASTARR_WSTARR, z_diff_ASTARR_TMPRA, z_diff_WSTARR_TMPRA) %>%
    tidyr::gather(Assay, z_Rank_Diff, -Peak, -Region) %>%
    dplyr::mutate(Assay = recode(
        Assay, 
        'z_diff_ASTARR_WSTARR' = 'ASTARR vs WSTARR', 
        'z_diff_ASTARR_TMPRA'  = 'ASTARR vs TMPRA',
        'z_diff_WSTARR_TMPRA'  = 'WSTARR vs TMPRA'
    ))

lst = split(dat, dat$Assay)

lst_gpt_rankdiff = lapply(lst, function(dat){
    dat$Peak = factor(dat$Peak, levels=peaks)
    txt = unique(dat$Assay)
    gpt = ggplot(dat, aes(x=Peak, y=z_Rank_Diff, fill=Region)) + 
        geom_bar(stat = "identity") +
        labs(title=txt, y="Standardized Rank Difference") +
        theme_cowplot() + 
        theme(axis.text.x = element_blank(),
              axis.text.y = element_text(size=20),
              axis.title.y = element_text(size=20),
              title =element_text(size=30, face='bold'),
             legend.position = "none")
    return(gpt)
})

options(repr.plot.height=15, repr.plot.width=20)
plot_grid(plotlist = lst_gpt_rankdiff, ncol=1, align = "v")
```

```r
lst = lst_peak_annot
lst_gpt_annot = lapply(lst, function(dat){
    dat$Peak = factor(dat$Peak, levels=peaks)
    gpt = ggplot(dat, aes(x=Peak, y=Name, fill=Label)) +
        geom_tile() +
        scale_fill_manual(values = c("1" = "black", "0" = "white")) +
        labs(x="", y="") +
        theme_cowplot() +
        theme(axis.text.x = element_blank(),
              axis.text.y = element_text(size=30),
              #axis.ticks.x = element_blank(),
              legend.position = "none")
    return(gpt)
})

dat = dat_peak_info
dat$Peak = factor(dat$Peak, levels=peaks)
print(sum(is.na(dat$Score)))

gpt = ggplot(dat, aes(x=Peak, y=Name, fill=Score)) +
        geom_tile() +
        scale_fill_gradient(low = "white", high = "red") +
        labs(x="", y="") +
        theme_cowplot() +
        theme(axis.text.x = element_blank(),
              axis.text.y = element_text(size=30),
              #axis.ticks.x = element_blank(),
              legend.position = "bottom",
              legend.key.size = unit(1, 'cm'),
              legend.title = element_text(size=20),
              legend.text = element_text(size=20))

lst_gpt_annot = append(lst_gpt_annot, list(gpt))
```

```r
options(repr.plot.height=20, repr.plot.width=30)
plot_grid(plotlist = lst_gpt_annot, 
          ncol=1, 
          align = "v",
          rel_heights = c(5, 5, 1, 1, 1, 2))
```

```r
lst_gpt = lst_gpt_Log2FC
lst_gpt = append(lst_gpt, lst_gpt_annot)

options(repr.plot.height=30, repr.plot.width=30)
plot_grid(plotlist = lst_gpt, 
          ncol=1, 
          align = "v",
          rel_heights = c(4, 4, 4, 5, 5, 1, 1, 1, 2))
```

```r
dat = dat_rank_diff 
dat = dat %>% 
    dplyr::select(Peak, Region, z_diff_ASTARR_WSTARR, z_diff_ASTARR_TMPRA, z_diff_WSTARR_TMPRA) %>%
    tidyr::gather(Assay, z_Rank_Diff, -Peak, -Region) %>%
    dplyr::mutate(Assay = recode(
        Assay, 
        'z_diff_ASTARR_WSTARR' = 'ASTARR vs WSTARR', 
        'z_diff_ASTARR_TMPRA'  = 'ASTARR vs TMPRA',
        'z_diff_WSTARR_TMPRA'  = 'WSTARR vs TMPRA'
    ))

dat_sum = dat %>%
    group_by(Peak) %>%
    summarise(Sum = sum(abs(z_Rank_Diff)), .groups="drop") %>% 
    arrange(Sum)

dat_sum$Peak = factor(dat_sum$Peak, levels=unique(dat_sum$Peak))
gpt_sum = ggplot(dat_sum, aes(x=Peak, y=Sum)) + 
    geom_bar(stat = "identity") +
    labs(y="", title="Sum(|Standardized Rank Difference|)") +
    theme_cowplot() + 
    theme(axis.text.x = element_blank(),
          axis.text.y = element_text(size=20),
          axis.title.y = element_text(size=20),
          title =element_text(size=30, face='bold'))
          #legend.position = "none")

options(repr.plot.height=5, repr.plot.width=20)
print(gpt_sum)
```
