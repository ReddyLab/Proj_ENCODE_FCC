## Import regions

```r
### set file path
fdiry = file.path(FD_RES, "region", "KS91_K562_ASTARRseq_peak_macs")
fname = "KS91_K562_hg38_ASTARRseq_Input.peaks.MYC.FADS.GATA1.tsv"
fpath = file.path(fdiry, fname)

### import data
dat_region = read_tsv(fpath, show_col_types = FALSE)
peaks = dat_region$Peak

print(dim(dat_region))
head(dat_region)
```

## Plot CPM

```r
ASSAYS = c("KS91_K562_ASTARRseq", "A001_K562_WSTARRseq", "Tewhey_K562_TileMPRA")
```

```r
lst = list(dat_log2fc_cpm_astarr, dat_log2fc_cpm_wstarr, dat_log2fc_cpm_tmpra)
dat = bind_rows(lst) %>% left_join(dat_region, ., by=c("Chrom", "Start", "End", "Peak"))
dat$Assay = factor(dat$Assay, levels=ASSAYS)
lst = split(dat, dat$Assay)
```

```r
lst = lapply(lst, function(dat){
    dat$Peak = factor(dat$Peak, levels=peaks)
    txt = unique(dat$Assay)
    gpt = ggplot(dat, aes(x=Peak, y=log2FoldChange, fill=Region)) + 
        geom_bar(stat = "identity") +
        labs(title=txt) +
        theme_cowplot() + 
        theme(axis.text.x = element_blank())
    return(gpt)
})
```

```r
options(repr.plot.height=15, repr.plot.width=20)
plot_grid(plotlist = lst, ncol=1, align = "v")
```

## Plot scatter plot (ASTARR)

```r
dat1 = dat_log2fc_cpm_astarr %>% dplyr::rename(Log2FC_cpm = log2FoldChange)
dat2 = dat_log2fc_deseq_astarr %>%
    dplyr::select(Peak, log2FoldChange) %>%
    dplyr::rename(Log2FC_deseq = log2FoldChange)
dat_comb = left_join(dat1, dat2, by="Peak")
print(dim(dat_comb))
head(dat_comb)
```

```r
dat = dat_comb %>% filter_all(all_vars(!is.infinite(.))) %>% na.omit %>% dplyr::slice_sample(n = 100000)

gpt_hist_top   = ggplot() + geom_histogram(data = dat, aes(x=Log2FC_cpm), bins = 50) +
    theme_bw()
gpt_hist_right = ggplot() + geom_histogram(data = dat, aes(x=Log2FC_deseq), bins = 50) + 
    coord_flip() +
    theme_bw() +
    theme(axis.text.x = element_text(vjust=0.5, angle=270))

gpt_scatter = ggplot() +
    geom_point(data = dat, aes(x=Log2FC_cpm, y=Log2FC_deseq), size=0.1, alpha=0.3) + 
    labs(x="Log2FC (CPM)", y="Log2FC (DESeq2)", title="ASTARR") +
    theme_cowplot() +
    theme(
        panel.grid.major = element_line(colour="grey90", linewidth=0.5),
        panel.grid.minor = element_line(colour="grey90", linewidth=0.2))

gpt_empty = ggplot()+
    geom_point(aes(1,1), colour="white")+
    theme(
        axis.ticks       = element_blank(), 
        panel.background = element_blank(), 
        axis.text.x      = element_blank(), 
        axis.text.y      = element_blank(),           
        axis.title.x     = element_blank(), 
        axis.title.y     = element_blank())
```

```r
options(repr.plot.height=7, repr.plot.width=7)
grid.arrange(
    gpt_hist_top, gpt_empty, gpt_scatter, gpt_hist_right, 
    ncol=2, nrow=2, 
    widths =c(4, 1), 
    heights=c(1, 4))
```
