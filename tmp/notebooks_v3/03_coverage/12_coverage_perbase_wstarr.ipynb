**RUN**

```bash
sbatch -p ${NODE} \
    --mem 20G \
    --array 0-6 \
    -o ${FD_LOG}/coverage_perbase_WSTARR.%a.txt \
    <<'EOF'
#!/bin/bash
### set directories & global variables
source ../config/config_duke.sh
ASSAY=A001_K562_WSTARRseq
PREFIX=A001_K562_WSTARRseq
REGIONS=(GATA1 MYC FADS)

### set input and output directory
FD_INP=${FD_RES}/results/${ASSAY}/fragment_count
FD_OUT=${FD_RES}/results/${ASSAY}/coverage

### print start message
timer_start=`date +%s`
echo "Hostname:          " $(hostname)
echo "Slurm Array Index: " ${SLURM_ARRAY_TASK_ID}
echo "Time Stamp:        " $(date +"%m-%d-%y+%T")
echo

### +++++++++++++++++++++++++++++++++++++++++++++++

for REGION in ${REGIONS[@]}; do

    ### set input and output file
    FP_REG=${FD_OUT}/region_${REGION}_perbase.bed
    
    FP_INPS=($(ls ${FD_INP}/*${REGION}*))
    FP_INP=${FP_INPS[${SLURM_ARRAY_TASK_ID}]}
    FN_INP=$(basename ${FP_INP})
    
    FN_OUT=${FN_INP}
    FP_OUT=${FD_OUT}/${FN_OUT}
    
    ### show input file
    echo ++++++++++++++++++++++++++++++++++++++
    echo
    echo "Input: " ${FP_INP}
    echo
    echo "show first few lines of input"
    zcat ${FP_INP} | head -n 5
    echo
    echo "show last few lines of output"
    zcat ${FP_INP} | tail -n 5
    echo
    echo "Input: " ${FP_REG}
    echo
    echo "show first few lines of input"
    cat  ${FP_REG} | head -n 5
    echo
    echo "show last few lines of output"
    cat  ${FP_REG} | tail -n 5
    echo
    echo
    
    ### execute
    bedtools map \
        -a ${FP_REG} \
        -b ${FP_INP} \
        -o sum |\
        gzip -c > ${FP_OUT}
    
    ### show output file
    echo "Output: " ${FP_OUT}
    echo
    echo "show first few lines of output:"
    zcat ${FP_OUT} | head -n 5
    echo
    echo "show last few lines of output"
    zcat ${FP_OUT} | tail -n 5
    echo
done

### print end message
timer=`date +%s`
runtime=$(echo "${timer} - ${timer_start}" | bc -l)
echo
echo 'Done!'
echo "Run Time: $(displaytime ${runtime})"
EOF
```
