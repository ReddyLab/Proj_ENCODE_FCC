**RUN**

```bash
sbatch -p ${NODE} \
    --exclude=dl-01 \
    --cpus-per-task 8 \
    --mem 20G \
    --output ${FD_LOG}/coverage_library_size_TMPRA.%a.txt \
    --array 0 \
    <<'EOF'
#!/bin/bash
### set directories & global variables
source ../config/config_duke.sh

### set input and output file paths
ASSAY=Tewhey_K562_TileMPRA

FD_INP=${FD_RES}/results/${ASSAY}/fragment_count
FP_INPS=($(ls ${FD_INP}/*Input*bed) $(ls ${FD_INP}/*Output*bed))

FD_OUT=${FD_RES}/results/${ASSAY}/coverage/summary
FN_OUT=library_size_output.csv
FP_OUT=${FD_OUT}/${FN_OUT}

### print start message
timer_start=`date +%s`
echo "Hostname:          " $(hostname)
echo "Slurm Array Index: " ${SLURM_ARRAY_TASK_ID}
echo "Time Stamp:        " $(date +"%m-%d-%y+%T")
echo

### init
mkdir -p   ${FD_OUT}
echo  -n > ${FP_OUT}

### get library size
for FP_INP in ${FP_INPS[@]}; do
    
    ### show I/O file
    echo ++++++++++++++++++++++++++++++++++++
    echo "Input file path: ${FP_INP}"
    echo "Show the first few lines"
    cat ${FP_INP} | head -n 3
    echo
    
    ### count the total counts and append the results to the output file
    FN_INP=$(basename ${FP_INP})
    cat ${FP_INP} |\
        awk -v OFS=, -v FNAME=${FN_INP} '{sum += $5} END {print FNAME, sum}' \
        >> ${FP_OUT}
done

### show I/O file
echo "Output file path: ${FP_OUT}"
echo "Show the lines"
cat ${FP_OUT}
echo

### print end message
timer=`date +%s`
runtime=$(echo "${timer} - ${timer_start}" | bc -l)
echo
echo 'Done!'
echo "Run Time: $(displaytime ${runtime})"

EOF
```
