## get the fragments

```r
get_strand = function(x){ifelse(is.na(x), "+", "-")}
fun_prep   = function(dat, pattern="^[0-9|X|Y]+:[0-9]*-[0-9]*", prefix_chrom = "chr") {
    ### Input:  dataframe
    ### Output: dataframe
    ### Description:
    ###     Preprocess the raw dataframe
    ###     - expand duplicates
    ###     - filter and extract only oligos
    ###     - get location information
    res = dat %>% 
        dplyr::mutate(Name = gsub("\\(|\\)", "", Name)) %>% 
        tidyr::separate_rows(Name, sep=";") %>%
        dplyr::filter(!grepl(pattern = "Nadav", Name)) %>%
        dplyr::filter(str_detect(string=Name, pattern=pattern)) %>%
        tidyr::separate(Name, c("Chrom", "Start", "End", "Strand"), remove=FALSE, fill="right") %>%
        dplyr::mutate(
            Chrom  = paste0(prefix_chrom, Chrom),
            Start  = as.integer(Start),
            End    = as.integer(End),
            Name   = paste0(prefix_chrom, Name),
            Strand = get_strand(Strand))
    return(res)
}
```

```r
lst_dat = list(
    "OL13" = dat_OL13_raw,
    "OL43" = dat_OL43_raw,
    "OL45" = dat_OL45_raw
)

print(lapply(lst_dat, dim))
```

```r
lst_dat[[1]] = fun_prep(lst_dat[[1]])
lst_dat[[2]] = fun_prep(lst_dat[[2]])
lst_dat[[3]] = fun_prep(lst_dat[[3]], pattern = "chr[0-9|X|Y]+:[0-9]*-[0-9]*", prefix_chrom = "")

print(lapply(lst_dat, dim))
```

```r
lapply(lst_dat, head)
```

```r
lst = lapply(lst_dat, function(dat){
    res = dat %>%
        group_by(Chrom) %>%
        summarize(
            Start = min(Start),
            End   = max(End),
            Count = n()
        ) %>%
        mutate(
            Range  = paste0(Chrom, ":", Start, "-", End),
            Length = round((End - Start) / 10^6, 3) 
        )
})
```

```r
dat = rbindlist(lst, idcol="File")
dat
```

```r
gpt = ggplot(dat, aes(x=Chrom, y=Count, fill=File)) + 
    geom_bar(stat="identity") + 
    labs(x="Chromosome") +
    theme_cowplot() +
    coord_flip()
    

options(repr.plot.height=4, repr.plot.width=10)
print(gpt)
```
