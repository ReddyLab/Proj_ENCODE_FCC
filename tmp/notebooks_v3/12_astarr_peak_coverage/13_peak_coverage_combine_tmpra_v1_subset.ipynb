```r
ASSAY  = "Tewhey_K562_TileMPRA"
FOLDER = "coverage_astarrseq_peak_macs_input"
REGIONS=c("GATA1", "MYC", "FADS")
TYPES=c("raw", "norm")

for (REGION in REGIONS) {
    for (TYPE in TYPES) {
        cat("^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^^\n")
        cat("Region: ", REGION, "\n")
        cat("Type:   ", TYPE, "\n")
        cat("++++++++++++++++++++++++++++++++++++++\n")
        
        ### get file paths
        fname  = paste("", TYPE, "rep", REGION, "bed", sep="*")
        fglob  = file.path(FD_RES, "results", ASSAY, FOLDER, fname)
        fpaths = Sys.glob(fglob)
        print(basename(fpaths))
        flush.console()
        
        ### set metadata
        dat_meta = data.frame(
            Sample = sapply(fpaths, get_sample),
            Group  = sapply(fpaths, get_group),
            Region = REGION,
            Type   = TYPE,
            FPath  = fpaths)
        rownames(dat_meta) = sapply(fpaths, get_sample)
        
        ### import count data
        cnames = c("Chrom", "Start", "End", "Count")
        ctypes = c(col_character(), col_integer(), col_integer(), col_double())
        lst = lapply(fpaths, function(fpath){

            dat = read_tsv(fpath, col_names = cnames, col_types = ctypes) %>% 
                dplyr::filter(Count != ".") %>% 
                dplyr::mutate(Count  = as.integer(Count)) %>%
                dplyr::mutate(
                    Peak   = paste(Chrom, Start, End, sep="_"),
                    Type   = TYPE,
                    Region = REGION,
                    Group  = get_group(fpath),
                    Sample = get_sample(fpath))

            return(dat)
        })

        ### Import library size
        fdiry = file.path(FD_RES, "results", ASSAY, "coverage", "summary")
        fname = "library_size_summary.csv"
        fpath = file.path(fdiry, fname)
        dat_lib = read_csv(fpath, show_col_types = FALSE)
        
        ### count matrix
        dat_count = bind_rows(lst) %>% dplyr::select(-Group) %>% spread(Sample, Count)
        
        ### Calculate CPM matrix and mean
        dat = dat_count %>% 
            tidyr::gather(Sample, Count, -Chrom, -Start, -End, -Peak, -Region, -Type) %>%
            dplyr::left_join(dat_lib, by=c("Type", "Region", "Sample")) %>%
            dplyr::mutate(CPM = Count * 1e6 / Size)

        dat_cpm = dat %>% 
           dplyr::select(Chrom, Start, End, Peak, Sample, CPM) %>%
           tidyr::spread(Sample, CPM)

        dat_log2fc = dat %>% 
           dplyr::group_by(Chrom, Start, End, Peak, Group) %>%
           dplyr::summarise(CPM = mean(CPM, na.omit=TRUE), .groups="drop") %>%
           tidyr::spread(Group, CPM) %>%
           dplyr::mutate(log2FoldChange = log2(Output) - log2(Input))
        
        ### store results
        cat("++++++++++++++++++++++++++++++++++++++\n")
        fdiry = file.path(FD_RES, "results", ASSAY, FOLDER, "summary")
        
        fname = paste("metadata", TYPE, REGION, "tsv", sep=".")
        fpath = file.path(fdiry, fname)
        write_tsv(dat_meta, fpath)
        print(fname)
        
        fname = paste("matrix", TYPE, "count", REGION, "tsv", sep=".")
        fpath = file.path(fdiry, fname)
        write_tsv(dat_count, fpath)
        print(fname)
        
        fname = paste("matrix", TYPE, "cpm", REGION, "tsv", sep=".")
        fpath = file.path(fdiry, fname)
        write_tsv(dat_cpm, fpath)
        print(fname)
        
        fname = paste("result", "Log2FC", TYPE, "cpm", REGION, "tsv", sep=".")
        fpath = file.path(fdiry, fname)
        write_tsv(dat_log2fc, fpath)
        print(fname)
    }
}
```
