```r
ASSAY   = "Tewhey_K562_TileMPRA"
REGIONS = c("GATA1", "MYC", "FADS")
TYPE    = "raw"
PROCESS = "deseq"
FOLDER  = "coverage_astarrseq_peak_macs_input"

fdiry = file.path(FD_RES, "results", ASSAY, FOLDER, "summary")
lst = lapply(REGIONS, function(REGION){
    ### set file path
    fname = paste("result", "Log2FC", TYPE, PROCESS, REGION, "tsv", sep=".")
    fpath = file.path(fdiry, fname)
    cat(fpath, "\n")

    ### import data
    dat = read_tsv(fpath) %>% dplyr::mutate(Assay = ASSAY, Region = REGION)
    print(head(dat))
    cat("\n")
    return(dat)
})
names(lst) = REGIONS

lst_log2fc_tmpra = lst
dat_log2fc_tmpra = bind_rows(lst)
```

```r
lst = list(dat_log2fc_astarr, dat_log2fc_wstarr, dat_log2fc_tmpra)
lst = lapply(lst, function(dat){return(dat$Peak)})
peaks = Reduce(intersect, lst)
print(length(peaks))
```

```r
dim(dat_log2fc_tmpra)
```

```r
dat_region = dat_log2fc_tmpra %>% 
    dplyr::select(Peak, Region) %>% 
    dplyr::filter(Peak %in% peaks)

tmp = left_join(dat_region, dat_log2fc_astarr, by="Peak")
lst = split(tmp, tmp$Region)
lst_log2fc_astarr = lst

tmp = left_join(dat_region, dat_log2fc_wstarr, by="Peak")
lst = split(tmp, tmp$Region)
lst_log2fc_wstarr = lst
```
