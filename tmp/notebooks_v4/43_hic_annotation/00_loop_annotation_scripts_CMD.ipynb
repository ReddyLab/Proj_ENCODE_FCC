```bash
cat > run_annot.sh << 'EOF'
#!/bin/bash

### print start message
timer_start=`date +%s`
echo "Hostname:          " $(hostname)
echo "Slurm Array Index: " ${SLURM_ARRAY_TASK_ID}
echo "Time Stamp:        " $(date +"%m-%d-%y+%T")
echo

### setup input and output
source ../config/config_duke.sh
FP_REG=$1
FP_ANT=$2
FP_OUT=$3

### show I/O file
echo "Input: " ${FP_REG}
echo
echo "show first few lines of input"
zcat ${FP_REG} | head -5
echo
echo "Input: " ${FP_ANT}
echo
echo "show first few lines of input"
zcat ${FP_ANT} | head -5
echo

### execute: annotation using intersect
bedtools intersect \
    -a ${FP_REG} \
    -b ${FP_ANT} \
    -wo \
    -f 0.7 \
    -F 0.7 \
    -e \
| gzip -c \
> ${FP_OUT}

### show output file
echo
echo "Output: " ${FP_OUT}
echo
echo "show first few lines of output:"
zcat ${FP_OUT} | head -5
echo

### print end message
timer=`date +%s`
runtime=$(echo "${timer} - ${timer_start}" | bc -l)
echo
echo 'Done!'
echo "Run Time: $(displaytime ${runtime})"

EOF

chmod +x run_annot.sh
```

## Annotate: TSS

```bash
FOLDER_ANT=annotation_tss_pol2
FD_ANT=${FD_RES}/results/region/${FOLDER_ANT}
FP_ANTS=($(ls ${FD_ANT}/*bed.gz))

for FP_ANT in ${FP_ANTS[@]}; do
    FN_ANT=$(basename ${FP_ANT})
    PREFIX_ANT=${FN_ANT%.bed.gz}
    echo ${FN_ANT}
    echo ${PREFIX_ANT}
    echo
    
    for FOLDER_LOOP in "hic_intact_ENCSR479XDG" "hic_intact_deep"; do
        FD_REG=${FD_RES}/results/region/${FOLDER_LOOP}/loop
        FD_OUT=${FD_RES}/results/region/${FOLDER_LOOP}/loop/${FOLDER_ANT}
        echo ${FOLDER_LOOP}
        echo

        for PREFIX_LOOP in "loopA" "loopB"; do
            FN_LOG=annotation.${FOLDER_LOOP}.${PREFIX_LOOP}.${PREFIX_ANT}.%a.txt
            FN_REG=${PREFIX_LOOP}.bed.gz
            FN_OUT=${PREFIX_LOOP}.annotation.${FN_ANT}

            FP_REG=${FD_REG}/${FN_REG}
            FP_OUT=${FD_OUT}/${FN_OUT}
            echo ${FN_LOG}
            echo ${FN_REG}
            echo ${FN_OUT}
            echo
        done
    done
    echo
done
```
