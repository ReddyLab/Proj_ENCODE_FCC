```bash
FD_INP=${FD_RES}/results/*/coverage/track_normalized
FN_INP=*pLog2FC.mean.SUBSET.unstranded.bed.gz
FP_TRACK_STARR=($(ls ${FD_INP}/${FN_INP}))

FD_INP=${FD_RES}/results/Tewhey_K562_TileMPRA/coverage/track_counts
FN_INP=OL*hg38*norm*mean*stranded_pos.bed.gz
FP_TRACK_TMPRA=($(ls ${FD_INP}/${FN_INP}))

for FP_INP in ${FP_TRACK_STARR[@]}; do
    FD_OUT=$(dirname  ${FP_INP})
    
    ### Original
    FN_INP=$(basename ${FP_INP})
    PREFIX=${FN_INP%.bed.gz}
    echo ${FD_OUT} 
    echo ${FN_INP}
    
    ### Input
    FN_OUT=${PREFIX/pLog2FC/Input}.bg
    echo ${FN_OUT}
    
    ### Output
    FN_OUT=${PREFIX/pLog2FC/Output}.bg
    echo ${FN_OUT}
    
    ### pLog2FC
    FN_OUT=${PREFIX}.bg
    echo ${FN_OUT}
    echo
done

for FP_INP in ${FP_TRACK_TMPRA[@]}; do
    FD_INP=$(dirname  ${FP_INP})
    FN_INP=$(basename ${FP_INP})
    PREFIX=${FN_INP%.bed.gz}
    FN_OUT=${PREFIX}.bg
    echo ${FD_INP}
    echo ${FN_INP}
    echo ${FN_OUT}
    echo
done
```

```bash
FD_INP=${FD_RES}/results/*/coverage/track_normalized
FN_INP=*pLog2FC.mean.SUBSET.unstranded.bed.gz
FP_TRACK_STARR=($(ls ${FD_INP}/${FN_INP}))

FD_INP=${FD_RES}/results/Tewhey_K562_TileMPRA/coverage/track_counts
FN_INP=OL*hg38*norm*mean*stranded_pos.bed.gz
FP_TRACK_TMPRA=($(ls ${FD_INP}/${FN_INP}))

for FP_INP in ${FP_TRACK_STARR[@]}; do
    FD_OUT=$(dirname  ${FP_INP})
    echo ${FD_OUT}
     
    ### Original
    FN_INP=$(basename ${FP_INP})
    PREFIX=${FN_INP%.bed.gz}
    echo ${FN_INP}
    
    zcat ${FP_INP} | head -10
    echo
    
    ### Input
    FN_OUT=${PREFIX/pLog2FC/Input}.bg
    FP_OUT=${FD_OUT}/${FN_OUT}
    echo ${FN_OUT}
    
    zcat ${FP_INP} \
    | awk '{print $1 "\t" $2 "\t" $3 "\t" $4}' \
    > ${FP_OUT}
    echo
    
    ### Output
    FN_OUT=${PREFIX/pLog2FC/Output}.bg
    FP_OUT=${FD_OUT}/${FN_OUT}
    echo ${FN_OUT}
    
    zcat ${FP_INP} \
    | awk '{print $1 "\t" $2 "\t" $3 "\t" $5}' \
    > ${FP_OUT}
    echo
    
    ### pLog2FC
    FN_OUT=${PREFIX}.bg
    FP_OUT=${FD_OUT}/${FN_OUT}
    echo ${FN_OUT}
    
    zcat ${FP_INP} \
    | awk '{print $1 "\t" $2 "\t" $3 "\t" $6}' \
    > ${FP_OUT}
    echo
done

for FP_INP in ${FP_TRACK_TMPRA[@]}; do
    FD_OUT=$(dirname  ${FP_INP})
    FN_INP=$(basename ${FP_INP})
    PREFIX=${FN_INP%.bed.gz}
    FN_OUT=${PREFIX}.bg
    FP_OUT=${FD_OUT}/${FN_OUT}
    echo ${FD_OUT}
    echo ${FN_INP}
    echo ${FN_OUT}
    zcat ${FP_INP} > ${FP_OUT}
    echo
done
```
