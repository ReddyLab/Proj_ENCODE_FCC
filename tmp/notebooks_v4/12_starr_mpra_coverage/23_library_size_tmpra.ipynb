## RUN

```bash
sbatch -p ${NODE} \
    --exclude=dl-01 \
    --cpus-per-task 8 \
    --mem 8G \
    --output ${FD_LOG}/coverage_library_size_TMPRA.%a.txt \
    --array 0 \
    <<'EOF'
#!/bin/bash
### set directories & global variables
source ../config/config_duke.sh
ASSAY=Tewhey_K562_TileMPRA

### print start message
timer_start=`date +%s`
echo "Hostname:          " $(hostname)
echo "Slurm Array Index: " ${SLURM_ARRAY_TASK_ID}
echo "Time Stamp:        " $(date +"%m-%d-%y+%T")
echo

### set input and output files
FD_INP=${FD_RES}/results/${ASSAY}/fragment_count
FP_INPS=($(ls ${FD_INP}/OL*hg*{Input,Output}*rep*bed))

FD_OUT=${FD_RES}/results/${ASSAY}/coverage/summary
FN_OUT=library_size_count.csv
FP_OUT=${FD_OUT}/${FN_OUT}

### init
mkdir -p   ${FD_OUT}
echo  -n > ${FP_OUT}

### get library size
for FP_INP in ${FP_INPS[@]}; do
    
    ### show I/O file
    echo ++++++++++++++++++++++++++++++++++++++
    echo "Input: " ${FP_INP}
    echo
    echo "show first few lines of file"
    cat ${FP_INP} | head -n 3
    echo
    
    ### count the total counts and append the results to the output file
    FN_INP=$(basename ${FP_INP})
    cat ${FP_INP} |\
        awk -v OFS=, -v FNAME=${FN_INP} '{sum += $5} END {OFMT="%f";print FNAME, sum}' \
        >> ${FP_OUT}
done

### show I/O file
echo "Output file: ${FP_OUT}"
echo
echo "Show the lines"
cat ${FP_OUT}
echo

### print end message
timer=`date +%s`
runtime=$(echo "${timer} - ${timer_start}" | bc -l)
echo
echo 'Done!'
echo "Run Time: $(displaytime ${runtime})"
EOF
```

Note: `OFMT="%f"` prevent awk from printing scientific notation

ref: [How can I make awk not use scientific notation when printing small values? - Stack Overflow](https://stackoverflow.com/questions/12714155/how-can-i-make-awk-not-use-scientific-notation-when-printing-small-values)
