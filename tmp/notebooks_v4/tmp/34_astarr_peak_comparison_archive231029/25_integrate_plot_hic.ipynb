```r
dat = dat_peak_score_summary_starrmpra_full
dat = dat %>% 
    dplyr::mutate(
        Assay = factor(Assay, levels=VEC_ASSAYS),
        Peak  = factor(Peak,  levels=VEC_PEAKS),
        Score = RankNorm_Total
    )

dat_score_starrmpra = dat
print(dim(dat))
head(dat)
```

```r
### setup theme
theme_fill = scale_fill_gradient2(
    low  = "#4575b4",
    mid  = "#ffffbf",
    high = "#d73027",
    midpoint = 0.5
)
theme_text = THEME_TEXT
theme_legend = theme(legend.position = "none")

### plot and assign
dat = dat_score_starrmpra
gpt = ggplot(dat, aes(x=Peak, y=Assay, fill=Score)) +
    geom_tile() +
    labs(x="Open Chromatin Regions", title = "STARR/TMPRA (Ranked-Normalized Log2FC)") +
    theme_cowplot() + 
    theme_text +
    theme_fill +
    theme_legend
gpt_score_starrmpra = gpt

### show the plot
options(repr.plot.height=3, repr.plot.width=15)
print(gpt)
```

```r
dat = data.frame(Peak = VEC_PEAKS)
dat = dplyr::left_join(
    dat,
    dat_region_pair_hic_score_tss_fads1,
    by="Peak"
)

dat = dat %>% 
    dplyr::select(Peak, Score) %>%
    tidyr::gather(Group, Score, -Peak) %>%
    dplyr::mutate(Peak = factor(Peak,  levels=VEC_PEAKS)) %>%
    replace(is.na(.), 0)

dat_score_hic_tss_fads1 = dat
print(dim(dat))
head(dat)
```

```r
### setup theme
theme_text = THEME_TEXT
theme_legend = theme(legend.position = "none")

### plot and assign
dat = dat_score_hic_tss_fads1
gpt = ggplot(dat, aes(x=Peak, y=Score)) + 
    #geom_bar(stat="identity", position="identity") + 
    geom_col(width=2, position="identity") + 
    labs(title = "Intact Hi-C (FADS1 TSS)") +
    theme_cowplot() + 
    theme_text +
    theme_legend
gpt_score_hic_tss_fads1 = gpt

### show the plot
options(repr.plot.height=3, repr.plot.width=15)
print(gpt)
```
