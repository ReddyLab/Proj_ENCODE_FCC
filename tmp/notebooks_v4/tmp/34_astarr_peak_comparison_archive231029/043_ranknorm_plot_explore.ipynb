## Plot screens

```r
dat = dat_peak_score_starrmpra_crispr
dat = dat %>% 
    dplyr::mutate(
        `ASTARR/WSTARR/TMPRA` = RankNorm, 
        `CRISPRi-HCRFF`       = CRISPRi_HCRFF_Tot, 
        `CRISPRi-Growth`      = CRISPRi_Growth_Tot, 
        `Lenti-MPRA`          = LMPRA_Tot
    ) %>%
    dplyr::select(
        `ASTARR/WSTARR/TMPRA`, 
        `CRISPRi-HCRFF`, 
        `CRISPRi-Growth`, 
        `Lenti-MPRA`
    ) 

lst = lapply(dat, function(vec){sum(!is.na(vec))})
dat = data.frame(Assay=names(lst), Count = unlist(lst))
dat = dat %>% 
    dplyr::arrange(Count) %>%
    dplyr::mutate(Assay=factor(Assay, levels=Assay))


gpt = ggplot(dat, aes(x=Assay, y=Count, label=Count)) +
    geom_col() +
    geom_text(hjust=-0.5) +
    labs(x="Assay", y="#OCR") +
    theme_cowplot() +
    coord_flip() +
    ylim(c(0, 2000))

options(repr.plot.height = 4, repr.plot.width=8)
print(gpt)
```

## Plot scores

**Setup plot parameters**

```r
dat = dat_peak_score_starrmpra_crispr
dat = dat %>% arrange(RankNorm)

VEC_PEAKS  = dat$Peak
VEC_ASSAYS = c("ASTARR", "WSTARR", "TMPRA")
```

```r
THEME_TEXT = theme(
    axis.ticks.x = element_blank(),
    axis.text.x  = element_blank(),
    axis.text.y  = element_text(size = 15),
    axis.title.x = element_blank(),
    axis.title.y = element_blank(),
    plot.title = element_text(face = "bold", size = 17)
    #plot.title = element_text(face = "plain", size = 17)
)
```

```r
dat = dat_peak_summary_starrmpra
dat = dat %>% 
    dplyr::mutate(
        Assay = factor(Assay, levels=VEC_ASSAYS),
        Peak  = factor(Peak,  levels=VEC_PEAKS),
        Score = RankNorm_Total
    )

dat_score_starrmpra = dat
print(dim(dat))
head(dat)
```

```r
### setup theme
theme_fill = scale_fill_gradient2(
    low  = "#4575b4",
    mid  = "#ffffbf",
    high = "#d73027",
    midpoint = 0.5
)
theme_text = THEME_TEXT
theme_legend = theme(legend.position = "none")

### plot and assign
dat = dat_score_starrmpra
gpt = ggplot(dat, aes(x=Peak, y=Assay, fill=Score)) +
    geom_tile() +
    labs(x="Open Chromatin Regions", title = "STARR/TMPRA (Ranked-Normalized Log2FC)") +
    theme_cowplot() + 
    theme_text +
    theme_fill +
    theme_legend
gpt_score_starrmpra = gpt + theme(legend.position = "none")

### show the plot
options(repr.plot.height=3, repr.plot.width=15)
print(gpt)
```
