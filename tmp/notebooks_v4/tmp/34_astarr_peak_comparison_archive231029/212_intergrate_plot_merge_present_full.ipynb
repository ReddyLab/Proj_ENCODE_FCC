```r
### init
lst = list()

###
txt = "Enhancer Calls (Unify Processing Pipeline)"
vec = c("ASTARR_A", "WSTARR_A", "TMPRA_A","LMPRA_A")
vec = paste("enhancer_zscore_junke", vec, sep=".")
lst[[txt]] = vec

#txt = "Enhancer"
#vec = c("ASTARR_AB", "WSTARR_AB", "TMPRA_A", "LMPRA_AB")
#vec = paste("enhancer_zscore_junke", vec, sep=".")
#lst[[txt]] = vec

txt = "CRISPRi"
vec = c("crispri_hcrff_casa", "crispri_growth_dhs", "crispri_e2g")
lst[[txt]] = vec

txt = "TSS (K562; NCBI RefSeq + POLR2A)"
lst[[txt]] = "tss_pol2"

txt = "cCREs (K562; ENCODE V4)"
lst[[txt]] = "ccre_v4"

txt = "ChIP-seq Histone (K562)"
lst[[txt]] = "chipseq_histone"

txt = "ChIP-seq TF (K562)"
lst[[txt]] = "chipseq_tf"

lst_vec_annot_group = lst
print(lst)
```

```r
lst_dat_annot  = list()
lst_dat_enrich = list()

for (idn in names(lst_vec_annot_group)){
    ###
    vec = lst_vec_annot_group[[idn]]
    cat(idn, "\n")
    print(vec)
    cat("\n")
    
    ### get annotation
    lst = lst_dat_annot_full
    vec = lst_vec_annot_group[[idn]]
    dat = bind_rows(lst[vec])
    lst_annot = split(dat, dat$Label)
    txt_annot = unique(dat$Annotation)
    vec_label = unique(dat$Label)
    
    ### create annotation table
    dat = data.frame(
        Peak = VEC_PEAKS, 
        Annotation = txt_annot)
    
    for (idx in names(lst_annot)){
        tmp = lst_annot[[idx]]
        dat[[idx]] = ifelse(
            dat$Peak %in% tmp$Peak, 
            "1", 
            "0")
    }
    dat_annot = dat
    
    ### create enrichment results
    dat = dat_enrich_overall
    dat = dat %>% 
        dplyr::filter(Annotation == txt_annot) %>%
        dplyr::filter(Label %in% vec_label) %>%
        dplyr::arrange(nLogP) 
    dat_enrich = dat
    
    ### arrange results
    dat_enrich = dat_enrich %>%
        dplyr::mutate(
            Label = factor(Label, levels=Label)
        )
    dat_annot  = dat_annot %>% 
        tidyr::gather(Label, Value, -Peak, -Annotation) %>%
        dplyr::mutate(
            Peak  = factor(Peak,  levels=VEC_PEAKS),
            Label = factor(Label, levels=dat_enrich$Label)
        )
    
    ### store results
    lst_dat_annot[[idn]]  = dat_annot
    lst_dat_enrich[[idn]] = dat_enrich
}
```

**Enhancer**

```r
### init
vec = c(
    "ASTARR_AB", "ASTARR_A", "ASTARR_R", 
    "WSTARR_AB", "WSTARR_A", "WSTARR_R", 
                 "TMPRA_A",  "TMPRA_R",
    "LMPRA_AB",  "LMPRA_A",  "LMPRA_R"
)
vec = paste("enhancer_zscore_junke", vec, sep=".")

###
lst = lst_dat_annot
dat = bind_rows(lst[vec])
lst = split(dat, dat$Label)
lst_annot = lst

### init
dat = data.frame(Peak = VEC_PEAKS)

### annotate
for (idx in names(lst_annot)){
    ant = lst_annot[[idx]]
    vec = ant$Peak
    dat[[idx]] = ifelse(dat$Peak %in% vec, "1", "0")
}

### arrange
dat = dat %>% 
    tidyr::gather(Label, Value, -Peak) %>%
    dplyr::mutate(Peak  = factor(Peak,  levels=VEC_PEAKS))

### assign and show
dat_annot_enhancer_starrmpra = dat
print(dim(dat))
head(dat)
```

```r
theme_fill   = scale_fill_manual(values = c("1" = "black", "0" = "white"))
theme_text   = THEME_TEXT
theme_legend = theme(legend.position = "none")

txt = "Enhancer (K562)"
dat = dat_annot_enhancer_starrmpra

gpt = ggplot(dat, aes(x=Peak, y=Label, fill=Value)) +
    geom_tile() +
    labs(title = txt) +
    theme_cowplot() +
    theme_fill +
    theme_text +
    theme_legend

#gpt_annot_ccre_v4 = gpt
options(repr.plot.height=4, repr.plot.width=15)
print(gpt)
```
