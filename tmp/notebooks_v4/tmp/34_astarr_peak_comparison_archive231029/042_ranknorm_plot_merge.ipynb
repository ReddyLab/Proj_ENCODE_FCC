## Set factor for plotting order

```r
dat = dat_peak_score_starrmpra_mean
dat = dat %>% arrange(RankNorm)

VEC_PEAKS  = dat$Peak
VEC_ASSAYS = c("ASTARR", "WSTARR", "TMPRA")
```

## Plot: STARR/MPRA

```r
dat = dat_peak_score_summary_starrmpra
dat = dat %>% 
    dplyr::mutate(
        Assay = factor(Assay, levels=VEC_ASSAYS),
        Peak  = factor(Peak,  levels=VEC_PEAKS),
        Score = RankNorm_Total
    )

dat_score_starrmpra = dat
print(dim(dat))
head(dat)
```

```r
theme_fill = scale_fill_gradient2(
    low  = "#4575b4",
    mid  = "#ffffbf",
    high = "#d73027",
    midpoint = 0.5
)

theme_text = theme(
    axis.ticks.x = element_blank(),
    axis.text.x  = element_blank(),
    axis.text.y  = element_text(size = 15),
    #axis.title.x = element_text(size = 20),
    axis.title.x = element_blank(),
    axis.title.y = element_blank()
)

theme_legend = theme(legend.position = "none")

dat = dat_score_starrmpra

gpt = ggplot(dat, aes(x=Peak, y=Assay, fill=Score)) +
    geom_tile() +
    labs(x="Open Chromatin Regions") +
    theme_cowplot() + 
    theme_text +
    theme_fill

gpt_score_starrmpra = gpt + theme(legend.position = "none")

options(repr.plot.height=3, repr.plot.width=15)
print(gpt)
```

```r
legend = cowplot::get_legend(gpt)
options(repr.plot.height=4, repr.plot.width=3)
grid.draw(legend)
```

## HiC

```r
fdiry = file.path(
    FD_RES, "results", 
    "hic_intact_K562_deep", 
    "coverage_astarrseq_peak_macs_input",
    "region_pair_selected"
)

cnames = c(
    "Chrom1", "Start1", "End1", 
    "Chrom2", "Start2", "End2",
    "Name"
)
cnames = c(
    cnames,
    "Mean", "SD", 
    "Q1", "Q2", "Q3", 
    "Min", "Max", 
    "Density")

fnames = dir(fdiry)
fpaths = file.path(fdiry, fnames)

lst = lapply(fpaths, function(fpath){
    dat = read_tsv(fpath, col_names = cnames, show_col_types = FALSE)
    return(dat)
})
dat = bind_rows(lst)
dat = dat %>%
    tidyr::separate(Name, c("Region1", "Region2"), sep = "\\|") %>%
    dplyr::mutate(Score = Mean) %>%
    dplyr::select(Region1, Region2, Score)

dat_region_pair_hic_score = dat
print(dim(dat))
head(dat)
```

```r
dat = tidyr::expand_grid(Region1 = VEC_PEAKS, Region2 = VEC_PEAKS)
head(dat)
```

```r
#vec = head(VEC_PEAKS, 100)
vec = VEC_PEAKS
dat = tidyr::expand_grid(Region1 = vec, Region2 = vec)
dat = dat %>%
    dplyr::left_join(
        dat_region_pair_hic_score, 
        by=c("Region1", "Region2")) %>%
    dplyr::mutate(Proximal = (Region1 == Region2)) %>%
    dplyr::mutate(
        Region1 = factor(Region1, levels=rev(VEC_PEAKS)),
        Region2 = factor(Region2, levels=VEC_PEAKS)
    )

dat_score = dat
print(dim(dat))
head(dat)
```

```r
theme_fill = scale_fill_gradient2(
    low  = "white", 
    mid  = "#cb181d", 
    high = "#a50f15",
    midpoint = 15,
    na.value = 'white'
)
theme_text = theme(
    axis.ticks = element_blank(),
    axis.text  = element_blank())

dat = dat_score
gpt = ggplot(dat, aes(x=Region1, y=Region2, fill=Score)) +
    geom_tile() +
    theme_cowplot() +
    theme_fill +
    theme_text

options(repr.plot.height=8, repr.plot.width=8.5)
print(gpt)
```
