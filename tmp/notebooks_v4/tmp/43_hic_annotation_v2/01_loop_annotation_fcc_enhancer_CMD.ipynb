## Loop annotations

**Test loop**

```bash
FOLDER_LOOPS=("hic_intact_ENCSR479XDG" "hic_intact_deep")
PREFIX_LOOPS=("Loop_A" "Loop_B")

for FOLDER_LOOP in ${FOLDER_LOOPS[@]}; do
    
    for PREFIX_LOOP in ${PREFIX_LOOPS[@]}; do
    
        ### get loop file path
        FD_REG=${FD_RES}/results/region/${FOLDER_LOOP}
        FP_REG=$(ls ${FD_REG}/*${PREFIX_LOOP}.bed.gz)
        FN_REG=$(basename ${FP_REG})
        
        for FP_ANT in ${FP_ANTS[@]}; do
            ### get annotation file folder and name
            FD_ANT=$(dirname  ${FP_ANT})
            FN_ANT=$(basename ${FP_ANT})
            
            FOLDER_ANT=$(basename ${FD_ANT})
            PREFIX_ANT=${FN_ANT%.bed.gz}
            
            ### set log file path
            FN_LOG=${PREFIX_LOOP}.annotation.${PREFIX_ANT}.txt
            FP_LOG=${FD_LOG}/${FN_LOG}

            ### set output file path
            FD_OUT=${FD_REG}/${FOLDER_ANT}
            FN_OUT=${PREFIX_LOOP}.annotation.${FN_ANT}
            FP_OUT=${FD_OUT}/${FN_OUT}
            
            ### show progress
            echo "File Folder (Loop):      " ${FOLDER_LOOP}
            echo "File Prefix (Loop):      " ${PREFIX_LOOP}
            echo "File Name   (Loop):      " ${FN_REG}
            
            echo "File Folder (Annotation):" ${FOLDER_ANT}
            echo "File Prefix (Annotation):" ${PREFIX_ANT}
            echo "File Name   (Annotation):" ${FN_ANT}
            
            echo "File Folder (Output):    " ${FOLDER_LOOP}/${FOLDER_ANT}
            echo "File Name   (Output):    " ${FN_OUT}
            
            echo "File Path   (Log):       " '${FD_LOG}/'${FN_LOG}
            echo
        done
    done
done
```

**Run**

```bash
FOLDER_LOOPS=("hic_intact_ENCSR479XDG" "hic_intact_deep")
PREFIX_LOOPS=("Loop_A" "Loop_B")

for FOLDER_LOOP in ${FOLDER_LOOPS[@]}; do
    
    for PREFIX_LOOP in ${PREFIX_LOOPS[@]}; do
    
        ### get loop file path
        FD_REG=${FD_RES}/results/region/${FOLDER_LOOP}
        FP_REG=$(ls ${FD_REG}/*${PREFIX_LOOP}.bed.gz)
        FN_REG=$(basename ${FP_REG})
        
        for FP_ANT in ${FP_ANTS[@]}; do
            ### get annotation file folder and name
            FD_ANT=$(dirname  ${FP_ANT})
            FN_ANT=$(basename ${FP_ANT})
            
            FOLDER_ANT=$(basename ${FD_ANT})
            PREFIX_ANT=${FN_ANT%.bed.gz}
            
            ### set log file path
            FN_LOG=${PREFIX_LOOP}.annotation.${PREFIX_ANT}.txt
            FP_LOG=${FD_LOG}/${FN_LOG}

            ### set output file path
            FD_OUT=${FD_REG}/${FOLDER_ANT}
            FN_OUT=${PREFIX_LOOP}.annotation.${FN_ANT}
            FP_OUT=${FD_OUT}/${FN_OUT}
            
            ### show progress
            echo "File Folder (Loop):      " ${FOLDER_LOOP}
            echo "File Prefix (Loop):      " ${PREFIX_LOOP}
            echo "File Name   (Loop):      " ${FN_REG}
            
            echo "File Folder (Annotation):" ${FOLDER_ANT}
            echo "File Prefix (Annotation):" ${PREFIX_ANT}
            echo "File Name   (Annotation):" ${FN_ANT}
            
            echo "File Folder (Output):    " ${FOLDER_LOOP}/${FOLDER_ANT}
            echo "File Name   (Output):    " ${FN_OUT}
            
            echo "File Path   (Log):       " '${FD_LOG}/'${FN_LOG}
            echo
            
            ### execute
            ### Remind: put annotation file first and loop file next
            mkdir  -p ${FD_OUT}
            sbatch -p ${NODE} \
                --exclude=dl-01 \
                --cpus-per-task 4 \
                --mem 4G \
                --output ${FP_LOG} \
                ../scripts/run_bedtools_closest.sh ${FP_ANT} ${FP_REG} ${FP_OUT}
            echo
        done
    done
done
```
