**Store separate bed files of counts (only positive strand)**

````r
lst = lst_dat_read
for (dat in lst){
    print(colnames(dat))
    cat("=============\n")
}
````

````r
fun_sortbed = function(dat){
    ### reset levels in chromosome
    #CHROMS = paste0("chr", c(1:22, "X", "Y"))
    #dat$Chrom = factor(dat$Chrom, levels=CHROMS)
    
    ### sort dataframe by location (Chrom, Start, End)
    dat = dat %>% arrange(Chrom, Start, End)
    return(dat)
}
````

````r
assay  = "Tewhey_K562_TileMPRA"
folder = "fragment_count"
strand = "stranded_pos"


lst_dat = lst_dat_read
for (sid in names(lst_dat)){
    ###
    dat = lst_dat[[sid]]
    dat = dat %>% 
        dplyr::filter(Strand == "+") %>% 
        gather(Sample, Score, -Chrom, -Start, -End, -Name, -Strand)
    lst = split(dat, dat$Sample)
    lst = lapply(lst, fun_sortbed)
    
    ###
    for (sam in names(lst)){
        ###
        dat = lst[[sam]]
        dat = dat %>% 
            dplyr::select(Chrom, Start, End, Name, Score, Strand)%>%
            dplyr::filter(!is.infinite(Score)) %>% 
            dplyr::mutate(Score = round(Score, 3))

        ###
        fdiry = file.path(FD_RES, "results", assay, folder)
        fname = paste(sid, sam, strand, "bed", sep=".")
        fpath = file.path(fdiry, fname)
        write_tsv(dat, fpath, col_names=FALSE)

        cat("\n++++++++++++++++++++++++++++++++++++++++++\n")
        cat(fname, "\n")
        print(head(dat))
        flush.console()
    }
}
````
