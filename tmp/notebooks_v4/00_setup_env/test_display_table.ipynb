```r
library(tidyverse)
library(kableExtra)
library(IRdisplay)
```

```r
num = 20
dat_tidy = data.frame(x = LETTERS[1:num], y = rnorm(num))
dat_wide = dat_tidy %>% tidyr::spread(x, y)
```

```r
head(dat_tidy)
```

```r
#options(repr.matrix.max.cols=10)
IRdisplay::display(head(dat_wide))
```

```r
dat = dat_tidy
dat %>% 
    kbl() %>% 
    kable_styling() %>%
    scroll_box(width = "500px", height = "200px") %>%
    as.character() %>%
    display_html()
```

```r
dat = dat_wide
dat |> 
    kbl() |>
    kable_styling() |>
    scroll_box(width = "500px", height = "200px") |>
    as.character() |>
    display_html()
```

```r
dat = dat_wide
dat %>% 
    kable() %>% 
    kable_styling() %>%
    scroll_box(width = "100%", height = "200px") %>%
    as.character() %>%
    display_html()
```

```r
dat = dat_wide
dat %>% kable("html") %>% kable_styling("striped") %>% scroll_box(width = "100%") %>%
    as.character() %>%
    display_html()
```

```r
fun_display_table = function(dat) {
    dat |> 
        kableExtra::kable("html") |> 
        kableExtra::kable_styling("striped") %>% 
        kableExtra::scroll_box(width = "100%") %>%
        as.character() %>%
        IRdisplay::display_html()    
}
```

```r
fun_display(dat_wide)
```
