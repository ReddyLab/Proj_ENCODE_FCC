```r
vec_dst = seq(0, 500)
vec_dst = vec_dst * 1000
print(length(vec_dst))
vec_dst
```

```r
dat = tribble(
    ~Name, ~Description,
    "Chrom", "Chromosome",
    "Start", "Start position",
    "End",   "End position",
    "Name",  "Name",
    "Score", "Pol2 score",
    "Gene",  "Gene of the TSS")

fdiry = file.path(FD_RES, "results", "region", "annotation_tss_pol2_span")
fname = "description.tsv"
fpath = file.path(fdiry, fname)

write_tsv(dat, fpath)
```

```r
dat = dat_tss_pol2
for(idx in 1:(length(vec_dst)-1)){
    ### 
    dst1 = vec_dst[idx]
    dst2 = vec_dst[idx+1]
    
    ###
    lab = paste0(as.integer(dst1/1000), "-", as.integer(dst2/1000), "K", ":", "+")
    cat(lab, "\n")
    
    dat_pos = dat %>% 
        dplyr::mutate(
            Start = Pos + dst1,
            End   = Pos + dst2,
            Name  = paste0(Loc, ":", lab)
        )
    
    ###
    lab = paste0(as.integer(dst1/1000), "-", as.integer(dst2/1000), "K", ":", "-")
    cat(lab, "\n");
    dat_neg = dat %>%
        dplyr::mutate(
            Start = Pos - dst2,
            End   = Pos - dst1,
            Name  = paste0(Loc, ":", lab)
        )
    
    ### combine both directions
    dat_both = bind_rows(dat_pos, dat_neg) %>% 
        dplyr::filter(Start > 0, End > 0)  %>%
        dplyr::select(Chrom, Start, End, Name, Score, Gene) %>%
        dplyr::arrange(Chrom, Start, End)
    
    ### save the results
    lab   = paste0(as.integer(dst1/1000), "_", as.integer(dst2/1000), "K")
    fdiry = file.path(FD_RES, "results", "region", "annotation_tss_pol2_span")
    fname = paste("K562.TSS.highest_Pol2", lab , "bed", "gz", sep = ".")
    fpath = file.path(fdiry, fname)
    
    cat(fname, "\n")
    #print(head(dat_both))
    write_tsv(dat_both, fpath, col_names = FALSE)
    cat("\n")
    flush.console()
}
```
