## Plot

```r
### init
dat = dat_coverage_crispri_hcrff_import
vec = dat$Score

### stats
val_mu = mean(vec)
val_sd = sd(vec)
vec_sd = c(val_sd, -val_sd)
res    = summary(vec)
cat("",
    "Max:    ", res['Max.'],    "\n",
    "3rd Qu.:", res['3rd Qu.'], "\n",
    "Mean:   ", res['Mean'],    "\n",
    "Median: ", res['Median'],  "\n",
    "1st Qu.:", res['1st Qu.'], "\n",
    "Min:    ", res['Min.'],    "\n",
    "Std:    ", val_sd,         "\n"
)

### visualize
dat = dat %>% dplyr::filter(abs(Score) < 5)
gpt = ggplot(dat, aes(x=Score)) + 
    geom_histogram(binwidth = 0.1, color="grey90", fill="grey30") +
    geom_vline(xintercept = val_mu,       linetype="solid",  color="black") +    
    geom_vline(xintercept = vec_sd * 1.0, linetype="dashed", color="red") +
    geom_vline(xintercept = vec_sd * 1.5, linetype="dashed", color="blue") +
    geom_vline(xintercept = vec_sd * 2.0, linetype="dashed", color="chartreuse4") +
    labs(title="CRISPRi-HCRFF (Z Score)") +
    theme_cowplot()

options(repr.plot.height=5, repr.plot.width=8)
print(gpt)
```
