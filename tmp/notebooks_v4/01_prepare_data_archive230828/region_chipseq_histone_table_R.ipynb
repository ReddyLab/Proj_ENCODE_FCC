## Import data

```r
fdiry = file.path(FD_RES, "source", "annotation_chipseq_histone")
fname = "metadata.tsv"
fpath = file.path(fdiry, fname)

dat = read_tsv(fpath)

dat_chipseq_import = dat
print(dim(dat))
head(dat)
```

```r
dat = dat_chipseq_import
colnames(dat)
```

## bigwig file

```r
dat = dat_chipseq_import
dat = dat %>% 
    dplyr::filter(`File format` == "bigWig") %>% 
    dplyr::mutate(`Experiment target` = str_remove(`Experiment target`, "-human")) %>%
    dplyr::select(
        `Experiment accession`,
        `File accession`,
        `Experiment target`,
        `Output type`,
        `Assay`,
        `Biosample term name`,
        `File analysis title`,
        `File assembly`,
        `Experiment date released`,
        `md5sum`,
        `File download URL`
    ) 
dat_bw = dat
print(dim(dat))
dat
```

```r
targets = c(
    "H3K27ac",
    "H3K27me3",
    "H3K4me1",
    "H3K4me2",
    "H3K4me3"
)
dat = dat_bw
dat = dat %>% dplyr::filter(Target %in% targets)
dat
```

## Bed peak file

```r
dat = dat_chipseq_import
dat = dat %>% 
    dplyr::filter(`File format` == "bed narrowPeak") %>% 
    dplyr::mutate(`Experiment target` = str_remove(`Experiment target`, "-human")) %>%
    dplyr::select(
        `Experiment accession`,
        `File accession`,
        `Experiment target`,
        `Output type`,
        `Assay`,
        `Biosample term name`,
        `File analysis title`,
        `File assembly`,
        `Experiment date released`,
        `md5sum`,
        `File download URL`
    ) 

colnames(dat) = c(
    "Experiment_Accession",
    "File_Accession",
    "Target",
    "Output",
    "Assay",
    "Biosample",
    "File_Analysis_Title",
    "File_Assembly",
    "Experiment_Data_Released",
    "md5sum",
    "URL"
)

dat_chipseq_bed = dat
print(dim(dat))
dat
```

```r
dat = dat_chipseq_bed
dat = dat %>%
    dplyr::mutate(
        FN_OUT = paste(
            "K562", 
            Experiment_Accession, 
            File_Accession,
            Target,
            "bed.gz",
            sep = ".")) %>%
    dplyr::mutate(
        CMD = paste(
            "wget",
            URL,
            "-O",
            FN_OUT))

dat_chipseq_bed_cmd = dat
print(dim(dat))
dat
```

```r
fdiry = file.path(FD_RES, "results", "region", "annotation_chipseq_histone")
fname = "accession_info_peaks.tsv"
fpath = file.path(fdiry, fname)

dat = dat_chipseq_bed_cmd
write_tsv(dat, fpath)
```
