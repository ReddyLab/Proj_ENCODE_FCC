## Import data

**ASTARR**

```r
ASSAY="ATAC_STARR_deep"
FOLDER="deep_ATAC_STARR_seq.full_overlap_100bp_bin_10bp_stepSize"
FN_INP="orientation_independent_up_regulated_peaks_union.bed.gz"

### set file path
### FP_INP=${FD_INP}/${ASSAY}/${FOLDER}/processed/merged_peak/${FN_INP}
fdiry = file.path(
    FD_RES,
    "source",
    "annotation_enhancer_junke",
    ASSAY,
    FOLDER,
    "processed",
    "merged_peak"
)
fname = FN_INP
fpath = file.path(fdiry, fname)

cnames = c(
    "Chrom",
    "Start",
    "End",
    "Score",
    "Score_across_all_the_bins",
    "Size"
)
dat = read_tsv(fpath, col_names = cnames)

dat_peak_astarr_import = dat
print(dim(dat))
head(dat, 3)
```

**WSTARR**

```r
ASSAY="WHG_STARR_seq_TR"
FOLDER="WHG_STARR_seq_TR.full_overlap_100_bp_bin_10_bp_stepSize"
FN_INP="orientation_independent_up_regulated_peaks_union.bed.gz"

### set file path
### FP_INP=${FD_INP}/${ASSAY}/${FOLDER}/processed/merged_peak/${FN_INP}
fdiry = file.path(
    FD_RES,
    "source",
    "annotation_enhancer_junke",
    ASSAY,
    FOLDER,
    "processed",
    "merged_peak"
)
fname = FN_INP
fpath = file.path(fdiry, fname)

cnames = c(
    "Chrom",
    "Start",
    "End",
    "Score",
    "Score_across_all_the_bins",
    "Size"
)
dat = read_tsv(fpath, col_names = cnames)

dat_peak_wstarr_import = dat
print(dim(dat))
head(dat, 3)
```

## Check

**ASTARR**

```r
dat = dat_peak_astarr_import
dat = dat %>% 
    dplyr::select(Score, Score_across_all_the_bins) %>%
    head %>%
    dplyr::rowwise() %>%
    dplyr::mutate(Mean = mean(as.double(unlist(str_split(Score_across_all_the_bins, ","))))) %>%
    dplyr::select(-Score_across_all_the_bins)
dat
```

```r
dat = dat_peak_astarr_import
dat = dat %>% 
    dplyr::select(Start, End, Size) %>%
    head %>%
    dplyr::mutate(Length = End - Start)
dat
```

**WSTARR**

```r
dat = dat_peak_wstarr_import
dat = dat %>% 
    dplyr::select(Score, Score_across_all_the_bins) %>%
    head %>%
    dplyr::rowwise() %>%
    dplyr::mutate(Mean = mean(as.double(unlist(str_split(Score_across_all_the_bins, ","))))) %>%
    dplyr::select(-Score_across_all_the_bins)
dat
```

```r
dat = dat_peak_wstarr_import
dat = dat %>% 
    dplyr::select(Start, End, Size) %>%
    head %>%
    dplyr::mutate(Length = End - Start)
dat
```

## Arrange

```r
dat = dat_peak_astarr_import
dat = dat %>% dplyr::select(-Score_across_all_the_bins, -Size)

dat_peak_astarr = dat
print(dim(dat))
head(dat)
```

```r
dat = dat_peak_wstarr_import
dat = dat %>% dplyr::select(-Score_across_all_the_bins, -Size)

dat_peak_wstarr = dat
print(dim(dat))
head(dat)
```
