```r
### set file path
fdiry = file.path(FD_RES, "source", "CRISPRi_FlowFISH")
fname = "K562_HCRFF_20genes_TSS_TES_coords.out"
fpath = file.path(fdiry, fname)

### read file
dat = read_tsv(fpath, col_names = TRUE, show_col_types = FALSE)

### assign and show info
dat_crispr_gene_coord = dat
print(dim(dat))
dat
```

```r
dat = dat_crispr_gene_coord
dat = dat %>% 
    dplyr::mutate(
        TSS = ifelse(Strand == "+", Start, End),
        TES = ifelse(Strand == "+", End, Start),
        Coord = paste0(Chrom, ":", Start, "-", End)
    ) %>%
    dplyr::select(Chrom, Start, End, Strand, TSS, TES, Coord, Gene)

dat_crispr_gene_coord_prep = dat
print(dim(dat))
dat
```

```r
### set file path
fdiry = file.path(FD_RES, "results", "region", "annotation_crispri_hcrff")
fname = "K562_HCRFF_20genes_TSS_TES_coords.tsv"
fpath = file.path(fdiry, fname)

### read file
write_tsv(dat_crispr_gene_coord_prep, fpath)
```
