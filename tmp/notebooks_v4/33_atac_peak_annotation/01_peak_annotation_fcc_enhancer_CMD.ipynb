**Get data files**

```bash
### STARR/MPRA: Enhancer calls from Junke
FOLDER=annotation_enhancer_zscore_junke
FD_ANT=${FD_RES}/results/region/${FOLDER}
FP_ANTS_STARRMPRA=($(ls ${FD_ANT}/*peak*bed.gz))

echo ${FOLDER}
basename -a ${FP_ANTS_STARRMPRA[@]}
echo

### CRISPRi-HCRFF
FOLDER=annotation_crispri_hcrff
FD_ANT=${FD_RES}/results/region/${FOLDER}
FP_ANTS_CRISPRi_HCRFF=($(ls ${FD_ANT}/*.casa.bed.gz))

echo ${FOLDER}
basename -a ${FP_ANTS_CRISPRi_HCRFF[@]}
echo

### CRISPRi-Growth
FOLDER=annotation_crispri_growth
FD_ANT=${FD_RES}/results/region/${FOLDER}
FP_ANTS_CRISPRi_GROWTH=($(ls ${FD_ANT}/*.active.bed.gz))

echo ${FOLDER}
basename -a ${FP_ANTS_CRISPRi_GROWTH[@]}
echo

### ENCODE-E2G benchmark
FOLDER=annotation_encode_e2g_benchmark
FD_ANT=${FD_RES}/results/region/${FOLDER}
#FP_ANTS_ENCODE_E2G_BENCHMARK=($(ls ${FD_ANT}/*.{active,total}.merge.bed.gz))
FP_ANTS_ENCODE_E2G_BENCHMARK=($(ls ${FD_ANT}/*.{active,total}.bed.gz))

echo ${FOLDER}
basename -a ${FP_ANTS_ENCODE_E2G_BENCHMARK[@]}
echo

### ENCODE-E2G prediction
FOLDER=annotation_encode_e2g_prediction
FD_ANT=${FD_RES}/results/region/${FOLDER}
#FP_ANTS_ENCODE_E2G_PREDICTION=($(ls ${FD_ANT}/*.{active,total}.merge.bed.gz))
FP_ANTS_ENCODE_E2G_PREDICTION=($(ls ${FD_ANT}/*.{active,total}.bed.gz))

echo ${FOLDER}
basename -a ${FP_ANTS_ENCODE_E2G_PREDICTION[@]}
echo
```

**Test loop**

```bash
### combine the file lists
FP_ANTS=(
    ${FP_ANTS_STARRMPRA[@]}
    ${FP_ANTS_CRISPRi_HCRFF[@]}
    ${FP_ANTS_CRISPRi_GROWTH[@]}
    ${FP_ANTS_ENCODE_E2G_BENCHMARK[@]}
    ${FP_ANTS_ENCODE_E2G_PREDICTION[@]}
)

for FP_ANT in ${FP_ANTS[@]}; do
    FD_ANT=$(dirname  ${FP_ANT})
    FOLDER=$(basename ${FD_ANT})
    
    FN_ANT=$(basename ${FP_ANT})
    PREFIX=${FN_ANT%.bed.gz}
    
    FN_LOG=peak.annotation.${PREFIX}.txt
    FP_LOG='${FD_LOG}/'peak.annotation.${PREFIX}.txt
    
    FD_OUT=${FD_REG}/${FOLDER}
    FN_OUT=peak.annotation.${FN_ANT}
    FP_OUT=${FD_OUT}/${FN_OUT}
    
    echo "File Folder (Annotation):" ${FOLDER}
    echo "File Prefix (Annotation):" ${PREFIX}
    echo "File Name   (Annotation):" ${FN_ANT}
    echo "File Name   (Output):    " ${FN_OUT}
    echo "File Path   (Log):       " '${FD_LOG}/'${FN_LOG}
    echo
done
```

**Execute**

```bash
### combine the file lists
FP_ANTS=(
    ${FP_ANTS_STARRMPRA[@]}
    ${FP_ANTS_CRISPRi_HCRFF[@]}
    ${FP_ANTS_CRISPRi_GROWTH[@]}
    ${FP_ANTS_ENCODE_E2G_BENCHMARK[@]}
    ${FP_ANTS_ENCODE_E2G_PREDICTION[@]}
)

for FP_ANT in ${FP_ANTS[@]}; do
    FD_ANT=$(dirname  ${FP_ANT})
    FOLDER=$(basename ${FD_ANT})
    
    FN_ANT=$(basename ${FP_ANT})
    PREFIX=${FN_ANT%.bed.gz}
    
    FN_LOG=peak.annotation.${PREFIX}.txt
    FP_LOG=${FD_LOG}/${FN_LOG}
    
    FD_OUT=${FD_REG}/${FOLDER}
    FN_OUT=peak.annotation.${FN_ANT}
    FP_OUT=${FD_OUT}/${FN_OUT}
    
    echo "File Folder (Annotation):" ${FOLDER}
    echo "File Prefix (Annotation):" ${PREFIX}
    echo "File Name   (Annotation):" ${FN_ANT}
    echo "File Name   (Output):    " ${FN_OUT}
    echo "File Path   (Log):       " '${FD_LOG}/'${FN_LOG}
    echo
    sbatch -p ${NODE} \
        --exclude=dl-01 \
        --cpus-per-task 4 \
        --mem 4G \
        --output ${FP_LOG} \
        ../scripts/run_bedtools_annot.sh ${FP_REG} ${FP_ANT} ${FP_OUT}
    echo
done
```
