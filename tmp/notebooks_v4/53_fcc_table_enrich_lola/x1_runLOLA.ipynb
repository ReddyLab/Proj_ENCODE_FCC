## Arrange regions files

**Split by assay type**

````r
lst_dat_peak_fcc_active = dat_peak_fcc_active %>% split(., .$Assay_Type)
lst_dat_peak_fcc_total  = dat_peak_fcc_total  %>% split(., .$Assay_Type)
````

**Total regions**

````r
### init
lst = lst_dat_peak_fcc_total

### create GRangeList for assays
lst = lapply(lst, function(dat) {
    grg = makeGRangesFromDataFrame(
        dat,
        #keep.extra.columns=FALSE,
        ignore.strand  = TRUE,
        seqinfo        = NULL,
        seqnames.field = "Chrom",
        start.field    = "Start",
        end.field      = "End",
        starts.in.df.are.0based=TRUE)
    return(grg)
})
lst = GRangesList(lst)

### assign and show
lst_grg_peak_fcc_total = lst
````

````r
lst = lst_dat_peak_fcc_total
dat = lst[[1]]

print(names(lst))
head(dat, 3)
````

````r
lst = lst_grg_peak_fcc_total
grg = lst[[1]]

print(names(lst))
print(grg)
````

**Significant regions**

````r
### init
lst = lst_dat_peak_fcc_active
vec = names(lst)
names(vec) = vec
vec_txt_assay = vec

### create a list of "GRangeList" for assays
lst = lapply(vec_txt_assay, function(txt_assay){

    ### filter
    dat = lst_dat_peak_fcc_total[[txt_assay]]
    vec = dat$Peak

    dat = lst_dat_peak_fcc_active[[txt_assay]]
    dat = dat %>% dplyr::filter(Peak %in% vec) %>% dplyr::distinct()
    
    ###
    lst_inner = split(dat, dat$Assay_Label)
    lst_inner = lapply(lst_inner, function(dat){
        grg = makeGRangesFromDataFrame(
            dat,
            ignore.strand  = TRUE,
            seqinfo        = NULL,
            seqnames.field = "Chrom",
            start.field    = "Start",
            end.field      = "End",
            starts.in.df.are.0based=TRUE)
        return(grg)
    })

    ###
    #txt_label = paste(txt_assay, "A", sep="_")
    #lst_inner = lst_inner[txt_label]
    
    ###
    lst_inner = GRangesList(lst_inner)
    return(lst_inner)
})

### assign and show
lst_grg_peak_fcc_active = lst
print(names(lst))
````

````r
lst_grg_peak_fcc_active[["ASTARR"]]
````

## Run the enrichment

````r
### init
lst = lst_dat_peak_fcc_active
vec = names(lst)
names(vec) = vec
vec_txt_assay = vec

### run LOLA for each assay
lst = lapply(vec_txt_assay, function(txt_assay){
    ###
    grg_peak_total = lst_grg_peak_fcc_total[[txt_assay]]
    grg_peak_test  = lst_grg_peak_fcc_active[[txt_assay]]

    ###
    cat("Run LOLA on assay:", txt_assay, "\n")
    res = runLOLA(
        grg_peak_test,
        grg_peak_total,
        regionDB
    )

    return(res)
})

### assign
lst_dat_lola_results = lst
````

````r
dat = lst[[1]]
vec = colnames(dat)
print(vec)
head(dat, 3)
````

````r
lst = lst_dat_lola_results
names(lst)
````

````r
lst = lst_dat_lola_results
dat = bind_rows(lst, .id = "Assay_Type")
vec_txt_assay = c(
    'ASTARR', 
    'WSTARR',
    'TMPRA',
    'LMPRA',
    'CRISPRi-Growth',
    'CRISPRi-HCRFF')

dat = dat %>% 
    dplyr::mutate(Assay_Label = paste(Assay_Type, "A", sep="_")) %>%
    dplyr::filter(Assay_Label == userSet) %>%
    dplyr::filter(Assay_Type %in% vec_txt_assay) %>%
    dplyr::select(Assay_Type, userSet, antibody, oddsRatio, pValueLog, maxRnk, meanRnk)

dat_lola_results = dat
print(dim(dat))
head(dat)
````

````r
dat = dat_lola_results
dat = dat %>% dplyr::mutate(pValue = 10^(-pValueLog)) 

lst = split(dat, dat$Assay_Type)
lst = lapply(names(lst), function(txt){
    dat = lst[[txt]]
    gpt = ggplot(dat, aes(x=pValue)) +
        geom_histogram(bins = 30) +
        labs(title=txt) +
        theme_cowplot()
    return(gpt)
})
````

````r
options(repr.plot.height=5, repr.plot.width=15)
plot_grid(plotlist = lst, nrow = 2, ncol = 3)
````

````r
dat = dat_lola_results
dat = dat %>% dplyr::mutate(pValue = 10^(-pValueLog))
lst = split(dat, dat$userSet)
lst = lapply(lst, function(dat){
    vec_num_pvalues  = dat$pValue
    dat$pValue_Adj_BH   = p.adjust(vec_num_pvalues, method="BH")
    dat$pValue_Adj_Bonf = p.adjust(vec_num_pvalues, method="bonferroni")
    return(dat)
})

dat = bind_rows(lst)
dat = dat %>% 
    dplyr::mutate(
        pValueLog_Adj_BH   = -log10(pValue_Adj_BH),
        pValueLog_Adj_Bonf = -log10(pValue_Adj_Bonf)
    ) %>%
    dplyr::select(-pValue, -pValue_Adj_BH, -pValue_Adj_Bonf)

dat_lola_results_padj = dat
print(dim(dat))
head(dat)
````

````r
dat = dat_lola_results_padj

dat = dat %>% 
    dplyr::filter(pValueLog_Adj_Bonf >= 10) %>% 
    dplyr::group_by(Assay_Type, userSet, antibody) %>%
    dplyr::summarise(
        oddsRatio = mean(oddsRatio), 
        maxRnk	  = max(maxRnk),
        meanRnk   = mean(meanRnk),
        .groups = "drop")

dat_lola_results_filter = dat
    
print(dim(dat))
print(length(unique(dat$antibody)))
head(dat)
````
