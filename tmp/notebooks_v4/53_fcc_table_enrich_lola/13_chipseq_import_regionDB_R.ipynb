## Remove Cached files to generate a new one

**Remove cached files**

```r
fdiry = file.path(
    FD_RES, 
    "results", 
    "region", 
    "KS91_K562_ASTARRseq_peak_macs_input",
    "enrichment_lola",
    FOLDER_REGION_COLLECT
)
fname = "*RData"
fglob = file.path(fdiry, "*", fname)
fpath = Sys.glob(fglob)

if (length(fpath) != 0) {
    cat("Delete cached files:", "\n")
    txt = paste("rm", fpath)
    system(txt)
    print(fpath)
} else {
    cat("No cached file found.", "\n")
}
```
