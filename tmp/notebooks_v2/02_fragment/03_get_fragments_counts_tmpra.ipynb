**Region: FADS**

```bash
FDIRY=${FD_RES}/source/MPRA/syn22348180
FNAME=FADS_Tile_K562_K562_20200905.hg38.txt
FP_INP=${FDIRY}/${FNAME}

### postive strand
FDIRY=${FD_RES}/Tewhey_K562_TileMPRA/fragment
FNAME=Tile_K562_hg38_20200905.FADS.stranded_pos.bed
FP_OUT=${FDIRY}/${FNAME}

cat ${FP_INP} |\
    tail -n +2 |\
    awk -v OFS='\t' {'if ($NF == "Tiles" && $6 == "+") {print}'} |\
    cut -f 1-6 > ${FP_OUT}
head ${FP_OUT}

echo ++++++++++++++++++++++++++++++++++++++++++++++++++++

### negative strand
FDIRY=${FD_RES}/Tewhey_K562_TileMPRA/fragment
FNAME=Tile_K562_hg38_20200905.FADS.stranded_neg.bed
FP_OUT=${FDIRY}/${FNAME}

cat ${FP_INP} |\
    tail -n +2 |\
    awk -v OFS='\t' {'if ($NF == "Tiles" && $6 == "-") {print}'} |\
    cut -f 1-6 > ${FP_OUT}
head ${FP_OUT}
```

## Extract counts

```bash
FD_OUT=${FD_RES}/Tewhey_K562_TileMPRA/fragment_count 
mkdir -p ${FD_OUT}
```

**Region: GATA1**

```bash
FDIRY=${FD_RES}/source/MPRA/syn26435687
FNAME=GATA-MYC_Tile_K562_K562_20210130.hg38.cleaned.GATA1.bed
FP_INP=${FDIRY}/${FNAME}

FNAME=Tile_K562_hg38_20210130.GATA1.stranded_pos.bed
FP_OUT=${FD_OUT}/${FNAME}

echo "Number of rows:     " $(cat ${FP_INP} | wc -l)
echo "Number of columns:  " $(awk '{print NF}' ${FP_INP} | sort -nu | tail -n 1)
echo "Number of Input 0:  " $(cat ${FP_INP} | awk -v OFS='\t' {'if ($8 == "0") {print}'} | wc -l)
echo "Number of Value NA :" $(cat ${FP_INP} | awk -v OFS='\t' {'if ($7 == "NA") {print}'} | wc -l)
echo ++++++++++++++++++++++++++++++++++++++++++++
cat ${FP_INP} | awk -v OFS='\t' {'if ($8 != "0") {print}'} > ${FP_OUT}
head ${FP_OUT}
```
