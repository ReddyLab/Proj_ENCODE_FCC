```r
### set globals
REGIONS = c("GATA1", "MYC") 

SAMPLES = c(
    paste0("Input_rep", 1:6),
    paste0("Output_rep", 1:4))

GROUPS = c(
    rep("Input", 6),
    rep("Output", 4))

### file directories
prefix = "KS91_K562_ASTARRseq"
fdiry  = file.path(FD_RES, prefix, "coverage")

### set columns
cnames = c("Chrom", "Start", "End", "Count")

### set columns
lst = lapply(REGIONS, function(region){
    fname = paste0("*", region, "*total_count*")
    fglob = file.path(fdiry, fname)
    fpath = Sys.glob(fglob)
    dat   = read_tsv(fpath, col_names=cnames)
    dat$Region = region
    dat$Sample = SAMPLES
    dat$Group  = GROUPS
    dat$Fpath  = fpath
    return(dat)
})
lst
```

```r
dat = bind_rows(lst)
dat
```

```r
fdiry = file.path(FD_RES, prefix, "coverage", "library_size")
fname = "region_total_counts.csv"
fpath = file.path(fdiry, fname)

write_csv(dat, fpath)
```
