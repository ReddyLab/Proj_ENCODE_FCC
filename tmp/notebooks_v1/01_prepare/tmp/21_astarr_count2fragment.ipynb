**Run**

````bash
sbatch -p ${NODE} \
    --mem 8G \
    --array 1-6 \
    -o ${FD_LOG}/ASTARR_count2fragment_input_dedups.%a.txt \
    <<'EOF'
#!/bin/bash
### set directories & global variables
source ../config_duke.sh
REPLICATE=rep${SLURM_ARRAY_TASK_ID}

REGIONS=(GATA1 MYC)
SAMPLE=Input
STRAND=unstranded

FD_INP=${FD_RES}/source/starrseq
FD_OUT=${FD_RES}/KS91_K562_ASTARRseq/fragment

### print start message
timer_start=`date +%s`
echo "Slurm Array Index: " ${SLURM_ARRAY_TASK_ID}
echo "Time Stamp:        " $(date +"%m-%d-%y+%T")
echo

### loop through count files and convert counts to fragments
for REGION in ${REGIONS[@]}; do

    ### set input/output files
    FP_INP=$(ls ${FD_INP}/KS91_K562_hg38_ASTARRseq_${SAMPLE}_${REPLICATE}*${REGION}*${STRAND}.bed)
    FN_INP=$(basename ${FP_INP})

    FN_OUT=KS91_K562_hg38_ASTARRseq_${SAMPLE}_${REPLICATE}.${REGION}.${STRAND}.dedups.bed
    FP_OUT=${FD_OUT}/${FN_OUT}

    ### print input/output message
    echo "Input  file:" ${FP_INP}
    echo "Output file:" ${FP_OUT}

    ### show input file
    echo "Show input file:"
    head -n 5 ${FP_INP}
    echo

    ### convert counts to fragments
    cat ${FP_INP} | awk -v OFS='\t' {'print $1, $2, $3, $4, 1, $6'} > ${FP_OUT}

    ### show output file
    echo "Show output file:"
    head -n 5 ${FP_OUT}
    echo
done

### print end message
timer=`date +%s`
runtime=$(echo "${timer} - ${timer_start}" | bc -l)
echo
echo 'Done!'
echo "Run Time: $(displaytime ${runtime})"

EOF
````
