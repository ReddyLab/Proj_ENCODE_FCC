```r
dat = dat_astarr
gpt = ggplot(dat, aes(x=Loc, y=log2((Output+1)/(Input+1)))) + 
    geom_point(alpha = 0.1, size=0.5) + 
    theme_cowplot() + 
    labs(x="Location", y="GC Contenct(%)", title="ASTARR")
options(repr.plot.width=10, repr.plot.height=5)
print(gpt)
```

```r
dat = dat_astarr
gpt = ggplot(dat, aes(x=(Start+End)/2, y=Pct_GC)) + 
    geom_point(alpha = 0.1, size=0.5) + 
    theme_cowplot() + 
    labs(x="Location", y="GC Contenct(%)", title="ASTARR")
options(repr.plot.width=10, repr.plot.height=5)
print(gpt)
```

```r
dat = dat_astarr
gpt = ggplot(dat, aes(x=Start, y=End - Start)) + 
    geom_point(alpha = 0.1, size=0.5) + 
    theme_cowplot() + 
    labs(x="Location", y="Length", title="ASTARR")
options(repr.plot.width=10, repr.plot.height=5)
print(gpt)
```

```r
dat = dat_astarr
gpt = ggplot(dat, aes(x=Pct_GC, y=End - Start)) + 
    geom_point(alpha = 0.1, size=0.5) + 
    scale_y_log10() +
    theme_cowplot() + 
    labs(x="GC Contenct(%)", y="Length", title="ASTARR")
options(repr.plot.width=8, repr.plot.height=5)
print(gpt)
```

```r
dat = dat_astarr
gpt = ggplot(dat, aes(x=Pct_GC, y=log2(Output/Input))) + 
    geom_point(alpha = 0.1) + 
    theme_cowplot() + 
    labs(x="GC Content (%)", y="Normalized log2(Output/Input)", title="ASTARR")
print(gpt)
```
