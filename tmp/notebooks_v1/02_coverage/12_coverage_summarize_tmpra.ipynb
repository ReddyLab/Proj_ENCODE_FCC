## Import coverage data of Tiling MPRA

**set variables**

```r
SAMPLES = c("Input", "Output")
CNAMES  = c("Chrom", "Start", "End")
```

**Region: GATA1**

```r
### import input and output coverage
prefix  = "Tile_K562_hg38_20210130.GATA1.stranded_pos"
lst_dat = lapply(SAMPLES, function(SAMPLE){
    fdiry = file.path(FD_RES, "Tewhey_K562_TileMPRA", "coverage")
    fname = paste0(prefix, ".", SAMPLE, ".perbase.tsv.gz")
    fpath = file.path(fdiry, fname)
    
    cnames = c(CNAMES, SAMPLE)
    dat = read_tsv(fpath, col_names = cnames, show_col_types = FALSE)
    return(dat)
})

### combine
fun = function(dat1, dat2){dat = inner_join(dat1, dat2, by=c("Chrom", "Start", "End"))}
dat = Reduce(fun, lst_dat)
dat = dat %>%
    mutate(
        Input  = replace(Input,  Input ==".", 0),
        Output = replace(Output, Output==".", 0)) %>% 
    mutate(
        Input  = as.integer(Input),
        Output = as.integer(Output)) %>%
    dplyr::filter(Input != 0) %>%
    mutate(FC = Output / Input)
        
### assign
dat_tmpra_gata1 = dat
head(dat)
```

## Export to bigwig files

```r
suppressMessages(suppressWarnings(library("GenomicRanges")))
suppressMessages(suppressWarnings(library("rtracklayer")))
suppressMessages(suppressWarnings(library("BSgenome.Hsapiens.UCSC.hg38")))
```

```r
genome_hg38 = BSgenome.Hsapiens.UCSC.hg38
```

```r
dat = dat_tmpra_gata1
head(dat)
```

```r
grg = GRanges(
    seqnames   = dat$Chrom,               
    ranges     = IRanges(
        start  = dat$Start+1,
        end    = dat$End), 
    strand     = "+",
    seqlengths = seqlengths(genome_hg38))
genome(grg) = "hg38"
```

```r
fdiry = file.path(FD_RES, "Tewhey_K562_TileMPRA", "coverage")
fname = "Tile_K562_hg38_20210130_Input.GATA1.perbase.bw"
fpath = file.path(fdiry, fname)

mcols(grg)$score = dat$Input
export(grg, fpath, format = "BigWig")
```

```r
fdiry = file.path(FD_RES, "Tewhey_K562_TileMPRA", "coverage")
fname = "Tile_K562_hg38_20210130_Output.GATA1.perbase.bw"
fpath = file.path(fdiry, fname)

mcols(grg)$score = dat$Output
export(grg, fpath, format = "BigWig")
```

```r
fdiry = file.path(FD_RES, "Tewhey_K562_TileMPRA", "coverage")
fname = "Tile_K562_hg38_20210130_FC.GATA1.perbase.bw"
fpath = file.path(fdiry, fname)

mcols(grg)$score = dat$FC
export(grg, fpath, format = "BigWig")
```

```r
fdiry = file.path(FD_RES, "Tewhey_K562_TileMPRA", "coverage")
fname = "Tile_K562_hg38_20210130_Log2FC.GATA1.perbase.bw"
fpath = file.path(fdiry, fname)

mcols(grg)$score = dat$log2FC
export(grg, fpath, format = "BigWig")
```
