**Test loop: Assay**

```bash
for ASSAY_NAME in ${ASSAY_NAMES[@]}; do
    
    ### set input and output files
    FD_INP=${FD_RES}/assay_fcc/${ASSAY_NAME}/fragment_counts
    FP_INPS=($(ls ${FD_INP}/*hg38*WGS*bed.gz))
    
    ### loop through each input file
    for FP_INP in ${FP_INPS[@]}; do
    
        ### set output file
        FN_INP=$(basename ${FP_INP})
        FD_OUT=${FD_INP}/summary
        FN_OUT="result.size.${FN_INP}.csv"
        FP_OUT=${FD_OUT}/${FN_OUT}
        
        ### set log file
        FN_LOG=coverage.library_size.${FN_INP}.txt
        FP_LOG=${FD_LOG}/${FN_LOG}
    
        ### show progress
        echo ======================================
        echo "Assay:      " ${ASSAY_NAME}
        echo "Input  file:" ${FP_INP}
        echo "Output file:" ${FP_OUT}
        echo "Log    file:" '${FD_LOG}/'${FN_LOG}
        echo
    done
done
```

## Execute

**Run: Calculate library size for each sample in each assay**

```bash
for ASSAY_NAME in ${ASSAY_NAMES[@]}; do
    
    ### set input and output files
    FD_INP=${FD_RES}/assay_fcc/${ASSAY_NAME}/fragment_counts
    FP_INPS=($(ls ${FD_INP}/*hg38*WGS*bed.gz))
    
    ### loop through each input file
    for FP_INP in ${FP_INPS[@]}; do
    
        ### set output file
        FN_INP=$(basename ${FP_INP})
        FD_OUT=${FD_INP}/summary
        FN_OUT="result.size.${FN_INP}.csv"
        FP_OUT=${FD_OUT}/${FN_OUT}
        
        ### set log file
        FN_LOG=coverage.library_size.${FN_INP}.txt
        FP_LOG=${FD_LOG}/${FN_LOG}
    
        ### show progress
        echo ======================================
        echo "Assay:      " ${ASSAY_NAME}
        echo "Input  file:" ${FP_INP}
        echo "Output file:" ${FP_OUT}
        echo "Log    file:" '${FD_LOG}/'${FN_LOG}
        echo
        
        ### execute
        mkdir -p ${FD_OUT}
        sbatch -p ${NODE} \
            --exclude=dl-01 \
            --cpus-per-task 4 \
            --mem 4G \
            --output ${FP_LOG} \
            ${FD_EXE}/run_calulate_lib_size.sh ${FD_PRJ} ${FP_INP} ${FP_OUT}
        echo
    done
done
```
