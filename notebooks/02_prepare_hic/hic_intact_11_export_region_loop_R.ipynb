## Rename columns

```r
fun_toupper_first = function(txt_input){
    txt_first  = toupper(substring(txt_input, 1, 1))
    txt_rest   = substring(txt_input, 2)
    txt_output = paste0(txt_first, txt_rest)
    return(txt_output)
}
```

```r
dat = dat_region_import

### rename location
vec = c(
    "Chrom_1", "ChromStart_1", "ChromEnd_1",
    "Chrom_2", "ChromStart_2", "ChromEnd_2"
)
colnames(dat)[1:6] = vec

### captialize the first letter
vec = colnames(dat)
vec = fun_toupper_first(vec)

### rename strand and centroid
vec = dplyr::recode(
    vec, 
    "Strand1"   = "Strand_1", 
    "Strand2"   = "Strand_2",
    "Centroid1" = "Centroid_1", 
    "Centroid2" = "Centroid_2"
)
colnames(dat) = vec

### assign and show
dat_reigon_rename = dat
print(dim(dat))
fun_display_table(head(dat))
```

## Arrange tables

```r
### init
dat = dat_reigon_rename

### arrange columns
dat = dat %>% 
    dplyr::mutate(
        Length_1 = ChromEnd_1 - ChromStart_1,
        Length_2 = ChromEnd_2 - ChromStart_2,
        Size     = (Length_1 + Length_2) / 2,
        Region_1 = fun_gen_region(Chrom_1, ChromStart_1, ChromEnd_1),
        Region_2 = fun_gen_region(Chrom_2, ChromStart_2, ChromEnd_2),
        Name     = paste0(Region_1, "|", Region_2),
        Distance = Centroid_2 - Centroid_1
    ) 

### assign and show
dat_region_arrange = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

**Explore**

```r
dat = dat_region_arrange
all(dat$Length_1 == dat$Length_2)
```

## Prepare region A and B

```r
### init
vec_txt_cname_A = c("Chrom_1", "ChromStart_1", "ChromEnd_1", "Name")
vec_txt_cname_B = c("Chrom_2", "ChromStart_2", "ChromEnd_2", "Name")
vec_txt_cname   = c("Chrom",   "ChromStart",   "ChromEnd",   "Name", "Group", "Label")
txt_assay = TXT_ASSAY

### prepare point A
dat = dat_region_arrange
dat = dat %>% 
    dplyr::select(!!!vec_txt_cname_A) %>% 
    dplyr::mutate(
        Group = txt_assay,
        Label = "Loop_A"
    ) %>%
    dplyr::distinct()
colnames(dat) = vec_txt_cname
dat_region_A  = dat

### prepare point B
dat = dat_region_arrange
dat = dat %>% 
    dplyr::select(!!!vec_txt_cname_B)%>% 
    dplyr::mutate(
        Group = txt_assay,
        Label = "Loop_B"
    ) %>%
    dplyr::distinct()
colnames(dat) = vec_txt_cname
dat_region_B  = dat
```

## Arrange loops

```r
### init
dat = dat_region_import
vec_txt_cname1 = c(
    "chr",  "x1", "x2", 
    "chr2", "y1", "y2", 
    "observed", "centroid1", "centroid2", "radius"
)
vec_txt_cname2 = c(
    "Chrom_A", "ChromStart_A", "ChromEnd_A",
    "Chrom_B", "ChromStart_B", "ChromEnd_B",
    "Observed", "Centroid_A", "Centroid_B", "Radius"
)
vec_txt_cname3 = c(
    "Chrom_A", "ChromStart_A", "ChromEnd_A", "Region_A", "Length_A",
    "Chrom_B", "ChromStart_B", "ChromEnd_B", "Region_B", "Length_B",
    "Name", 
    "Observed", "Centroid_A", "Centroid_B", 
    "Radius", "Distance", "Size"
)

### get columns
dat = dat %>% dplyr::select(!!!vec_txt_cname1)
colnames(dat) = vec_txt_cname2

### arrange columns
dat = dat %>% 
    dplyr::mutate(
        Length_A = ChromEnd_A - ChromStart_A,
        Length_B = ChromEnd_B - ChromStart_B,
        Size     = (Length_A + Length_B) / 2,
        Region_A = paste0(Chrom_A, ":", ChromStart_A, "-", ChromEnd_A),
        Region_B = paste0(Chrom_B, ":", ChromStart_B, "-", ChromEnd_B),
        Name     = paste0(Region_A, "|", Region_B),
        Distance = Centroid_B - Centroid_A
    ) %>%
    dplyr::select(!!!vec_txt_cname3) %>%
    dplyr::distinct()

### assign and show
dat_region_arrange = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

**Check**

```r
dat = dat_region_arrange
all(dat$Length_A == dat$Length_B)
```
