**Import data**

```r
### set file path
txt_folder = TXT_FOLDER_REGION
txt_fdiry  = file.path(FD_DAT, "external", txt_folder)
txt_fname = "K562.hg38.ENCSR545YBD.ENCFF693XIL.hic_insitu.loop.bedpe.gz"
txt_fpath = file.path(txt_fdiry, txt_fname)

### read table
dat = read_tsv(txt_fpath, show_col_types = FALSE)

### arrange columns
### the first line contains juicer version
dat = dat %>% dplyr::filter(!row_number() == 1)
colnames(dat)[1] = "chr"

### assign and show
dat_region_import = dat
print(dim(dat))
fun_display_table(head(dat))
```

## Rename columns

```r
fun_toupper_first = function(txt_input){
    txt_first  = toupper(substring(txt_input, 1, 1))
    txt_rest   = substring(txt_input, 2)
    txt_output = paste0(txt_first, txt_rest)
    return(txt_output)
}
```

```r
dat = dat_region_import

### rename location
vec = c(
    "Chrom_1", "ChromStart_1", "ChromEnd_1",
    "Chrom_2", "ChromStart_2", "ChromEnd_2"
)
colnames(dat)[1:6] = vec

### captialize the first letter
vec = colnames(dat)
vec = fun_toupper_first(vec)

### rename strand and centroid
vec = dplyr::recode(
    vec, 
    "Strand1"   = "Strand_1", 
    "Strand2"   = "Strand_2",
    "Centroid1" = "Centroid_1", 
    "Centroid2" = "Centroid_2"
)
colnames(dat) = vec

### assign and show
dat_reigon_rename = dat
print(dim(dat))
fun_display_table(head(dat))
```

## Arrange tables

```r
### init
dat = dat_reigon_rename

### arrange columns
dat = dat %>% 
    dplyr::mutate(
        Length_1 = ChromEnd_1 - ChromStart_1,
        Length_2 = ChromEnd_2 - ChromStart_2,
        Size     = (Length_1 + Length_2) / 2,
        Region_1 = fun_gen_region(Chrom_1, ChromStart_1, ChromEnd_1),
        Region_2 = fun_gen_region(Chrom_2, ChromStart_2, ChromEnd_2),
        Name     = paste0(Region_1, "|", Region_2),
        Distance = Centroid_2 - Centroid_1
    ) 

### assign and show
dat_region_arrange = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

**Explore**

```r
dat = dat_region_arrange
all(dat$Length_1 == dat$Length_2)
```

## Prepare region A and B

```r
### init
vec_txt_cname_A = c("Chrom_1", "ChromStart_1", "ChromEnd_1", "Name")
vec_txt_cname_B = c("Chrom_2", "ChromStart_2", "ChromEnd_2", "Name")
vec_txt_cname   = c("Chrom",   "ChromStart",   "ChromEnd",   "Name", "Group", "Label")
txt_assay = TXT_ASSAY

### prepare point A
dat = dat_region_arrange
dat = dat %>% 
    dplyr::select(!!!vec_txt_cname_A) %>% 
    dplyr::mutate(
        Group = txt_assay,
        Label = "Loop_A"
    ) %>%
    dplyr::distinct()
colnames(dat) = vec_txt_cname
dat_region_A  = dat

### prepare point B
dat = dat_region_arrange
dat = dat %>% 
    dplyr::select(!!!vec_txt_cname_B)%>% 
    dplyr::mutate(
        Group = txt_assay,
        Label = "Loop_B"
    ) %>%
    dplyr::distinct()
colnames(dat) = vec_txt_cname
dat_region_B  = dat
```

**Show table**
