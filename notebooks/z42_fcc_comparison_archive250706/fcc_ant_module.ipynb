```r
txt_region = "fcc_astarr_macs_input_overlap"
txt_fdiry  = file.path(FD_RES, "region_coverage_fcc", txt_region, "summary")
txt_fname  = "result.coverage.zscore.share.tsv"
txt_fpath  = file.path(txt_fdiry, txt_fname)

dat = read_tsv(txt_fpath)
dat_zscore_share = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

```r
dat = dat_zscore_share
table(dat$Assay)
print(unique(dat$Assay))
```

```r
vec = c("ATAC-STARR", "WHG-STARR", "Lenti-MPRA", "CRISPRi-Growth")

dat = dplyr::left_join(
    dat_zscore_share, 
    dat_region_annot, 
    by = c("Chrom", "ChromStart", "ChromEnd", "Region"),
    relationship = "many-to-many"
)

dat = dat %>% dplyr::filter(Assay %in% vec) %>% na.omit
dat_merge = dat
print(dim(dat))
head(dat)
```

```r
dat = dat_merge
dat = dat %>% 
    dplyr::group_by(Assay, Label_B) %>%
    dplyr::summarise(
        Score   = mean(Score),
        .groups = "drop"
    )
dat_merge_summary = dat
print(dim(dat))
head(dat)
```

```r
dat = dat_merge
gpt = ggplot(dat, aes(x=Score, y=Label_B, color=Assay)) + 
    geom_boxplot() + 
    theme_cowplot() + 
    background_grid()
options(repr.plot.height=20, repr.plot.width=12)
print(gpt)
```

```r
vec_txt_color =  colorRampPalette(rev(brewer.pal(10, "RdYlBu")))
theme_color   = scale_fill_gradientn(colours = vec_txt_color(50), limits=c(-1, 1))
```

```r
dat = dat_merge_summary
gpt = ggplot(dat, aes(x=Assay, y=Label_B, fill=Score)) + 
    geom_tile() + 
    theme_cowplot() + 
    background_grid() +
    theme_color
options(repr.plot.height=20, repr.plot.width=10)
print(gpt)
```
