```r
dat = dat_tfmodule
lst = split(dat$TF, dat$Module)

gostres = gost(
    query             = lst,
    organism          = "gp__5a4w_txOm_zPI", 
    domain_scope      = "custom_annotated",
    correction_method = "fdr",
    custom_bg         = vec_txt_tf,
    multi_query       = FALSE,
    user_threshold    = 0.05,
    significant       = TRUE
)
```

```r
lst = gostres
dat = lst$result
dat = dat %>% dplyr::select(query, term_id, term_name, query_size, term_size, p_value)
dat = dat %>% dplyr::mutate(nlog10p = -log10(p_value))

dat_summary = dat
print(length(unique(dat$query)))
print(length(unique(dat$term_id)))
print(dim(dat))
head(dat)
```

```r
dat = dat_summary
dat = dat %>% 
    dplyr::select(query, term_name, nlog10p) %>%
    tidyr::spread(term_name, nlog10p) %>%
    tibble::column_to_rownames(var = "query") %>%
    replace(is.na(.), 0)

mat_summary = dat
head(dat)
```

```r
options(repr.plot.height=10, repr.plot.width=15)
vec = rev(hcl.colors(10, "Reds"))
vec = vec[2:length(vec)]
vec = c("white", vec)
vec_txt_color = vec

dat = mat_summary
plt = pheatmap(
    dat,
    #cluster_rows = res_hclust_module,
    #cluster_cols = res_hclust_tf,
    color = vec_txt_color,
    #breaks = c(0, 0.5, 1),
    #legend = FALSE,
    fontsize_col = 10
)

plt
```
