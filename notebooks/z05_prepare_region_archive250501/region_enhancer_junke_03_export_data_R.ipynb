```r
VEC_TXT_REGION_NAME = c(
    "ATAC_STARR_seq_100bp_genomic_bin_in_ATAC_seq_peak_cpm_20_avg_DNA_lib_",
    "WHG_STARR_TR",
    "TilingMPRA_ENCSR363XER_DNA_filtering_avg_DNA_lib_count_10",
    "TilingMPRA_ENCSR394HXI_DNA_filtering_avg_DNA_lib_count_10",
    "TilingMPRA_ENCSR917SFD_DNA_filtering_avg_DNA_lib_count_10",
    "LentiMPRA_DNA_filtering_avg_DNA_lib_count_10",
    "eSTARR_seq"
)

VEC_TXT_ASSAY_NAME = c(
    "ASTARR_TR",
    "WSTARR_TR",
    "TMPRA_OL45",
    "TMPRA_OL13",
    "TMPRA_OL43",
    "LMPRA_Nadav",
    "eSTARR"
)

VEC_TXT_ASSAY_GROUP = c(
    "ASTARR",
    "WSTARR",
    "TMPRA",
    "TMPRA",
    "TMPRA",
    "LMPRA",
    "eSTARR"
)

fun1 = function(txt){
    vec1 = VEC_TXT_REGION_NAME
    vec2 = VEC_TXT_ASSAY_NAME
    res  = fun_str_map_detect(txt, vec1, vec2)
    return(res)
}

fun2 = function(txt){
    vec1 = VEC_TXT_REGION_NAME
    vec2 = VEC_TXT_ASSAY_GROUP
    res  = fun_str_map_detect(txt, vec1, vec2)
    return(res)
}

fun3 = function(txt){
    vec1 = c("enhancer", "repressor")
    vec2 = c("A", "R")
    res  = fun_str_map_detect(txt, vec1, vec2)
    return(res)
}

fun4 = function(txt){
    vec1 = c("orientation_independent", "in_either_orientation")
    vec2 = c("B", "")
    res  = fun_str_map_detect(txt, vec1, vec2)
    return(res)
}
```

## Metadata

```r
txt_fdiry = file.path(
    FD_DAT, "processed", 
    "fcc_enhancer_Junke_zscore_240611", 
    "decompressed",
    "*",
    "merged_peak"
)
txt_fname = "*bed.gz"
txt_fglob = file.path(txt_fdiry, txt_fname)
vec_txt_fpath = Sys.glob(txt_fglob)
vec_txt_fname = basename(vec_txt_fpath)

dat = data.frame(
    FPath = vec_txt_fpath,
    FName = vec_txt_fname
)

dat = dat %>%
    dplyr::mutate(
        X1 = fun1(vec_txt_fpath),
        X2 = fun2(vec_txt_fpath),
        X3 = fun3(vec_txt_fname), 
        X4 = fun4(vec_txt_fname) 
    ) %>% 
    na.omit

dat = dat %>% 
    dplyr::mutate(
        Dataset = X1,
        Group   = X2,
        Label   = paste0(X2, "_", X3, X4)
    ) %>% 
    dplyr::select(
        Dataset,
        Group,
        Label,
        FName,
        FPath)

dat_meta = dat
fun_display_table(dat)
```

## Import data and arrange

```r
txt_fdiry = file.path(
    FD_DAT, "processed", 
    "fcc_enhancer_Junke_zscore_240611", 
    "decompressed",
    "*",
    "merged_peak"
)
txt_fname = "*bed.gz"
txt_fglob = file.path(txt_fdiry, txt_fname)
vec_txt_fpath = Sys.glob(txt_fglob)

vec_txt_cname = c(
    "Seqnames",
    "Start",
    "End",
    "Name",
    "Mean logFC across all the bins",
    "Strand",
    "Z_score based on mean logFC",
    "List of all logFCs",
    "Summit seqnames",
    "Summit start",
    "Summit end",
    "Summit logFC",
    "Summit z-score",
    "Peak size"
)

lst = lapply(vec_txt_fpath, function(txt_fpath){
    txt_fname = basename(txt_fpath)
    
    dat = read_tsv(txt_fpath, col_names = vec_txt_cname, show_col_types = FALSE)
    dat = dat %>% 
        dplyr::mutate(
            X1 = fun1(txt_fpath),
            X2 = fun2(txt_fpath),
            X3 = fun3(txt_fname), 
            X4 = fun4(txt_fname) 
        ) %>% 
        dplyr::mutate(
            Dataset = X1,
            Group   = X2,
            Label   = paste0(X2, "_", X3, X4)
        ) %>% 
        dplyr::select(
            Seqnames, 
            Start, 
            End, 
            Name,
            `Z_score based on mean logFC`,
            Strand,
            Group,
            Label,
            Dataset
        )
    return(dat)
})

dat = bind_rows(lst)
colnames(dat) = c(
    "Chrom", "ChromStart", "ChromEnd", "Name", "Score", "Strand", 
    "Group", 
    "Label", 
    "Dataset"
)
lst = split(dat, dat$Group)

dat_region = dat
lst_dat_region = lst
head(dat)
```
