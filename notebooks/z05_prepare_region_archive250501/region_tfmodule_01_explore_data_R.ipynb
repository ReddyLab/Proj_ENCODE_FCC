```r
library(pheatmap)
```

```r
dat = dat_module_tfs
print(length(unique(dat$module)))
print(length(unique(dat$TF)))
head(dat)
```

```r
dat = dat_module_tfs
dat = dat %>% 
    dplyr::mutate(Label = 1) %>%
    tidyr::spread(module, Label) %>%
    replace(is.na(.), 0) %>%
    column_to_rownames("TF")
head(dat)
```

```r
options(repr.plot.height=10, repr.plot.width=10)
pheatmap(
    t(dat), 
    color = c("white", "black"),
    breaks = c(0, 0.5, 1),
    legend = FALSE,
    fontsize_col = 5
)
```

```r
dat = dat_module_reg
dat = dat %>% dplyr::filter(chr == "chrX")
print(dim(dat))
```

```r
dat = dat_module_reg
dat = dat %>% dplyr::filter(chr == "chrX")

dat = dat %>% dplyr::mutate(Region = fun_gen_region(chr, start, end))
print(length(unique(dat$Region)))

mat = dat %>% dplyr::select(-chr, -start, -end, -TFs)
mat_region_module = mat

mat = dat %>% dplyr::select(Region, TFs)
mat_region_tfs = mat
```

```r
head(mat_region_module)
```

```r
set.seed(123)
mat = mat_region_module
#idx = sample(1:nrow(mat), size=10000) #173997)
#mat = mat[idx,]
mat = mat %>% na.omit
mat = mat %>% dplyr::select(-Region)

options(repr.plot.height=10, repr.plot.width=10)
pheatmap(
    t(mat), 
    color = c("white", "black"),
    breaks = c(0, 0.5, 1),
    legend = FALSE,
    fontsize_col = 5
)
```

```r
set.seed(123)
mat = mat_region_tfs
#idx = sample(1:nrow(mat), size=1000) #173997)
#mat = mat[idx,]
mat = mat %>% 
    tidyr::separate_rows(TFs) %>% 
    dplyr::mutate(Label = 1) %>%
    dplyr::distinct() %>% 
    tidyr::spread(TFs, Label) %>%
    dplyr::select(-Region) %>%
    replace(is.na(.), 0)
print(dim(mat))
head(mat)
```

```r
options(repr.plot.height=10, repr.plot.width=10)
pheatmap(
    t(mat), 
    color = c("white", "black"),
    breaks = c(0, 0.5, 1),
    legend = FALSE,
    fontsize_row = 5,
    fontsize_col = 5
)
```
