```r
set.seed(123)
dat = data.frame(
    x = rnorm(1000),
    y = rnorm(1000),
    z = sample(c("a", "b"), size = 1000, replace = TRUE)
)
```

```r
gpt = ggplot(dat, aes(x = x, y = y, color = z)) +
    geom_point() +
    theme_cowplot()
```

```r
options(repr.plot.height=5, repr.plot.width=6)
print(gpt)
```
