```r
set.seed(123)
num = 10000
dat = dat_cnt
idx = sample(1:nrow(dat), size = num)
dat = dat[idx,]
mat = dat %>% dplyr::select(starts_with("Input"), starts_with("Output"))

mat_cnt = mat
print(dim(mat))
head(mat)
```

```r
options(repr.plot.width = 7, repr.plot.height = 7)
ggpairs(
    mat, 
    aes(alpha = 0.5),
    upper = list(continuous = wrap("cor", size = 5))
    ) +
    theme(
        strip.text = element_text(size = 10))
```

```r
str_extract("Input.rep1", "Input|Output")
```

```r
dat = mat_cnt
dat = dat %>% correlate(method = "spearman") %>% shave() %>% stretch() %>% na.omit
dat = dat %>% 
    dplyr::mutate(
        Lib_x = str_extract(x, "Input|Output"),
        Lib_y = str_extract(y, "Input|Output")) %>%
    dplyr::mutate(
        Compare = paste(Lib_x, "x", Lib_y)
    )

set.seed(123)
gpt = ggplot(dat, aes(x=Compare, y=r)) + 
    geom_boxplot(width = 0.25) +
    geom_jitter(width = 0.25) + 
    ylim(0.7, 1.0) +
    theme_cowplot()
gpt
```
