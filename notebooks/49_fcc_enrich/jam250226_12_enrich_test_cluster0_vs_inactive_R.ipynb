## Arrange matrix

```r
dat = mat_region_score_import
dat = dat %>% dplyr::select(-Chrom, -ChromStart, -ChromEnd) %>% column_to_rownames("Region")
mat_region_score = dat
head(dat, 3)
```

```r
vec_txt_assay = c("ASTARR", "WSTARR", "TMPRA", "LMPRA", "CRISPRi-HCRFF", "CRISPRi-Growth")
print(vec_txt_assay)
```

```r
dat = mat_region_score
vec = colnames(dat)
vec = setdiff(vec, vec_txt_assay)

vec_txt_feature = vec
print(length(vec))
print(head(vec))
print(tail(vec))
```

## Prepare for test

```r
###
mgr  = mgr_region_target_vs_others
dat  = mat_region_score
vec1 = focal(mgr)$Region
vec2 = matched(mgr)$Region
dat1 = dat[vec1,]
dat2 = dat[vec2,]

###
dat1$Group = "1"
dat2$Group = "0"
dat = bind_rows(dat1, dat2)

###
dat_region_test = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

## Test all features for the targeted cluster

```r
### init: parameters
num_iter = 10000

### loop through each region group
set.seed(123)

### show progress
dat = dat_region_test
cat(nrow(dat), "\n")
flush.console()

### perform permutation test for each feature
lst = lapply(vec_txt_feature, function(txt_feature){
    
    ### select the feature and group lable
    dat = dat_region_test
    vec = c(txt_feature, "Group")
    dat = dat[, vec]
    colnames(dat) = c("Score", "Group")

    ### perform a permutation test
    dat = dat %>% dplyr::mutate(Group = factor(Group, levels = c("1", "0")))
    res = coin::oneway_test(
        Score ~ Group, 
        data = dat, 
        distribution = approximate(nresample = num_iter)
    )

    return(res)
})
names(lst) = vec_txt_feature
    

### assign and show
lst_fit_object = lst
print(names(lst))
```

## Summarize the results

```r
lst = lst_fit_object
lst = lapply(lst, function(fit){
    res = c(
        coin::statistic(fit), 
        coin::pvalue(fit) 
    )
    names(res) = c("Stat", "Pval")
    return(res)
})

###
dat = bind_rows(lst, .id = "Feature")
dat = dat %>% 
    dplyr::mutate(Score = as.numeric(scale(Stat))) %>%
    dplyr::mutate(Padj  = p.adjust(Pval, method = "BH"))

### assign and show
dat_fit_summary = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```
