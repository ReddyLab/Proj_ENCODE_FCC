## Loop and run differential analysis

```r
### loop through each region group
lst = lapply(lst_mgr_region, function(mgr){
    
    ### extract targeted and matched region sets
    vec1 = focal(mgr)$Region
    vec2 = matched(mgr)$Region

    ### setup data for testing
    dat  = mat_region_score
    dat1 = dat[vec1,]
    dat2 = dat[vec2,]
    dat1$Group = "1"
    dat2$Group = "0"

    ### combine data
    dat = bind_rows(dat1, dat2)
    dat_region_test = dat

    return(dat)
})

lst_dat_region_test = lst
print(names(lst))
```

```r
### init: parameters
num_iter = 10000

### loop through each region group
set.seed(123)
lst_outer = lapply(lst_dat_region_test[-1], function(dat_region_test){
    ### show progress
    dat = dat_region_test
    cat(nrow(dat), "\n")
    flush.console()
    
    ### perform permutation test for each feature
    lst_inner = lapply(vec_txt_feature, function(txt_feature){
        
        ### select the feature and group lable
        dat = dat_region_test
        vec = c(txt_feature, "Group")
        dat = dat[, vec]
        colnames(dat) = c("Score", "Group")
    
        ### perform a permutation test
        dat = dat %>% dplyr::mutate(Group = factor(Group, levels = c("1", "0")))
        res = coin::oneway_test(
            Score ~ Group, 
            data = dat, 
            distribution = approximate(nresample = num_iter)
        )
    
        return(res)
    })
    names(lst_inner) = vec_txt_feature
    
    return(lst_inner)
})

### assign and show
lst_fit_object = lst_outer
print(names(lst_outer))
```
