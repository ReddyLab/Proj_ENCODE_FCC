## sampling

```r
suppressMessages(suppressWarnings(library("nullranges")))
suppressMessages(suppressWarnings(library("plyranges")))
suppressMessages(suppressWarnings(library("patchwork")))
```

```r
vec = c("Chrom", "ChromStart", "ChromEnd", "Region")
dat = dplyr::inner_join(
    mat_region_focal,
    mat_region_covariate,
    by = vec
)
print(dim(dat))

lst = split(dat, dat$Group)
lst_dat_region_focal = lst
names(lst)
```

```r
vec = c("Chrom", "ChromStart", "ChromEnd", "Region")
dat = dplyr::inner_join(
    mat_region_pool,
    mat_region_covariate,
    by = vec
)
print(dim(dat))

lst = split(dat, dat$Note)
lst_dat_region_pool = lst
names(lst)
```

```r
lst = lst_dat_region_focal
lst = lapply(lst, function(dat){
    grg = GRanges(
        seqnames = dat$Chrom,
        ranges = IRanges(
            start = dat$ChromStart, 
            end   = dat$ChromEnd
        ),
        Region = dat$Region,
        ATAC   = dat$ATAC,
        pGC    = dat$pGC,
        Length = dat$Length
    )
    return(grg)
})

lst_grg_region_focal = lst
```

```r
lst = lst_dat_region_pool
lst = lapply(lst, function(dat){
    grg = GRanges(
        seqnames = dat$Chrom,
        ranges = IRanges(
            start = dat$ChromStart, 
            end   = dat$ChromEnd
        ),
        Region = dat$Region,
        ATAC   = dat$ATAC,
        pGC    = dat$pGC,
        Length = dat$Length
    )
    return(grg)
})

lst_grg_region_pool = lst
```

```r
vec_txt_focal = c(
    'Distal:Active',
    'Distal:Repressive',
    #'Proximal:Active',
    'Proximal:Repressive'
)

vec_txt_pool = c(
    "Distal:Inactive", 
    "Distal:Inactive", 
    #"Proximal:Inactive", 
    "Proximal:Inactive"
)
```

```r
lst = list()
for (idx in 1:3){
    txt_focal = vec_txt_focal[idx]
    txt_pool  = vec_txt_pool[idx]
    cat("Index:", idx, "\n")
    cat(txt_focal, "\n")
    cat(txt_pool,  "\n")
    cat("\n")

    grg_region_focal = lst_grg_region_focal[[txt_focal]]
    grg_region_pool  = lst_grg_region_pool[[txt_pool]]
    print(length(grg_region_focal))
    print(length(grg_region_pool))
    
    ### Match ranges
    set.seed(123)
    mgr = matchRanges(
        focal = grg_region_focal,
        pool  = grg_region_pool,
        covar = ~ ATAC + pGC,
        method = 'stratified',
        replace = FALSE
    )
    txt_title = paste(txt_focal, txt_pool, sep="_")
    lst[[txt_title]] = mgr
    print(txt_title)
    
    print(length(mgr))
    print(length(unique(mgr$Region)))
    cat("\n")
    flush.console()
}

lst_mgr_region = lst
```

```r
lst = lst_mgr_region
for (idx in names(lst_mgr_region)){
    
    txt = str_remove(idx, pattern = ":")
    cat(idx, "\n")
    cat(txt, "\n")
}
```

```r
txt_fdiry = file.path(
    FD_RES, 
    "region_integration",
    "fcc_astarr_macs_input_overlap",
    "analysis_enrichment_v2",
    "fcc_starrmpra_vote2_v2_split_pool_by_tss"
)

for (idx in names(lst_mgr_region)){
    ###
    txt = str_remove_all(idx, pattern = ":")
    cat("Progress:", idx, "\n")
    
    ###
    mgr = lst_mgr_region[[idx]]
    cat("#Region: ", length(mgr), "\n")
    
    ### save results
    txt_fname = paste("MatchedGRanges", txt, sep = "_")
    txt_fname = paste(txt_fname, "rds", sep = ".")
    txt_fpath = file.path(txt_fdiry, txt_fname)
    
    cat("Saved:   ", txt_fname, "\n")
    obj = mgr
    saveRDS(obj, txt_fpath)
    cat("\n")
}
```
