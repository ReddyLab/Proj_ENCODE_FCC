```r
txt_fdiry = file.path(
    FD_RES, 
    "region_integration",
    "fcc_astarr_macs_input_overlap",
    "analysis_enrichment_v2",
    "fcc_starrmpra_vote2_v2_split_pool_by_tss"
)
vec_txt_fname = c(
    "results_DistalActive_perm10k.rds",
    "results_DistalRepressive_perm10k.rds",
    "results_ProximalActive_perm10k.rds",
    "results_ProximalRepressive_perm10k.rds"
)
vec_txt_fpath = file.path(txt_fdiry, vec_txt_fname)
                          
vec_txt_group = c("Distal:Active", "Distal:Repressive", "Proximal:Active", "Proximal:Repressive")
```

```r
lst = lapply(vec_txt_fpath, function(txt_fpath){
    obj = readRDS(txt_fpath)
    return(obj)
})
names(lst) = vec_txt_group

lst_fit_object = lst
```

```r
lst = lst_fit_object
lst = lst[[1]]
names(lst)
```

```r
lst_outer = lst_fit_object
lst_outer = lapply(lst_outer, function(lst_inner){
    ###
    lst_inner = lapply(lst_inner, function(fit){
        res = c(
            coin::statistic(fit), 
            coin::pvalue(fit) 
        )
        names(res) = c("Stat", "Pval")
        return(res)
    })

    ###
    dat = bind_rows(lst_inner, .id = "Feature")
    dat = dat %>% 
        dplyr::mutate(Score = as.numeric(scale(Stat))) %>%
        dplyr::mutate(Padj  = p.adjust(Pval, method = "BH"))
    return(dat)
})

###
dat = bind_rows(lst_outer, .id = "Group")

###
dat_fit_summary = dat
print(dim(dat))
fun_display_table(head(dat))
```
