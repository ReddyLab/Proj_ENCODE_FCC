```r
txt_feature = "H3K27ac"

dat = dat_region_test
vec = c(txt_feature, "Group")
dat = dat[, vec]
colnames(dat) = c("Score", "Group")

dat = dat %>% 
    dplyr::group_by(Group) %>% 
    dplyr::summarise(
        Mean   = mean(Score, na.rm = TRUE),
        Median = median(Score, na.rm = TRUE),
        SD     = sd(Score, na.rm = TRUE),
        Min    = min(Score, na.rm = TRUE),
        Max    = max(Score, na.rm = TRUE),
        N      = dplyr::n()
    ) %>%
    dplyr::ungroup()
dat
```

```r
txt_feature = "H3K27ac"

dat = dat_region_test
vec = c(txt_feature, "Group")
dat = dat[, vec]
colnames(dat) = c("Score", "Group")

# Perform a permutation test
num_iter = 10000
dat = dat %>% dplyr::mutate(Group = factor(Group, levels = c("1", "0")))
res_perm = coin::oneway_test(
    Score ~ Group, 
    data = dat, 
    #distribution = "asymptotic"
    distribution = approximate(nresample = num_iter)
)

# Print results
print(res_perm)
```

**Test multiple features**

```r
vec_txt_feature = c("CTCF", "SMC3", "RAD21", "POLR2A", "EP300", "EZH2", "REST", "SIN3A",
        'H3K27ac', 'H3K4me1', 'H3K4me2', 'H3K4me3', 'H3K27me3')
vec = c(vec_txt_feature, "Group")

###
dat = dat_region_test
dat = dat[, vec]

head(dat, 3)
```

```r
num_iter = 10000

lst = lapply(vec_txt_feature, function(txt_feature){
    ### select the feature and group lable
    dat = dat_region_test
    vec = c(txt_feature, "Group")
    dat = dat[, vec]
    colnames(dat) = c("Score", "Group")

    ### perform a permutation test
    dat = dat %>% dplyr::mutate(Group = factor(Group, levels = c("1", "0")))
    res = coin::oneway_test(
        Score ~ Group, 
        data = dat, 
        distribution = approximate(nresample = num_iter)
    )

    return(res)
})
names(lst) = vec_txt_feature

lst_fit_object = lst
```

```r
lst = lst_fit_object
lst = lapply(lst, function(fit){
    res = c(
        coin::statistic(fit), 
        coin::pvalue(fit) 
    )
    names(res) = c("Stat", "Pval")
    return(res)
})
dat = bind_rows(lst, .id = "Feature")
dat = dat %>% dplyr::mutate(Group = "TSS:Silencer")

dat_fit_summary = dat
dat
```
