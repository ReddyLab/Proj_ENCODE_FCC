```r
dat = dplyr::left_join(
    dat_annot %>% rownames_to_column("Region"),
    mat_score %>% rownames_to_column("Region"),
    by = "Region")

lst = split(dat, dat$Group)
```

```r
dat = lst[[2]]
rownames(dat) = NULL
dat = dat %>% dplyr::select(-Group) %>% column_to_rownames(var="Region")
print(dim(dat))
head(dat)
```

```r
mat = cor(t(dat), use="pairwise.complete.obs")
mat = as.dist(1 - mat)
res = hclust(mat, method="complete")
```

```r
vec = res$labels[res$order]
head(vec)
```

```r
head(res$labels)
```

```r
dat = dplyr::left_join(
    dat_annot %>% rownames_to_column("Region"),
    mat_score %>% rownames_to_column("Region"),
    by = "Region")

lst = split(dat, dat$Group)
lst = lapply(lst, function(dat){
    rownames(dat) = NULL
    dat = dat %>% dplyr::select(-Group) %>% column_to_rownames(var="Region")
    mat = cor(t(dat), use="pairwise.complete.obs")
    mat = as.dist(1 - mat)
    res = hclust(mat, method="complete")
    vec = res$labels[res$order]
    return(vec)
})
```

```r
lst_res_hclust = lst
```

```r
lst = lst_res_hclust
vec = c(
    lst[["Proximal:Active"]], 
    lst[["Distal:Active"]], 
    lst[["Proximal:Repressive"]], 
    lst[["Distal:Repressive"]]
)
vec_txt_region_ordered = vec
```

```r
lst = lst_res_hclust
vec = c(
    lst[["Proximal:Active"]] %>% head(100), 
    lst[["Distal:Active"]] %>% head(100), 
    lst[["Proximal:Repressive"]] %>% head(100), 
    lst[["Distal:Repressive"]] %>% head(100)
)
vec_txt_region_subset = vec
```

```r
dat = tribble(
    ~Group,                 ~Color,
    "Proximal:Active",      "#e41a1c", 
    "Distal:Active",        "#ff7f00",
    "Distal:Repressive",    "#377eb8",
    "Proximal:Repressive",  "#f781bf"
)
dat_group_color = dat
```

```r
dat = dat_group_color
vec = dat$Color
names(vec) = dat$Group
lst = list(Group = vec)

lst_txt_color = lst
```

```r
n = 10
vec = brewer.pal(n = 7, name = "Reds")
vec = c(vec, "white", "white")
vec = colorRampPalette(vec)(n)
vec_txt_color = vec
```

```r
dat = mat_score
idx = vec_txt_region_ordered
dat = dat[idx,]

set.seed(123)

pheatmap(
    dat, 
    color  = vec_txt_color,
    cluster_rows      = FALSE,
    annotation_row    = dat_annot,
    annotation_colors = lst_txt_color,
    treeheight_row = 0,
    treeheight_col = 0,
    show_rownames = FALSE, 
    show_colnames = FALSE)
```

```r
dat = mat_score
idx = vec_txt_region_subset
dat = dat[idx,]

set.seed(123)

pheatmap(
    dat, 
    color  = vec_txt_color,
    cluster_rows      = TRUE,
    annotation_row    = dat_annot,
    annotation_colors = lst_txt_color,
    treeheight_row = 0,
    treeheight_col = 0,
    show_rownames = FALSE, 
    show_colnames = FALSE)
```

```r
n = 100
vec_txt_color = colorRampPalette(rev(brewer.pal(n = 7, name = "RdYlBu")))(n)
vec_txt_break = c(
    seq(-1,    0, length.out=ceiling(n/2) + 1), 
    seq(10/n, 10, length.out=floor(n/2)))
```

```r
set.seed(123)
dat = mat_score
dat = dat %>% dplyr::slice_sample(n = 1000)

pheatmap(
    dat, 
    color  = vec_txt_color,
    breaks = vec_txt_break,
    show_rownames = FALSE, 
    show_colnames = FALSE)
```
