## Merge

```r
fun = function(dat1, dat2){
    dat = dplyr::full_join(dat1, dat2, by = c("Chrom", "ChromStart", "ChromEnd", "Region"))
    return(dat)
}

lst = list(
    mat_annot_fcc,
    mat_annot_tss,
    mat_annot_chipseq
)

dat = Reduce(fun, lst)
dat = dat %>% replace(is.na(.), 0)

mat_annot_merge_full_tot = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

```r
fun = function(dat1, dat2){
    dat = dplyr::full_join(dat1, dat2, by = c("Chrom", "ChromStart", "ChromEnd", "Region"))
    return(dat)
}

lst = list(
    mat_annot_fcc,
    mat_annot_chipseq
)

dat = Reduce(fun, lst)
dat = dat %>% replace(is.na(.), 0)

mat_annot_merge_notss_tot = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```
