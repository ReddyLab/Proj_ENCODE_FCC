## Summary

```r
dat = bind_rows(
    dat_region_annot_chipseq_histone,
    dat_region_annot_chipseq_flagship
)
dat = dat %>% dplyr::mutate(Group = "ChIP-seq:Full")

dat_region_annot_chipseq_full = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

```r
### init
lst = list(
    "encode_chipseq_subset"   = dat_region_annot_chipseq_subset,
    "encode_chipseq_histone"  = dat_region_annot_chipseq_histone,
    "encode_chipseq_flagship" = dat_region_annot_chipseq_flagship,
    "encode_chipseq_full"     = dat_region_annot_chipseq_full
)

### arrange columns
lst = lapply(lst, function(dat){

    dat = dat %>%
        dplyr::select(Chrom, ChromStart, ChromEnd, Region, Group, Label) %>%
        dplyr::distinct()

    return(dat)
})

### assign
lst_dat_summary = lst
```

```r
### init
lst = lst_dat_summary

### spread to matrix
lst = lapply(lst, function(dat){

    dat = dat %>%
        dplyr::select(Chrom, ChromStart, ChromEnd, Region, Label) %>%
        dplyr::mutate(Value  = 1) %>%
        dplyr::distinct()

    dat = dat %>%
        tidyr::spread(Label, Value) %>%
        replace(is.na(.), 0)

    return(dat)
})

### assign
lst_dat_matrix = lst
```

## Save results

```r
txt_fdiry = file.path(
    FD_RES, 
    "region_annotation", 
    "fcc_astarr_macs_input_overlap",
    "summary"
)

lst = lst_dat_summary
for (idx in names(lst)){
    ### get table
    dat = lst[[idx]]

    ### set file path
    txt_fname = paste("region", "summary", idx, "tsv", sep = ".")
    txt_fpath = file.path(txt_fdiry, txt_fname)
    print(txt_fname)
    
    ### write table
    write_tsv(dat, txt_fpath)
}
```
