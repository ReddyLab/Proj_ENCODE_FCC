**Read region annotation**

```r
### loop to import data
lst = lapply(VEC_TXT_FOLDER, function(txt_folder){
    ### set file directory
    txt_fdiry = file.path(FD_RES, "region_annotation", txt_folder, "summary")
    txt_fname = "region.annotation.encode_e2g_benchmark.regulated.tsv"
    txt_fpath = file.path(txt_fdiry, txt_fname)

    ### read table
    dat = read_tsv(txt_fpath, show_col_types = FALSE)
    return(dat)
})
names(lst) = VEC_TXT_FOLDER

### assign and show
lst_dat_region_annot_e2g_label = lst

res = lapply(lst, dim)
print(res)

dat = lst[[1]]
fun_display_table(head(dat, 3))
```

## Arrange table

```r
lst = lapply(VEC_TXT_FOLDER, function(idx){
    ### get tables
    dat_region_link  = lst_dat_region_annot_e2g_link[[idx]]
    dat_region_label = lst_dat_region_annot_e2g_label[[idx]]

    ### get region significant for more than one target
    dat = dat_region_label
    dat = dat %>% dplyr::filter(Regulated_Count > 1)
    vec_txt_region = dat$Region

    ### get the singificant link with region targeted more than one gene
    dat = dat_region_link
    dat = dat %>% 
        dplyr::select(Region, Target, Regulated) %>%
        dplyr::distinct() %>%
        dplyr::filter(Regulated) %>%
        dplyr::filter(Region %in% vec_txt_region)
    
    return(dat)
})

### assign and show
lst_dat_region_merge = lst

res = lapply(lst, dim)
print(res)

dat = lst[[1]]
fun_display_table(head(dat, 3))
```

## Explore: Count table

```r
lst = lst_dat_region_annot_e2g_label
lst = lapply(lst, function(dat){
    dat = dat %>% dplyr::mutate(Note = paste("Regulated", Regulated_Ifany, sep = ":"))
    res = table(dat$Note, dnn = "Note")
    dat = as.data.frame(res)
    return(dat)
})

dat = bind_rows(lst, .id = "Region")
dat = dat %>% 
    tidyr::spread(Note, Freq) %>% 
    dplyr::mutate(Total = `Regulated:FALSE` + `Regulated:TRUE`)
#fun_display_table(dat)
dat %>% kableExtra::kable("markdown")
#dat
```

## Explore: Region overlap

```r
### prepare list
lst = lst_dat_region_merge
dat = lst[[1]]
lst = split(dat$Region, dat$Target)

### generate plot
plt = upset(
    fromList(lst), 
    order.by = "freq", 
    text.scale=1.7, 
    nsets = length(lst),
    nintersects = NA,
    mb.ratio = c(0.3, 0.7)
)

### assign and show
plt_export_upset_gene = plt
options(repr.plot.height=20, repr.plot.width=10)
print(plt)
```
