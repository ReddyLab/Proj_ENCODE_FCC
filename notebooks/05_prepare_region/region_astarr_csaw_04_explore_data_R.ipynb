## Explore data

### Compare the number of significant regions

````r
### combine two tables
lst = list(
    "KSMerge (6Dna7Rna)" = dat_region_ksmerge,
    "KS91 (6Dna4Rna)"    = dat_region_ks91
)
dat = bind_rows(lst, .id = "Label")

### label postive, negative, and total
tmp = dat %>% dplyr::mutate(Sign = "Total")
dat = dat %>% dplyr::mutate(
    Sign = ifelse(
        Log2FC > 0, 
        "Positive",
        "Negative"
    )
)
dat = bind_rows(dat, tmp)

### label by q-values
dat = dat %>% dplyr::mutate(
    Signif0_3 = MinusLog10QValue >= 0.3,
    Signif0_4 = MinusLog10QValue >= 0.4,
    Signif0_5 = MinusLog10QValue >= 0.5,
    Signif1   = MinusLog10QValue >= 1,
    Signif2   = MinusLog10QValue >= 2,
    Signif3   = MinusLog10QValue >= 3
)

### assign and show
dat_region_arrange = dat
fun_display_table(head(dat, 3))
````

````r
dat = dat_region_arrange
dat = dat %>%
    dplyr::group_by(Label, Sign) %>%
    dplyr::summarise(
        Count          = n(),
        `-log10Q >= 2` = sum(Signif2),
        `-log10Q >= 3` = sum(Signif3),
        .groups = "drop"
    )

fun_display_table(dat)
````

````r
dat = dat_region_arrange
dat = dat %>%
    dplyr::filter(Signif3) %>%
    dplyr::group_by(Label, Sign) %>%
    dplyr::summarise(
        Count = n(),
        .groups = "drop"
    ) %>%
    tidyr::spread(Sign, Count)

fun_display_table(dat)
````

````r
dat = dat_region_arrange
dat = dat %>%
    dplyr::filter(Signif3) %>%
    dplyr::group_by(Label, Sign) %>%
    dplyr::summarise(
        Count = n(),
        .groups = "drop"
    ) %>%
    tidyr::spread(Sign, Count) %>%
    dplyr::mutate(
        Positive = round(Positive / Total, 2),
        Negative = round(Negative / Total, 2)
    )

fun_display_table(dat)
````

```

Number of significant regions increased. Negative regions increased more than positive regions.
```

## Save results

### Figure: Compare score distribution (Log2FC + Q-Value)

````r
### combine table
lst = list(
    "KSMerge (6Dna7Rna)" = dat_region_ksmerge,
    "KS91 (6Dna4Rna)"    = dat_region_ks91
)
dat = bind_rows(lst, .id = "Label")

### label by q-values
dat = dat %>% dplyr::mutate(Signif3 = MinusLog10QValue >= 3)

### assign and show
dat_region_arrange = dat
print(dim(dat))
fun_display_table(head(dat, 3))
````

````r
### set text size
theme_text = theme(
    title      = element_text(size = 16),
    axis.title = element_text(size = 16),
    axis.text  = element_text(size = 14)
)

### label and set factor
dat = dat_region_arrange
vec = c("KSMerge (6Dna7Rna)", "KS91 (6Dna4Rna)")
dat = dat %>% dplyr::mutate(Label = factor(Label, levels = vec))

### summarize the number of regions (total)
tmp = dat
vec = tmp %>%
    group_by(Label) %>%
    summarise(Count = n(), .groups="drop") %>%
    dplyr::mutate(
        line1 = Label,
        line2 = paste("Count =", scales::comma(Count)),
        combined = paste0(line1, "\n", line2)
    ) %>%
    { set_names(.$combined, .$Label) }
vec_txt_label = vec

### plot the score distribution (total)
gpt = ggplot(tmp, aes(x=Log2FC, color=Label)) +
    geom_density(linewidth = 1, alpha = 0.5) +
    scale_color_discrete(
        ### replace each original level by our multiline version
        labels = vec_txt_label
    ) +
    labs(x="Log2(fold-change)", y="Density", title="ASTARR CSAW Regions (Total)")
gpt1 = gpt


### summarize the number of regions (significant)
tmp = dat %>% dplyr::filter(Signif3)
vec = tmp %>%
    group_by(Label) %>%
    summarise(Count = n(), .groups="drop") %>%
    dplyr::mutate(
        line1 = Label,
        line2 = paste("Count =", scales::comma(Count)),
        combined = paste0(line1, "\n", line2)
    ) %>%
    { set_names(.$combined, .$Label) }
vec_txt_label = vec

### plot the score distribution (significant)
gpt = ggplot(tmp, aes(x=Log2FC, color=Label)) +
    geom_density(linewidth = 1, alpha = 0.5) +
    scale_color_discrete(
        ### replace each original level by our multiline version
        labels = vec_txt_label
    ) +
    labs(x="Log2(fold-change)", y="Density", title="ASTARR CSAW Regions (q-value < 0.001)")
gpt2 = gpt

options(repr.plot.height=4, repr.plot.width=10)
lst = list(gpt1, gpt2)
lst = lapply(lst, function(gpt){
    gpt = gpt +
        theme_cowplot() +
        background_grid() +
        theme_text +
        theme(
            legend.position = "inside", 
            legend.position.inside = c(0.1, 0.6),
            legend.background = element_rect(fill = "white"),
            legend.key.spacing.y = unit(0.5, 'cm')
        )
    return(gpt)
})

plt = plot_grid(plotlist = lst, ncol = 1, align = "v")
plt_export = plt

options(repr.plot.height=6, repr.plot.width=6)
print(plt)
````
