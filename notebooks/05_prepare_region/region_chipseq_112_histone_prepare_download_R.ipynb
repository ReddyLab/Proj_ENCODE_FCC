```r
fun = function(txt){
    lst = str_split(txt, ",")
    lst = lapply(lst, length)
    vec = unlist(lst)
    return(vec)
}

dat = dat_metadata_chipseq_v2
dat = dat %>% dplyr::filter(`File type` == "bigWig", `Output type` == "fold change over control")
dat = dat %>% dplyr::mutate(Count_Replicate = fun(`Biological replicate(s)`))
dat = dat %>% dplyr::select(
    `File accession`,
    `Experiment accession`,
    `Biological replicate(s)`,
    Count_Replicate)
fun_display_table(dat)
```
