```bash
### loop through each region set
while read FOLDER_REG FNAME_REG LABEL_REG; do
    
    ### Set directory
    FD_INP_A=${FD_RES}/region/${FOLDER_REG}
    FN_INP_A=${FNAME_REG}
    FP_INP_A=${FD_INP_A}/${FN_INP_A}
    
    ### loop through each assay
    for ASSAY_NAME in ${ASSAY_NAMES[@]}; do
        for FOLDER_RES in overlap_fragment_unique overlap_fragment_one2one; do
            ### Set directory
            #FD_INP_B=${FD_RES}/${ASSAY_TYPE}/${ASSAY_NAME}/coverage/${FOLDER_REG}/${LABEL_REG}/overlap_fragment_one2one
            FD_INP_B=${FD_RES}/${ASSAY_TYPE}/${ASSAY_NAME}/coverage/${FOLDER_REG}/${LABEL_REG}/${FOLDER_RES}
            FP_INPS=($(ls ${FD_INP_B}/*hg38*rep*bed.gz))
            
            ### loop through each sample
            LEN=${#FP_INPS[@]}
            for (( IDX=0; IDX < ${LEN}; IDX++ )); do
                ### setup input file
                FP_INP_B=${FP_INPS[${IDX}]}
                FD_INP_B=$(dirname  ${FP_INP_B})
                FN_INP_B=$(basename ${FP_INP_B})

                ### setup output file
                FD_OUT=$(dirname ${FD_INP_B})
                FD_OUT=${FD_RES}/${ASSAY_TYPE}/${ASSAY_NAME}/coverage/${FOLDER_REG}/${LABEL_REG}/${FOLDER_RES}/summary
                FN_OUT=result.size.${IDX}.csv
                FP_OUT=${FD_OUT}/${FN_OUT}

                ### setup log file
                FN_LOG=region.coverage.size.${LABEL_REG}.${FOLDER_RES}.${FN_INP_B}.txt
                FP_LOG=${FD_LOG}/${FN_LOG}

                ### show progress
                echo ==============================
                echo "Region folder:" ${FOLDER_REG}
                echo "Region label: " ${LABEL_REG}
                echo "Assay  name:  " ${ASSAY_NAME}
                echo "Result folder:" ${FOLDER_RES}
                echo
                echo "Input fdiry: " ${FD_INP_B}
                echo "Input fname: " ${FN_INP_B}
                echo
                echo "Output fdiry: " ${FD_OUT}
                echo "Output fname: " ${FN_OUT}
                echo
                echo "Log file:" '${FD_LOG}/'${FN_LOG}
                echo
            done
        done
    done
done < <(cat ${FP_REGION_LABEL} | awk 'NR >=2 {print}' | grep fcc_astarr_macs)
```

## Execute

```bash
### loop through each region set
while read FOLDER_REG FNAME_REG LABEL_REG; do
    
    ### Set directory
    FD_INP_A=${FD_RES}/region/${FOLDER_REG}
    FN_INP_A=${FNAME_REG}
    FP_INP_A=${FD_INP_A}/${FN_INP_A}
    
    ### loop through each assay
    for ASSAY_NAME in ${ASSAY_NAMES[@]}; do
        for FOLDER_RES in overlap_fragment_unique overlap_fragment_one2one; do
            ### Set directory
            #FD_INP_B=${FD_RES}/${ASSAY_TYPE}/${ASSAY_NAME}/coverage/${FOLDER_REG}/${LABEL_REG}/overlap_fragment_one2one
            FD_INP_B=${FD_RES}/${ASSAY_TYPE}/${ASSAY_NAME}/coverage/${FOLDER_REG}/${LABEL_REG}/${FOLDER_RES}
            FP_INPS=($(ls ${FD_INP_B}/*hg38*rep*bed.gz))
            
            ### loop through each sample
            LEN=${#FP_INPS[@]}
            for (( IDX=0; IDX < ${LEN}; IDX++ )); do
                ### setup input file
                FP_INP_B=${FP_INPS[${IDX}]}
                FD_INP_B=$(dirname  ${FP_INP_B})
                FN_INP_B=$(basename ${FP_INP_B})

                ### setup output file
                FD_OUT=$(dirname ${FD_INP_B})
                FD_OUT=${FD_RES}/${ASSAY_TYPE}/${ASSAY_NAME}/coverage/${FOLDER_REG}/${LABEL_REG}/${FOLDER_RES}/summary
                FN_OUT=result.size.${IDX}.csv
                FP_OUT=${FD_OUT}/${FN_OUT}

                ### setup log file
                FN_LOG=region.coverage.size.${LABEL_REG}.${FOLDER_RES}.${FN_INP_B}.txt
                FP_LOG=${FD_LOG}/${FN_LOG}

                ### show progress
                echo ==============================
                echo "Region folder:" ${FOLDER_REG}
                echo "Region label: " ${LABEL_REG}
                echo "Assay  name:  " ${ASSAY_NAME}
                echo "Result folder:" ${FOLDER_RES}
                echo
                echo "Input fdiry: " ${FD_INP_B}
                echo "Input fname: " ${FN_INP_B}
                echo
                echo "Output fdiry: " ${FD_OUT}
                echo "Output fname: " ${FN_OUT}
                echo
                echo "Log file:" '${FD_LOG}/'${FN_LOG}
                echo
                
                ### execute
                mkdir -p ${FD_OUT}
                sbatch -p ${NODE} \
                    --exclude=dl-01 \
                    --cpus-per-task 4 \
                    --mem 4G \
                    --output ${FP_LOG} \
                    ${FD_EXE}/run_calulate_lib_size.sh ${FD_PRJ} ${FP_INP_B} ${FP_OUT}
                echo
            done
        done
    done
done < <(cat ${FP_REGION_LABEL} | awk 'NR >=2 {print}' | grep fcc_astarr_macs)
```
