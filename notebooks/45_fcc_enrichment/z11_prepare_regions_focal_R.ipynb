## Label regions: Promoter, Enhancer, and Repressor

```r
dat = dat_region_tss
table(dat$Label)
```

```r
dat = dat_region_tss
dat = dat %>% 
    dplyr::mutate(TSS = 1) %>%
    dplyr::select(Chrom, ChromStart, ChromEnd, Region, TSS)

dat_region_tss_arrange = dat
print(dim(dat))
head(dat)
```

```r
dat = dat_region_fcc_assayvote
table(dat$Type)
```

```r
dat = dplyr::left_join(
    dat_region_fcc_assayvote,
    dat_region_tss_arrange,
    by = c("Chrom", "ChromStart", "ChromEnd", "Region")
)
dat = dat %>% replace(is.na(.), 0)

lst = split(dat, dat$Type)

txt = "Enhancer"
dat = lst[[txt]]
dat = dat %>% dplyr::mutate(Label = ifelse(TSS > 0, "Promoter_Enhanced", Type))
lst[[txt]] = dat

txt = "Repressor"
dat = lst[[txt]]
dat = dat %>% dplyr::mutate(Label = ifelse(TSS > 0, "Promoter_Repressed", Type))
lst[[txt]] = dat

txt = "Repressor_GCFilter"
dat = lst[[txt]]
dat = dat %>% dplyr::mutate(Label = ifelse(TSS > 0, "Promoter_Repressed_GCFilter", Type))
lst[[txt]] = dat

dat = bind_rows(lst)
dat_region_label = dat
print(dim(dat))
head(dat)
```

```r
dat = dat_region_label
mat = addmargins(table(dat$Label, dat$Type))

vec = c(
    "Enhancer", "Promoter_Enhanced", 
    "Repressor", "Promoter_Repressed",
    "Repressor_GCFilter", "Promoter_Repressed_GCFilter", "Sum")
fun_display_table(mat[vec,])
```

```r
dat = dat_region_label
#dat = dat %>% dplyr::filter(Num_Assay > 1)
mat = addmargins(table(dat$Label, dat$Num_Assay))

vec = c(
    "Enhancer", "Promoter_Enhanced", 
    "Repressor", "Promoter_Repressed",
    "Repressor_GCFilter", "Promoter_Repressed_GCFilter")
fun_display_table(mat[vec,])
```

```r
dat = dat_region_label
dat = dat %>% 
    dplyr::mutate(Label2 = Label) %>%
    dplyr::filter(Num_Assay >= 2)
lst = split(dat, dat$Type)

dat = lst[["Enhancer"]]
dat = dat %>% dplyr::mutate(
    Label2 = ifelse(
        Num_Assay > 2, 
        paste0(Label, "_Vote3or4"), 
        paste0(Label, "_Vote2")
    )
)
lst[["Enhancer"]] = dat

dat = bind_rows(lst)
lst = split(dat$Region, dat$Label)
lst = lapply(lst, unique)

mat = as.data.frame(table(dat$Label2))
colnames(mat) = c("Group", "Count")
fun_display_table(mat)
```
