## Import data

```r
lst = lapply(VEC_TXT_FOLDER, function(txt_folder){
    ### set file directory
    txt_fdiry = file.path(
        FD_RES, 
        "analysis_fcc_enrichment", 
        txt_folder,
        "summary"
    )
    txt_fname = "matrix.coverage.encode_chipseq_merge.by_target.log2p.fcc_starrmpra_group.tsv"
    txt_fpath = file.path(txt_fdiry, txt_fname)

    ### read table
    dat = read_tsv(txt_fpath, show_col_types = FALSE)
    return(dat)
})

### assign and show
lst_dat_region_score_chipseq = lst

res = lapply(lst, dim)
print(res)

dat = lst[[1]]
fun_display_table(head(dat, 3))
```

```r
lst =  lst_dat_region_score_chipseq
dat = lst[["fcc_astarr_macs_input_overlap"]]

dat_region_score_chipseq = dat
print(dim(dat))
fun_display_table(head(dat))
```

```r
dat = dat_region_score_chipseq
lst = split(dat, dat$Group)
lapply(lst, nrow)
```

```r
### init
set.seed(123)
vec_txt_region = c()

### split data into groups
dat = dat_region_score_chipseq
lst = split(dat, dat$Group)

### sample region from each group
idx = "Proximal:Repressive"
dat = lst[[idx]]
vec = dat$Region
vec_txt_region = c(vec_txt_region, vec)
lst[[idx]] = dat

idx = "Distal:Repressive"
dat = lst[[idx]]
dat = dat %>% dplyr::filter(!(Region %in% vec_txt_region)) %>% dplyr::slice_sample(n = 1000)
lst[[idx]] = dat
vec = dat$Region
vec_txt_region = c(vec_txt_region, vec)

idx = "Proximal:Active"
dat = lst[[idx]]
dat = dat %>% dplyr::filter(!(Region %in% vec_txt_region)) %>% dplyr::slice_sample(n = 1500)
lst[[idx]] = dat
vec = dat$Region
vec_txt_region = c(vec_txt_region, vec)


idx = "Distal:Active"
dat = lst[[idx]]
dat = dat %>% dplyr::filter(!(Region %in% vec_txt_region)) %>% dplyr::slice_sample(n = 2000)
lst[[idx]] = dat
vec = dat$Region
vec_txt_region = c(vec_txt_region, vec)

idx = "Proximal:Inactive"
dat = lst[[idx]]
dat = dat %>% dplyr::filter(!(Region %in% vec_txt_region)) %>% dplyr::slice_sample(n = 1000)
lst[[idx]] = dat
vec = dat$Region
vec_txt_region = c(vec_txt_region, vec)

idx = "Distal:Inactive"
dat = lst[[idx]]
dat = dat %>% dplyr::filter(!(Region %in% vec_txt_region)) %>% dplyr::slice_sample(n = 1000)
lst[[idx]] = dat
vec = dat$Region
vec_txt_region = c(vec_txt_region, vec)

lst_dat_region_score_subset = lst
lapply(lst, dim)
```

```r
lst = lst_dat_region_score_subset
lst = lapply(lst, function(dat){
    dat = dat %>% 
        dplyr::select(-Chrom, -ChromStart, -ChromEnd, -Group) %>%
        tibble::column_to_rownames(var = "Region")
    return(dat)
})
lst_mat_region_score_subset = lst
lapply(lst, dim)
```

```r
lst = lst_mat_region_score_subset
dat = bind_rows(lst)

mat_region_score_subset = dat
fun_display_table(head(dat))
```

```r
lst = lst_mat_region_score_subset
lst = lapply(lst, function(dat){
    mat = cor(t(dat), use="pairwise.complete.obs")
    mat = as.dist(1 - mat)
    obj = hclust(mat, method="complete")    
    return(obj)
})
lst_obj_clust_row = lst
```

```r
lst = lst_obj_clust_row
print(names(lst))
cat("\n")

idx = c(
    "Proximal:Active", "Proximal:Repressive", "Proximal:Inactive",
    "Distal:Active",   "Distal:Repressive",   "Distal:Inactive"
)
lst = lst[idx]
print(names(lst))
cat("\n")

lst = lapply(lst, function(obj){
    vec = obj$labels[obj$order]
    return(vec)
})
vec = unlist(lst)

vec_txt_region = vec
print(length(vec))
print(head(vec))
```

```r
dat = mat_region_score_subset
dat = as.matrix(dat)

mat = cor(dat, use="pairwise.complete.obs")
mat = as.dist(1 - mat)
obj = hclust(mat, method="complete")

obj_clust_col = obj
print(obj)
```

```r
obj = obj_clust_col
vec = obj$labels[obj$order]

vec_txt_feature = vec
print(length(vec))
print(head(vec))
```

```r
dat = mat_region_score_subset

idx = vec_txt_region
dat = dat[idx,]

idx = vec_txt_feature
dat = dat[,idx]

mat_region_score_arrange = dat
print(dim(dat))
fun_display_table(head(dat))
```

```r
lst = lst_dat_region_score_subset
dat = bind_rows(lst)
dat = dat %>% 
    dplyr::select(Region, Group) %>%
    tibble::column_to_rownames(var = "Region")

dat_region_annot = dat
print(dim(dat))
fun_display_table(head(dat))
```

```r
dat = tribble(
    ~Group,                 ~Color,
    "Proximal:Active",      "#e41a1c", 
    "Distal:Active",        "#ff7f00",
    "Distal:Repressive",    "#377eb8",
    "Proximal:Repressive",  "#4daf4a", #"#f781bf",
    "Distal:Inactive",      "#bdbdbd",
    "Proximal:Inactive",    "#252525"
)
dat_annot_color = dat
```

```r
dat = dat_annot_color
vec = dat$Color
names(vec) = dat$Group
lst = list(Group = vec)

lst_txt_color = lst
```

```r
dat = mat_region_score_arrange
mat = as.matrix(dat)
dat = data.frame(values = as.numeric(mat))

gpt = ggplot(dat, aes(values)) + 
    geom_density(bw = "sj") + 
    theme_bw(base_size=16)

options(repr.plot.height=3, repr.plot.width=5)
print(gpt)
```

```r
num = 50
vec_txt_color = viridis::inferno(num)
vec_txt_break = seq(0, 7, length.out = length(vec_txt_color))
```

```r
plt = pheatmap(
    mat_region_score_arrange, 
    color  = vec_txt_color,
    breaks = vec_txt_break,
    cluster_rows      = FALSE,
    cluster_cols      = FALSE,
    annotation_row    = dat_region_annot,
    annotation_colors = lst_txt_color,
    treeheight_row = 0,
    treeheight_col = 0,
    show_rownames = FALSE, 
    show_colnames = FALSE)

### assign and show
plt_export_heatmap_condense = plt
options(repr.plot.height=10, repr.plot.width=10)
print(plt)
```

```r
plt_export = plt_export_heatmap_condense
txt_export = "fig.fcc_starrmpra_group.astarr_input_overlap.coverage.chipseq.heatmap"

txt_fdiry = "./"
txt_fname = paste(txt_export, "png", sep = ".")
txt_fpath = file.path(txt_fdiry, txt_fname)

png(txt_fpath, height = 10, width = 10, units = "in", res = 300)
print(plt_export)
dev.off()

txt_fdiry = "./"
txt_fname = paste(txt_export, "svg", sep = ".")
txt_fpath = file.path(txt_fdiry, txt_fname)

svg(txt_fpath, height = 10, width = 10)
print(plt_export)
dev.off()
```

```r
plt = pheatmap(
    mat_region_score_arrange,
    color  = vec_txt_color,
    breaks = vec_txt_break,
    cluster_rows      = FALSE,
    cluster_cols      = FALSE,
    annotation_row    = dat_region_annot,
    annotation_colors = lst_txt_color,
    treeheight_row = 0,
    treeheight_col = 0,
    show_rownames = FALSE, 
    show_colnames = TRUE
)

### assign and show
plt_export_heatmap_extend = plt
options(repr.plot.height=10, repr.plot.width=50)
print(plt)
```

```r
save_pheatmap_pdf <- function(x, filename, height=7, width=7) {
   stopifnot(!missing(x))
   stopifnot(!missing(filename))
   pdf(filename, width=width, height=height)
   grid::grid.newpage()
   grid::grid.draw(x$gtable)
   dev.off()
}
```

```r
plt_export = plt_export_heatmap_extend
txt_export = "fig.fcc_starrmpra_group.astarr_input_overlap.coverage.chipseq.heatmap"

txt_fdiry = "./"
txt_fname = paste(txt_export, "pdf", sep = ".")
txt_fpath = file.path(txt_fdiry, txt_fname)

save_pheatmap_pdf(plt_export, txt_fpath, height=10, width=70)
```
