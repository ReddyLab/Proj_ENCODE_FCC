```r
fun_str_map = function(vec_txt_input){
    vec_txt_pattern = c("Enhancer", "Silencer", "TSS:Enhancer", "TSS:Silencer")
    vec_txt_replace = c("Distal:Active", "Distal:Repressive", "Proximal:Active", "Proximal:Repressive")
    vec_txt_output = fun_str_map_match(
        vec_txt_input,
        vec_txt_pattern,
        vec_txt_replace
    )
    return(vec_txt_output)
}
```

```r
dat = dat_region_annot_fcc
vec = c("Enhance", "Repress")
dat = dat %>% 
    dplyr::filter(Type %in% vec) %>% 
    dplyr::mutate(Group = fun_str_map(Label3)) %>%
    dplyr::select(Chrom:Region, Group, Num_Assay)

dat_region_fcc_label_vote = dat
print(dim(dat))
print(table(dat$Group))
head(dat)
```

```r
dat = dat_region_fcc_label_vote
lst = split(dat, dat$Group)
lst[["Distal:Active"]]       = lst[["Distal:Active"]]       %>% dplyr::filter(Num_Assay > 2)
lst[["Distal:Repressive"]]   = lst[["Distal:Repressive"]]   %>% dplyr::filter(Num_Assay > 1)
lst[["Proximal:Active"]]     = lst[["Proximal:Active"]]     %>% dplyr::filter(Num_Assay > 2)
lst[["Proximal:Repressive"]] = lst[["Proximal:Repressive"]] %>% dplyr::filter(Num_Assay > 1)

dat = bind_rows(lst)
vec = dat$Region

dat_region_annot_fcc_subset = dat
print(table(dat$Group))
print(length(vec))
print(length(unique(vec)))
fun_display_table(head(dat))
```

## Clustering for each group

```r
dat = dat_region_annot_merge
dat = dat %>% 
    dplyr::select(Region, Group) %>% 
    dplyr::left_join(mat_coverage_chipseq_full_log2p, by = "Region")

set.seed(123)
lst = split(dat, dat$Group)
lst = lapply(lst, function(dat){
    ###
    dat = dat %>% dplyr::select(-Group) %>% column_to_rownames(var="Region")
    dat = dat %>% dplyr::slice_sample(n=100)
    print(dim(dat))
    ###
    mat = cor(t(dat), use="pairwise.complete.obs")
    mat = as.dist(1 - mat)
    obj = hclust(mat, method="complete")

    return(obj)
})

lst_obj_hclust = lst
names(lst)
```

```r
dat = dat_region_annot_merge
dat = dat %>% 
    dplyr::select(Region, Group) %>% 
    dplyr::left_join(mat_coverage_chipseq_full_log2p, by = "Region")

lst = split(dat, dat$Group)
dat = lst[[1]]
dat = dat %>% dplyr::select(-Group) %>% column_to_rownames(var="Region")
dat = dat %>% dplyr::slice_sample(n=100)
print(dim(dat))
head(dat)
```

```r
mat = cor(t(dat), use="pairwise.complete.obs")
dim(mat)
```

```r
mat = as.dist(1 - mat)
obj = hclust(mat, method="complete")
```

```r
print(length(obj$labels))
head(obj$labels)
```

```r
print(length(obj$order))
head(obj$order)
```

```r
lst = lst_obj_hclust
lst = lapply(lst, function(obj){
    vec = obj$labels[obj$order]    
    return(vec)
})

vec = c(
    lst[["Proximal:Active"]], 
    lst[["Distal:Active"]], 
    lst[["Proximal:Repressive"]], 
    lst[["Distal:Repressive"]],
    lst[["Proximal:Inactive"]], 
    lst[["Distal:Inactive"]]
)
vec_txt_region_ordered = vec
print(length(vec))
```
