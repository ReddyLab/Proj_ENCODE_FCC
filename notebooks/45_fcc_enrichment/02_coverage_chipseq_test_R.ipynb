## Import data

```r
lst = lapply(VEC_TXT_FOLDER, function(txt_folder){
    ### set file directory
    txt_fdiry = file.path(
        FD_RES, 
        "analysis_fcc_enrichment", 
        txt_folder,
        "summary"
    )
    txt_fname = "matrix.coverage.encode_chipseq_merge.by_target.log2p.fcc_starrmpra_group.tsv"
    txt_fpath = file.path(txt_fdiry, txt_fname)

    ### read table
    dat = read_tsv(txt_fpath, show_col_types = FALSE)
    return(dat)
})

### assign and show
lst_dat_region_score_chipseq = lst

res = lapply(lst, dim)
print(res)

dat = lst[[1]]
fun_display_table(head(dat, 3))
```

```r
lst =  lst_dat_region_score_chipseq
dat = lst[["fcc_astarr_macs_input_overlap"]]

dat_region_score_chipseq = dat
print(dim(dat))
fun_display_table(head(dat))
```

```r
dat = dat_region_score_chipseq
lst = split(dat, dat$Group)
lapply(lst, nrow)
```

```r
set.seed(123)
num = 100
dat = dat_region_score_chipseq
lst = split(dat, dat$Group)
lst = lapply(lst, function(dat){
    idx = sample(1:nrow(dat), size = num)
    dat = dat[idx,]
    return(dat)
})

lst_dat_region_score_subset = lst
lapply(lst, dim)
```

```r
lst = lst_dat_region_score_subset
lst = lapply(lst, function(dat){
    dat = dat %>% 
        dplyr::select(-Chrom, -ChromStart, -ChromEnd, -Group) %>%
        tibble::column_to_rownames(var = "Region")
    return(dat)
})
lst_mat_region_score_subset = lst
lapply(lst, dim)
```

```r
lst = lst_mat_region_score_subset
dat = bind_rows(lst)

mat_region_score_subset = dat
fun_display_table(head(dat))
```

```r
dat = tribble(
    ~Group,                 ~Color,
    "Proximal:Active",      "#e41a1c", 
    "Distal:Active",        "#ff7f00",
    "Distal:Repressive",    "#377eb8",
    "Proximal:Repressive",  "#4daf4a", #"#f781bf",
    "Distal:Inactive",      "#bdbdbd",
    "Proximal:Inactive",    "#252525"
)
dat_annot_color = dat
```

```r
lst = lst_mat_region_score_subset
lst = lapply(lst, function(dat){
    mat = cor(t(dat), use="pairwise.complete.obs")
    mat = as.dist(1 - mat)
    obj = hclust(mat, method="complete")    
    return(obj)
})
lst_obj_clust_row = lst
```

```r
lst = lst_obj_clust_row
print(names(lst))
cat("\n")

idx = c(
    "Proximal:Active", "Proximal:Repressive", "Proximal:Inactive",
    "Distal:Active",   "Distal:Repressive",   "Distal:Inactive"
)
lst = lst[idx]
print(names(lst))
cat("\n")

lst = lapply(lst, function(obj){
    vec = obj$labels[obj$order]
    return(vec)
})
vec = unlist(lst)

vec_txt_region = vec
print(length(vec))
print(head(vec))
```

```r
dat = mat_region_score_subset
dat = as.matrix(dat)

mat = cor(dat, use="pairwise.complete.obs")
mat = as.dist(1 - mat)
obj = hclust(mat, method="complete")

obj_clust_col = obj
print(obj)
```

```r
obj = obj_clust_col
vec = obj$labels[obj$order]

vec_txt_feature = vec
print(length(vec))
print(head(vec))
```

```r
dat = mat_region_score_subset

idx = vec_txt_region
dat = dat[idx,]

idx = vec_txt_feature
dat = dat[,idx]

mat_region_score_arrange = dat
print(dim(dat))
fun_display_table(head(dat))
```
