```r
### init
set.seed(123)

#for (txt_cluster in vec_txt_cluster){
for (txt_cluster in c("Cluster00")){
    ### create granges for targeted cluster and other clusters
    dat = dat_region_covariate_cluster
    vec = ifelse(dat$Group == txt_cluster, txt_cluster, "others")
    lst = split(dat, vec)
    lst = lapply(lst, function(dat){
        grg = GRanges(
            seqnames = dat$Chrom,
            ranges = IRanges(
                start = dat$ChromStart, 
                end   = dat$ChromEnd
            ),
            Region = dat$Region,
            ATAC   = dat$ATAC,
            pGC    = dat$pGC,
            Length = dat$Length
        )
        return(grg)
    })
    
    grg_region_target = lst[[txt_cluster]]
    grg_region_others = lst[["others"]]
    
    ### show progress and double check the numbers
    num1 = length(grg_region_target)
    num2 = length(grg_region_others)
    num3 = num1 + num2
    txt1 = paste0("#{", txt_cluster, "})", " = ", num1)
    txt2 = paste0("#{", "others",    "})", " = ", num2)
    txt3 = paste0("#{", "total",     "})", " = ", num3)

    cat(txt_cluster, "\n")
    cat(txt1, "\n")
    cat(txt2, "\n")
    cat(txt3, "\n")
    cat("\n")
    flush.console()
    
    ### sampling 01
    mgr = matchRanges(
        focal = grg_region_target,
        pool  = grg_region_others,
        covar = ~ ATAC + pGC
    )
    mgr_target_vs_others = mgr

    ### sampling 02
    mgr = matchRanges(
        focal = grg_region_target,
        pool  = grg_region_inactive,
        covar = ~ ATAC + pGC
    )
    mgr_target_vs_inactive = mgr
}
```

```r
### init
set.seed(123)

for (txt_cluster in vec_txt_cluster){
#for (txt_cluster in c("Cluster00")){
    ### create granges for targeted cluster and other clusters
    dat = dat_region_covariate_cluster
    vec = ifelse(dat$Group == txt_cluster, txt_cluster, "others")
    lst = split(dat, vec)
    lst = lapply(lst, function(dat){
        grg = GRanges(
            seqnames = dat$Chrom,
            ranges = IRanges(
                start = dat$ChromStart, 
                end   = dat$ChromEnd
            ),
            Region = dat$Region,
            ATAC   = dat$ATAC,
            pGC    = dat$pGC,
            Length = dat$Length
        )
        return(grg)
    })
    
    grg_region_target = lst[[txt_cluster]]
    grg_region_others = lst[["others"]]
    
    ### show progress and double check the numbers
    num1 = length(grg_region_target)
    num2 = length(grg_region_others)
    num3 = num1 + num2
    txt1 = paste0("#{", txt_cluster, "})", " = ", num1)
    txt2 = paste0("#{", "others",    "})", " = ", num2)
    txt3 = paste0("#{", "total",     "})", " = ", num3)

    cat(txt_cluster, "\n")
    cat(txt1, "\n")
    cat(txt2, "\n")
    cat(txt3, "\n")
    cat("\n")
    flush.console()
    
    ### sampling 01
    mgr = matchRanges(
        focal = grg_region_target,
        pool  = grg_region_others,
        covar = ~ ATAC + pGC
    )
    mgr_target_vs_others = mgr

    ### sampling 02
    mgr = matchRanges(
        focal = grg_region_target,
        pool  = grg_region_inactive,
        covar = ~ ATAC + pGC
    )
    mgr_target_vs_inactive = mgr

    ### set output file directory
    txt_fdiry = file.path(
        FD_RES, 
        "region_integration",
        "fcc_astarr_macs_input_overlap",
        "analysis_enrichment_v2",
        "fcc_starrmpracrispri_vote2_phenograph_neighbors070"
    )

    ### save results
    txt_fname = paste("MatchedGRanges", txt_cluster, "vs", "others", sep = "_")
    txt_fname = paste(txt_fname, "rds", sep = ".")
    txt_fpath = file.path(txt_fdiry, txt_fname)
    
    cat("Saved:", txt_fname, "\n")
    obj = mgr_target_vs_others
    saveRDS(obj, txt_fpath)

    ### save results
    txt_fname = paste("MatchedGRanges", txt_cluster, "vs", "inactive", sep = "_")
    txt_fname = paste(txt_fname, "rds", sep = ".")
    txt_fpath = file.path(txt_fdiry, txt_fname)
    
    cat("Saved:", txt_fname, "\n")
    obj = mgr_target_vs_inactive
    saveRDS(obj, txt_fpath)
    cat("\n")
}
```
