## Arrange

```r
dat = dat_interpro_gene
dat = dat %>% dplyr::select(starts_with("Interpro")) %>% dplyr::distinct()

dat_interpro_info = dat
print(dim(dat))
head(dat)
```

```r
lst = lst_vec_txt_tfs_filter 
lst = lapply(lst, function(vec_txt_tfs){
    ###
    vec = c("Region", vec_txt_tfs)

    ###
    dat = mat_region_annot_fcc
    dat = dat %>% dplyr::select(any_of(vec))
    dat = dat %>% column_to_rownames("Region")
    vec = apply(dat, 1, sum)

    ###
    dat = data.frame(
        Region = names(vec),
        Count  = vec
    )
    dat = dat %>% dplyr::mutate(Score = Count / max(Count))
    
    return(dat)
})

lst_dat_interpro_score = lst
```

```r
lst = lst_dat_interpro_score
head(lst[[1]])
```

```r
lst = lst_dat_interpro_score
dat = bind_rows(lst, .id = "Interpro_ID")
dat = dat %>% 
    dplyr::select(Region, Interpro_ID, Score) %>%
    tidyr::spread(Interpro_ID, Score)

dat_region_score_interpro = dat
print(dim(dat))
head(dat)
```

```r
dat = dat_region_score_fcc
lst = split(dat, dat$Assay)
names(lst)
```

```r
dat = lst[["Lenti-MPRA"]]
dat = dat %>% 
    dplyr::select(Region, Score) %>%
    dplyr::left_join(dat_region_score_interpro, by = "Region") %>% 
    column_to_rownames(var = "Region") %>%
    replace(is.na(.), 0)
    
head(dat)
```

```r
fit = lm(Score ~ ., data = dat)
res = summary(fit)
res
```

```r
dat = res$coefficients[-1,]
dat = dat %>% 
    as.data.frame %>% 
    rownames_to_column(var = "Interpro_ID") %>% 
    dplyr::mutate(Padj   = p.adjust(`Pr(>|t|)`, method = "BH")) %>%
    dplyr::mutate(Signif = ifelse(Padj < 0.01, "1", "0"))

dat = dat %>%
    dplyr::left_join(dat_interpro_info, by = "Interpro_ID")
   
dat_fit_summary = dat
print(dim(dat))
head(dat)
```

```r
dat = dat_fit_summary
dat = dat %>% 
    dplyr::arrange(Estimate) %>% 
    dplyr::mutate(Interpro_ID = factor(Interpro_ID, levels=Interpro_ID)) %>%
    dplyr::mutate(Interpro_Desc_Short = factor(Interpro_Desc_Short, levels=Interpro_Desc_Short)) %>%
    dplyr::mutate(Interpro_Desc_Long  = factor(Interpro_Desc_Long,  levels=Interpro_Desc_Long))

theme_color = scale_fill_manual(
    values = c("#fb8072", "grey50"),
    breaks = c("1", "0"),
    labels = c("Padj < 0.01", "Not significant")
)
gpt = ggplot(dat, aes(x=Estimate, y=Interpro_Desc_Short, fill = Signif)) + 
    geom_col() + 
    labs(x="Effects", y="Protein domains/families", title="Mean(STARR/MPRA) ~ Interpro Term") +
    theme_color +
    theme_cowplot() + 
    background_grid()

options(repr.plot.height=15, repr.plot.width=10)
print(gpt)
```
