```r
fun_corr = function(x, y){
    fit     = glm(y ~ x, family = "binomial")
    num_cor = cor(x, y, method = "spearman")
    num_acc = sum(round(predict(fit, type = "response")) == y) / length(y)

    res = c(num_cor, num_acc)
    names(res) = c("Correlation", "Accuracy")
    return(res)
}
```

```r
fun_corr = function(x, y){
    fit     = glm(y ~ x, family = "binomial")
    num_cor = cor(x, y, method = "spearman")
    num_acc = sum(round(predict(fit, type = "response")) == y) / length(y)

    res = c(num_cor, num_acc)
    names(res) = c("Correlation", "Accuracy")
    return(res)
}
```
