## Import data

```r
txt_fdiry = file.path(
    FD_RES, 
    "annotation", 
    "protein_interpro"
)
txt_fname = "dat.interpro_gene.mart_export.tsv"
txt_fpath = file.path(txt_fdiry, txt_fname)

dat = read_tsv(txt_fpath, show_col_types = FALSE)

dat_interpro_gene = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

```r
txt_fdiry = file.path(
    FD_RES, 
    "region_annotation", 
    "fcc_astarr_macs_input_overlap",
    "summary"
)
txt_fname = "matrix.annotation.merge.chipseq_full.fcc.tss.tsv"
txt_fpath = file.path(txt_fdiry, txt_fname)

dat = read_tsv(txt_fpath, show_col_types = FALSE)

mat_region_annot_fcc = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

```r
dat = dat_interpro_gene
dat = dat %>% dplyr::filter(!str_detect(Interpro_Desc_Long, "superfamily"))

lst = split(dat, dat$Interpro_ID)
lst = lapply(lst, function(dat){
    vec = c(dat$Ensembl_Name, dat$UniProt_Name)
    vec = unique(vec)
    return(vec)
})
lst_vec_txt_tfs_total = lst

vec_txt_tfs_annot = colnames(mat_region_annot_fcc)
lst = lapply(lst, function(vec){
    res = intersect(vec_txt_tfs_annot, vec)
    return(res)
})
lst_vec_txt_tfs_match = lst

print(length(lst))
```

```r
lst = lst_vec_txt_tfs_match
lst = lapply(lst, length)
vec = unlist(lst)

dat = data.frame(
    ID    = names(vec),
    Count = vec
)
dat = dat %>%
    dplyr::filter(Count >= 5) %>% 
    dplyr::arrange(-Count)
vec = dat$ID

lst = lst_vec_txt_tfs_match
lst = lst[vec]

lst_vec_txt_tfs_filter = lst
print(dim(dat))
print(length(lst))
head(dat)
```

```r
dat = dat_interpro_gene
dat = dat %>% dplyr::select(starts_with("Interpro")) %>% dplyr::distinct()

dat_interpro_info = dat
print(dim(dat))
head(dat)
```

```r
dat = dat_interpro_gene
dat = dat %>% 
    dplyr::group_by(Interpro_ID) %>% 
    dplyr::summarize(Count = n(), .groups = "drop")

dat_interpro_size_total = dat
print(dim(dat))
head(dat, 3)
```

```r
lst = lst_vec_txt_tfs_match
lst = lapply(lst, length)
vec = unlist(lst)

dat = data.frame(
    ID    = names(vec),
    Count = vec
)

dat_interpro_size_match = dat
print(dim(dat))
head(dat, 3)
```

```r
lst = lst_vec_txt_tfs_filter
vec = names(lst)
mat = t(combn(vec, 2))
mat = rbind(mat[,c(1,2)], mat[,c(2,1)])

dat = mat %>% as.data.frame %>% dplyr::distinct()
dat = apply(dat, 1, function(row){
    ###
    idx1 = row[1]
    idx2 = row[2]

    ###
    vec1 = lst[[idx1]]
    vec2 = lst[[idx2]]
    vec_int = intersect(vec1, vec2) 
    vec_uni = union(    vec1, vec2)

    ###
    num = length(vec_int) / length(vec_uni)
    res = c(idx1, idx2, num)
    return(res)
})
dat = as.data.frame(t(dat))
colnames(dat) = c("Term1", "Term2", "Similarity")

###
tmp = dat_interpro_info
tmp = tmp %>% dplyr::select(Interpro_ID, Interpro_Desc_Long)

colnames(tmp) = c("Term1", "Desc1")
dat = dat %>% dplyr::left_join(tmp, by = "Term1")

colnames(tmp) = c("Term2", "Desc2")
dat = dat %>% dplyr::left_join(tmp, by = "Term2")

###
tmp = dat_interpro_size_total

colnames(tmp) = c("Term1", "Size_Total1")
dat = dat %>% dplyr::left_join(tmp, by = "Term1")

colnames(tmp) = c("Term2", "Size_Total2")
dat = dat %>% dplyr::left_join(tmp, by = "Term2")

###
tmp = dat_interpro_size_match

colnames(tmp) = c("Term1", "Size_Match1")
dat = dat %>% dplyr::left_join(tmp, by = "Term1")

colnames(tmp) = c("Term2", "Size_Match2")
dat = dat %>% dplyr::left_join(tmp, by = "Term2")


dat_interpro_cor = dat
print(dim(dat))
head(dat)
```

```r
lst = lst_vec_txt_tfs_filter
vec = names(lst)

dat = dat_interpro_info
dat = dat %>% dplyr::filter(Interpro_ID %in% vec)
vec1 = dat$Interpro_ID
vec2 = dat$Interpro_Desc_Long

mat1 = matrix(NA, nrow = length(vec1), ncol = length(vec1))
rownames(mat1) = vec1
colnames(mat1) = vec1

mat2 = matrix(NA, nrow = length(vec2), ncol = length(vec2))
rownames(mat2) = vec2
colnames(mat2) = vec2

dat = dat_interpro_cor
for (idx in 1:nrow(dat)){
    vec = dat[idx,]
    num  = as.numeric(vec$Similarity)
    
    idx1 = vec$Term1
    idx2 = vec$Term2
    mat1[idx1, idx2] = num
    #mat1[idx2, idx1] = num
    mat1[idx1, idx1] = 1
    mat1[idx2, idx2] = 1
    
    idx1 = vec$Desc1
    idx2 = vec$Desc2
    mat2[idx1, idx2] = num
    #mat2[idx2, idx1] = num
    mat2[idx1, idx1] = 1
    mat2[idx2, idx2] = 1
}
```

```r
col = c("#fb8072", "white", "#80b1d3")
col = rev(col)
col = colorRampPalette(col)(20)

options(repr.plot.height=15, repr.plot.width=15)
corrplot(mat1, type="upper", order="hclust", col=col)
```
