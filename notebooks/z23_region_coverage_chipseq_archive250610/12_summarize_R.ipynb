```r
### loop through region A
for (idx in seq_along(lst_region_label_A[1])){
    
    ### get region A
    lst = lst_region_label_A[[idx]]
    txt_region_folder_A = lst$Folder
    txt_region_label_A  = lst$Label

    ### loop through region B
    for (txt_region_folder_B in vec_txt_region_folder_B){
        
        ### get directory of annotation table
        txt_fdiry = file.path(
            FD_RES, 
            "region_coverage_chipseq",
            txt_region_label_A, 
            txt_region_folder_B
        )
        txt_fname = "*tsv"
        txt_fglob = file.path(txt_fdiry, txt_fname)

        ### get annotation table files
        vec_txt_fpath = Sys.glob(txt_fglob)
        vec_txt_fname = basename(vec_txt_fpath)

        ### show progress
        cat("===========================================\n")
        cat("Read file...", "\n")
        cat("Region A:", txt_region_folder_A, "|", txt_region_label_A, "\n")
        cat("Region B:", txt_region_folder_B, "\n")
        cat("FDiry:   ", "\n")
        print(txt_fdiry)
        cat("#File:   ", length(vec_txt_fname), "\n")
        cat("FName:   ", "\n")
        print(head(vec_txt_fname))
        cat("\n")
        flush.console()

        ###
        #vec_txt_fname = head(vec_txt_fname)
        
        ### import annotation tables
        lst = lapply(vec_txt_fname, function(txt_fname){
            ### set directory
            txt_fpath = file.path(txt_fdiry, txt_fname)
            
            ### get annotation of region A and region B
            vec = str_split(txt_fname, "\\.")[[1]]
            txt_annot_A = vec[1]
            txt_annot_B = vec[2]

            ### get target and file index
            ### ex: txt_annot_B: "encode_chipseq_LARP7_ENCFF689OKC"
            vec = unlist(str_split(txt_annot_B, pattern = "_"))
            txt_target = vec[3]
            txt_index  = vec[4]
            txt_label  = paste(txt_target, txt_index, sep = "_")
            
            ### read table and add annotation labels
            dat = read_tsv(txt_fpath, col_names = vec_txt_cname, show_col_types = FALSE)
            dat = dat %>% dplyr::select(Region, Mean0)
            colnames(dat) = c("Region", txt_label)
            return(dat)
        }) 
        names(lst) = vec_txt_fname
        lst_dat = lst
        
        ### Check if there is any empty table
        cat("Check if empty table exist:", "\n")
        lst = lst_dat
        lst = lapply(lst, nrow)
        vec = unlist(lst)
        print(table(vec))
        cat("\n")
        
        ### filtering out empty table if it exist
        idx = which(vec == 0)
        
        if (length(idx) > 0){
            lst_dat = lst_dat[-idx]
            cat("Filter out:", "\n")
            print(vec[idx])
            cat("\n")
            
            cat("After filtering:", "\n")
            lst = lst_dat
            lst = lapply(lst, nrow)
            vec = unlist(lst)
            print(table(vec))
            cat("\n")
        } else {
            cat("No filtering.", "\n")
            cat("\n")
        }

        ### join columns for each chip-seq file
        fun = function(dat1, dat2){
        dat = dplyr::full_join(dat1, dat2, by = "Region")
            return(dat)
        }
        lst = lst_dat
        dat = Reduce(fun, lst)
        mat_region_score_file = dat

        ### average scores by TF
        dat = mat_region_score_file
        dat = dat %>% column_to_rownames(var = "Region")
        
        txt = colnames(dat)
        lst = str_split(txt, "_")
        lst = lapply(lst, function(vec){return(vec[1])})
        vec = unlist(lst)
        vec_txt_target = vec

        dat = apply(dat, 1, function(vec_num_row){
            vec = tapply(vec_num_row, vec_txt_target, mean)
            return(vec)
        })
        dat = t(dat)
        dat = dat %>% as.data.frame %>% rownames_to_column("Region")
        mat_region_score_tfs = dat
    } ### end loop region B
} ### end loop region A
```

```r
### loop through region A
for (idx in seq_along(lst_region_label_A)){
    
    ### get region A
    lst = lst_region_label_A[[idx]]
    txt_region_folder_A = lst$Folder
    txt_region_label_A  = lst$Label

    ### loop through region B
    for (txt_region_folder_B in vec_txt_region_folder_B){
        
        ### get directory of annotation table
        txt_fdiry = file.path(
            FD_RES, 
            "region_coverage_chipseq",
            txt_region_label_A, 
            txt_region_folder_B
        )
        txt_fname = "*tsv"
        txt_fglob = file.path(txt_fdiry, txt_fname)

        ### get annotation table files
        vec_txt_fpath = Sys.glob(txt_fglob)
        vec_txt_fname = basename(vec_txt_fpath)

        ### show progress
        cat("===========================================\n")
        cat("Read file...", "\n")
        cat("Region A:", txt_region_folder_A, "|", txt_region_label_A, "\n")
        cat("Region B:", txt_region_folder_B, "\n")
        cat("FDiry:   ", "\n")
        print(txt_fdiry)
        cat("#File:   ", length(vec_txt_fname), "\n")
        cat("FName:   ", "\n")
        print(head(vec_txt_fname))
        cat("\n")
        flush.console()

        ### import annotation tables
        lst = lapply(vec_txt_fname, function(txt_fname){
            ### set directory
            txt_fpath = file.path(txt_fdiry, txt_fname)
            
            ### get annotation of region A and region B
            vec = str_split(txt_fname, "\\.")[[1]]
            txt_annot_A = vec[1]
            txt_annot_B = vec[2]

            ### read table and add annotation labels
            dat = read_tsv(txt_fpath, col_names = vec_txt_cname, show_col_types = FALSE)
            dat = dat %>% 
                dplyr::mutate(
                    Score        = Mean0,
                    Annotation_A = txt_annot_A,
                    Annotation_B = txt_annot_B
                ) %>%
                dplyr::select(Region, Score, Annotation_A, Annotation_B)
            return(dat)
        }) 
        names(lst) = vec_txt_fname
        lst_dat = lst
        
        ### Check if there is any empty table
        cat("Check if empty table exist:", "\n")
        lst = lst_dat
        lst = lapply(lst, nrow)
        vec = unlist(lst)
        print(table(vec))
        cat("\n")
        
        ### filtering out empty table if it exist
        idx = which(vec == 0)
        
        if (length(idx) > 0){
            lst_dat = lst_dat[-idx]
            cat("Filter out:", "\n")
            print(vec[idx])
            cat("\n")
            
            cat("After filtering:", "\n")
            lst = lst_dat
            lst = lapply(lst, nrow)
            vec = unlist(lst)
            print(table(vec))
            cat("\n")
        } else {
            cat("No filtering.", "\n")
            cat("\n")
        }
        
        ### concatenate tables and assign
        lst = lst_dat
        dat = bind_rows(lst)
        dat = dat %>% 
            tidyr::separate(
                col  = Annotation_B, 
                into = c("X1", "X2", "Target", "Index_File"), 
                remove = FALSE) %>%
            dplyr::select(-X1, -X2)
        dat_region_annot_pair = dat
        
        ### set directory
        txt_fdiry = file.path(
            FD_RES, 
            "region_coverage_chipseq",
            txt_region_label_A,
            "summary"
        )
        txt_fname = paste("region", "coverage",  txt_region_folder_B, "tsv", sep = ".")
        txt_fpath = file.path(txt_fdiry, txt_fname)

        ### show progress
        cat("Save file...",  "\n")
        cat("FDiry:", txt_fdiry, "\n")
        cat("FName:", txt_fname, "\n")
        cat("\n")
        flush.console()

        ### write tables
        dir.create(txt_fdiry, showWarnings = FALSE)
        write_tsv(dat_region_annot_pair, txt_fpath)
    } ### end loop region B
} ### end loop region A
```
