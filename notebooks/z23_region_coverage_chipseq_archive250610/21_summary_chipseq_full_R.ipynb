## Arrange table and merge

```r
dat = dat_coverage_chipseq_histone
dat = dat %>% 
    dplyr::group_by(Region, Target) %>% 
    dplyr::summarise(Score = mean(Mean0), .groups = 'drop')

dat = dat %>% tidyr::spread(Target, Score) 

dat_region_score_chipseq_histone = dat
print(dim(dat))
fun_display_table(head(dat))
```

```r
dat = dplyr::inner_join(
    dat_region_score_chipseq_histone, 
    dat_region_score_chipseq_tf_by_target,
    by = "Region"
)

dat_region_score_chipseq_merge = dat
print(dim(dat))
fun_display_table(head(dat))
```

```r
txt_fdiry = file.path(
    FD_RES, 
    "region_coverage_chipseq", 
    "fcc_astarr_macs_input_overlap",
    "summary"
)
txt_fname = "matrix.coverage.encode_chipseq_full.mean0.target.tsv"
txt_fpath = file.path(txt_fdiry, txt_fname)

dat = dat_region_score_chipseq_merge
write_tsv(dat, txt_fpath)
```
