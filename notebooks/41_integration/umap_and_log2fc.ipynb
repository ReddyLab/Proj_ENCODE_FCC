```r
### save deseq object
txt_fdiry = file.path(
    FD_RES, 
    "region_coverage_fcc", 
    "fcc_astarr_macs_input_overlap",
    "STARR_ATAC_K562_Reddy_KS91",
    "overlap_count",
    "summary"
)
txt_fname = "data_deseq2.rds"
txt_fpath = file.path(txt_fdiry, txt_fname)

obj = readRDS(txt_fpath)
```

```r
obj
```

```r
res = results(obj, contrast=c("Group","Output","Input"))
dat = as.data.frame(res)
dat = dat %>% 
    rownames_to_column(var =  "Region_A") %>% 
    dplyr::mutate(ZScore    = scale(log2FoldChange)) %>%
    dplyr::mutate(ZScoreAbs = abs(ZScore)) %>%
    na.omit

dat_region_score_astarr = dat
print(dim(dat))
head(dat)
```

```r
dat = dat_region_score_astarr
gpt = ggplot(dat, aes(x=ZScore)) + 
    geom_histogram(bins = 70) +
    xlim(-5, 5) +
    theme_cowplot() +
    background_grid()

options(repr.plot.height=5, repr.plot.width=8)
print(gpt)
```

```r
dat = dplyr::inner_join(
    dat_region_annot_umap,
    dat_region_score_astarr,
    by = "Region_A"
)

dat = dat %>% 
    dplyr::arrange(ZScoreAbs) %>%
    dplyr::mutate(Region_A = factor(Region_A, levels=Region_A))

dat_region_merge = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

```r
dat = dat_region_merge
dat_back = dat 

txt = "ATAC Regions"
gpt = ggplot() + 
    geom_point(data=dat_back, aes(x=UMAP1, y=UMAP2), size=0.5, alpha=0.1) + 
    labs(title = txt) +
    theme_cowplot() +
    background_grid()

options(repr.plot.height=6, repr.plot.width=7)
print(gpt)
```

```r
vec_txt_color =  colorRampPalette(rev(brewer.pal(10, "RdYlBu")))
theme_color   = scale_colour_gradientn(colours = vec_txt_color(100), limits=c(-6, 6))
```

```r
dat = dat_region_merge
dat_back = dat 

txt = "UMAP: ATAC Regions; Color: ASTARR (6DNA4RNA)"
gpt = ggplot() + 
    geom_point(data=dat_back, aes(x=UMAP1, y=UMAP2, color=ZScore), size=0.5, alpha=0.1) + 
    labs(title = txt) +
    theme_cowplot() +
    background_grid() +
    theme_color

options(repr.plot.height=6, repr.plot.width=7)
print(gpt)
```

## CRISPRi-Growth

```r
txt_assay  = "CRISPRi_Growth_K562_Gersbach_JinWoo"
txt_region = "fcc_astarr_macs_input_overlap"
txt_fdiry  = file.path(FD_RES, "region_coverage_fcc", txt_region, txt_assay, "overlap_score", "summary")
txt_fname  = "result.score.tsv"
txt_fpath  = file.path(txt_fdiry, txt_fname)
dat = read_tsv(txt_fpath, show_col_types = FALSE)

dat_region_score_cgrowth = dat
print(dim(dat))
head(dat)
```

```r
dat = dat_region_score_cgrowth
dat = dat %>% 
    dplyr::mutate(
        Region_A  = Region,
        ZScore    = Mean,
        ZScoreAbs = abs(Mean)
    ) %>%
    dplyr::select(Region_A, ZScore, ZScoreAbs)

dat = dplyr::inner_join(
    dat_region_annot_umap,
    dat,
    by = "Region_A"
)

dat = dat %>% 
    dplyr::arrange(ZScoreAbs) %>%
    dplyr::mutate(Region_A = factor(Region_A, levels=Region_A))

dat_region_merge = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

```r
range(dat_region_merge$ZScore)
```

```r
vec_txt_color =  colorRampPalette(rev(brewer.pal(10, "RdYlBu")))
theme_color   = scale_colour_gradientn(colours = vec_txt_color(100), limits=c(-3, 3))
```

```r
dat = dat_region_merge
dat_back = dat 

txt = "UMAP: ATAC Regions; Color: CRISPRi-Growth"
gpt = ggplot() + 
    geom_point(data=dat_back, aes(x=UMAP1, y=UMAP2, color=ZScore), size=0.5, alpha=0.1) + 
    labs(title = txt) +
    theme_cowplot() +
    background_grid() +
    theme_color

options(repr.plot.height=6, repr.plot.width=7)
print(gpt)
```
