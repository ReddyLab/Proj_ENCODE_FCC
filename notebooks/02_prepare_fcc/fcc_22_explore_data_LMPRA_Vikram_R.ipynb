## Import data

**Import library design**

```r
### set directory
fdiry = file.path(FD_DAT, "processed", "MPRA_Lenti_K562_Nadav_Vikram_230621")
fname = "LentiMPRA.K562.Library_Design.csv"
fpath = file.path(fdiry, fname)

### read table
dat = read_csv(fpath, show_col_types = FALSE)

### assign and show
dat_lmpra_library = dat
print(dim(dat))
fun_display_table(head(dat))
```

**Import effect size**

```r
### set directory
fdiry = file.path(FD_DAT, "processed", "MPRA_Lenti_K562_Nadav_Vikram_230621")
fname = "LentiMPRA.K562.Log2FC.matrix.csv"
fpath = file.path(fdiry, fname)

### read table
dat = read_csv(fpath, show_col_types = FALSE)

### assign and show
dat_lmpra_score_matrix = dat
print(dim(dat))
fun_display_table(head(dat))
```

## Merge and arrange data

```r
### merge score by oligo's ID
dat = dplyr::left_join(
    dat_lmpra_score_matrix,
    dat_lmpra_library,
    by="name"
)

### arrange table
dat = dat %>% 
    dplyr::select(
        chr.hg38, 
        start.hg38, 
        stop.hg38, 
        name, 
        mean, 
        str.hg38, 
        category)
colnames(dat) = c("Chrom", "ChromStart", "ChromEnd", "Name", "Score", "Strand", "Category")

### assign and show
dat_lmpra_arrange = dat
print(dim(dat))
fun_display_table(head(dat))
```

## Explore distributions of effect sizes

```r
theme_text = theme(
    text         = element_text(size=10), #change font size of all text
    axis.text    = element_text(size=10), #change font size of axis text
    axis.title   = element_text(size=15), #change font size of axis titles
    plot.title   = element_text(size=15), #change font size of plot title
    legend.text  = element_text(size= 5), #change font size of legend text
    legend.title = element_text(size= 7), #change font size of legend text
    strip.text   = element_text(size= 5)
) #change font size of legend title
```

**Distribution of scores (Total)**

```r
### create plot
dat = dat_lmpra_score_matrix
gpt = ggplot(dat, aes(x=mean)) + 
    geom_histogram(binwidth = 0.1) + 
    labs(
        x = "Log2 Fold Change", 
        y = "Count", 
        title="Lenti-MPRA Log2FC\n(Total)") +
    theme_cowplot() +
    background_grid() +
    theme_text

### show plot
options(repr.plot.height=5, repr.plot.width=7)
print(gpt)

### plotting parameters
txt_prefix = "fig.fcc_assay.LMPRA.hist.Log2fc.total"
num_height = 10
num_width  = 14

### save plot
for (txt_fext in c("png", "svg")){
    txt_fdiry  = "./"
    txt_fname = paste(txt_prefix, txt_fext, sep = ".")
    txt_fpath = file.path(txt_fdiry, txt_fname)
    ggsave(
        txt_fpath,
        plot   = gpt,
        height = num_height,
        width  = num_width,
        units = "cm",
        dpi = 300)    
}
```

## Explore distributions of z scores

**Extract Log2FC of negative control (Shuffled sequences)**

```r
dat = dat_lmpra_arrange
lst = split(dat, dat$Category)
dat = lst[["negative control, shuffled"]]

dat_lmpra_score_negative_shuffled = dat
print(dim(dat))
head(dat)
```

**Z Score transformation**

```r
dat = dat_lmpra_score_negative_shuffled
vec = dat$Score
val_mu = mean(vec)
val_sd = sd(vec)

dat = dat_lmpra_arrange
dat = dat %>% dplyr::mutate(ZScore = (Score - val_mu) / val_sd)

dat_lmpra_zscore = dat
print(val_mu)
print(val_sd)
head(dat)
```

**Visualization**

```r
### arrange data for plotting
dat = dat_lmpra_zscore

vec = c("potential enhancer", "promoter", "intergenic (7 loci)")
dat = dat %>% dplyr::filter(Category %in% vec)

dat = dat %>% dplyr::select(Score, ZScore)
colnames(dat) = c("Log2FC", "ZScore")
dat = dat %>% tidyr::gather(Group, Score)

### create plot
gpt = ggplot(dat, aes(x=Score, fill=Group)) + 
    geom_density(alpha=0.5) + 
    geom_vline(xintercept = 0, color = "grey30", linewidth=1) +
    labs(
        y="Density",
        title="Lenti-MPRA Log2FC vs Z-Score\n(Promoters/Enhancers/Intergenic regions)") +
    theme_cowplot() +
    background_grid() +
    theme_text

### show plot
options(repr.plot.height=6, repr.plot.width=10)
print(gpt)

### plotting parameters
txt_prefix = "fig.fcc_assay.LMPRA.hist.Log2fc.zscore"
num_height = 10
num_width  = 14

### save plot
for (txt_fext in c("png", "svg")){
    txt_fdiry = "./"
    txt_fname = paste(txt_prefix, txt_fext, sep = ".")
    txt_fpath = file.path(txt_fdiry, txt_fname)
    ggsave(
        txt_fpath,
        plot   = gpt,
        height = num_height,
        width  = num_width,
        units = "cm",
        dpi = 300)    
}
```
