```r
txt_fdiry = file.path(FD_DAT, "processed", "STARR_ATAC_K562_Reddy_KS*", "summary")
txt_fname = "summary.listchr.*"
txt_fglob = file.path(txt_fdiry, txt_fname)
vec_txt_fpath = Sys.glob(txt_fglob)


vec = basename(vec_txt_fpath)
for (txt in vec){cat(txt, "\n")}
```

```r
vec_txt_pattern = c("KS91", "KS274")
vec_txt_replace = vec_txt_pattern
vec_txt_assay   = fun_str_map_detect(vec_txt_fpath, vec_txt_pattern, vec_txt_replace)

vec_txt_pattern = c("repeat", "Input", "Output")
vec_txt_replace = c("Output", "Input", "Output")
vec_txt_group   = fun_str_map_detect(vec_txt_fpath, vec_txt_pattern, vec_txt_replace)

vec_txt_pattern = paste0("rep", 1:6)
vec_txt_replace = vec_txt_pattern
vec_txt_rep     = fun_str_map_detect(vec_txt_fpath, vec_txt_pattern, vec_txt_replace)

vec_txt_pattern = c("chr_main", "rmChrM", "corrected")
vec_txt_replace = vec_txt_pattern
vec_txt_label   = fun_str_map_detect(vec_txt_fpath, vec_txt_pattern, vec_txt_replace, .default = "Original")

vec_txt_sample  = paste("ASTARR", vec_txt_assay, vec_txt_group, vec_txt_rep, vec_txt_label, sep = "_")
names(vec_txt_sample) = vec_txt_fpath

dat_meta = data.frame(
    "Assay"  = vec_txt_assay,
    "Group"  = vec_txt_group,
    "Rep"    = vec_txt_rep,
    "Label"  = vec_txt_label,
    "Sample" = vec_txt_sample
)
```

```r
lst = lapply(vec_txt_fpath, function(txt_fpath){
    ###
    vec_txt_cname = c("Chrom", "Count")

    ###
    dat = read_tsv(txt_fpath, col_names = vec_txt_cname, show_col_types = FALSE)
    return(dat)
})
names(lst) = vec_txt_sample

dat = bind_rows(lst, .id = "Sample")
dat = dat %>% dplyr::left_join(dat_meta, by="Sample")
dat = dat %>% tidyr::spread(Chrom, Count)
dat
```
