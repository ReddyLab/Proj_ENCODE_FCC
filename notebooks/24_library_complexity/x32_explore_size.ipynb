```r
VEC_TXT_ASSAY = c(
    "STARR_ATAC_K562_Reddy_KS91",
    "STARR_ATAC_K562_Reddy_KS274",
    "STARR_WHG_K562_Reddy_A001"
)
```

```r
txt_region_folder     = "fcc_astarr_macs"
vec_txt_region_label  = c("astarr_macs_input_overlap", "astarr_macs_input_union")
#vec_txt_result_folder = c("overlap_fragment_unique", "overlap_fragment_one2one")
vec_txt_result_folder = c("overlap_fragment_unique")
```

```r
lst = list()
for (txt_assay in VEC_TXT_ASSAY) {
    for (txt_region_label in vec_txt_region_label) {
        for (txt_result_folder in vec_txt_result_folder) {

            ### show progress
            cat("====================================\n")
            cat("Assay        ", txt_assay,         "\n")
            cat("Region Folder", txt_region_folder, "\n")
            cat("Region Label ", txt_region_label,  "\n")
            cat("Result Folder", txt_result_folder, "\n")
            flush.console()
            
            ###
            txt_fdiry = file.path(
                FD_RES, "assay_fcc", txt_assay, "coverage", 
                txt_region_folder, 
                txt_region_label,
                txt_result_folder,
                "summary"
            )
            txt_fname = "result.size.summary.tsv"
            txt_fpath = file.path(txt_fdiry, txt_fname)
            dat = read_tsv(txt_fpath, show_col_types = FALSE)

            ###
            txt = paste(txt_assay, txt_region_label, txt_result_folder, sep=";")
            lst[[txt]] = dat
            
            ### show progress
            fun_display_table(dat)
            flush.console()
            cat("\n")
        }
    }
}
```

```r
dat = bind_rows(lst)
dat_size = dat
head(dat)
```

```r
fun_str_map_inner = function(
    txt_string, 
    vec_txt_pattern1, 
    vec_txt_pattern2) {
    idx = stringr::str_detect(string = txt_string, pattern = vec_txt_pattern1)
    txt = vec_txt_pattern2[idx]
    return(txt)
}

fun_str_map = function(vec_txt_string, vec_txt_pattern1, vec_txt_pattern2){
    fun = function(txt_string){
        res = fun_str_map_inner(
            txt_string,
            vec_txt_pattern1,
            vec_txt_pattern2
        )
        return(res)
    }
        
    res = sapply(vec_txt_string, fun)
    return(res)
}
```

```r
dat = dat_size

vec1 = c("KS91", "KS274", "A001")
vec2 = c("(ASTARR; KS91)", "(ASTARR; KS274)", "(WSTARR)")
dat = dat %>% 
    dplyr::mutate(Assay_Label  = fun_str_map(Assay, vec1, vec2)) %>%
    dplyr::mutate(Sample_Label = paste(Sample, Assay_Label))

dat = dat %>% dplyr::mutate(Freq = Count_Total * 100 / Size_Total)

#vec = dat %>% 
#    dplyr::arrange(Assay, Sample) %>% 
#    dplyr::pull(var = "Sample_Label")
#dat = dat %>% 
#    dplyr::mutate(Sample_Label = factor(Sample_Label, levels=vec))

dat = dat %>% 
    dplyr::arrange(Assay, Sample) %>% 
    dplyr::mutate(Sample_Label = factor(Sample_Label, levels=unique(Sample_Label)))
head(dat)
```

```r
gpt = ggplot(dat, aes(x=Sample_Label, y=Freq)) +
    geom_col() +
    coord_flip() + 
    theme_cowplot() +
    background_grid() +
    facet_wrap(~Region_Label)

options(repr.plot.width = 10, repr.plot.height=8)
print(gpt)
```
