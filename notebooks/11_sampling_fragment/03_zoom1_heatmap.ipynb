```r
library(pheatmap)
```

```r
fdiry = file.path(FD_RES, "regulatory_landscape")
fname = "dat_output_ASTARR_zoom1_count_1e5.csv"
fpath = file.path(fdiry, fname)

dat_astarr_count = read_csv(fpath, show_col_types = FALSE)
head(dat_astarr_count)
```

```r
fdiry = file.path(FD_RES, "regulatory_landscape")
fname = "dat_output_ASTARR_zoom1_score_1e5.csv"
fpath = file.path(fdiry, fname)

dat_astarr_score = read_csv(fpath, show_col_types = FALSE)
head(dat_astarr_score)
```

```r
fdiry = file.path(FD_RES, "regulatory_landscape")
fname = "dat_output_TileMPRA_zoom1_count.csv"
fpath = file.path(fdiry, fname)

dat_tmpra_count = read_csv(fpath, show_col_types = FALSE)
head(dat_tmpra_count)
```

```r
fdiry = file.path(FD_RES, "regulatory_landscape")
fname = "dat_output_TileMPRA_zoom1_score.csv"
fpath = file.path(fdiry, fname)

dat_tmpra_score = read_csv(fpath, show_col_types = FALSE)
head(dat_tmpra_score)
```

## Heatmap (Count)

**Count**

```r
set.seed(123)

df1 = dat_astarr_count
df1 = df1 %>% select(starts_with("Mtf_")) %>% sample_n(1000)
print(dim(df1))

df2 = dat_tmpra_count
df2 = df2 %>% select(starts_with("Mtf_")) %>% sample_n(1000)
print(dim(df2))
```

```r
options(repr.plot.height=8, repr.plot.width=8)
pheatmap(df1, 
         color = cols(20),
         show_rownames = FALSE, 
         show_colnames = FALSE)
```

```r
options(repr.plot.height=8, repr.plot.width=8)
pheatmap(df2, 
         color = cols(20),
         show_rownames = FALSE, 
         show_colnames = FALSE)
```

**Count (Probability)**

```r
set.seed(123)

df1 = dat_astarr_count
df1 = df1 %>% select(starts_with("Mtf_")) %>% sample_n(1000)
df1 = apply(df1, 1, function(x){x/sum(x)})
df1 = t(df1)
print(dim(df1))

df2 = dat_tmpra_count
df2 = df2 %>% select(starts_with("Mtf_")) %>% sample_n(1000)
df2 = apply(df2, 1, function(x){x/sum(x)})
df2 = t(df2)
print(dim(df2))
```

```r
options(repr.plot.height=8, repr.plot.width=8)
pheatmap(df1, 
         color = cols(10),
         show_rownames = FALSE, 
         show_colnames = FALSE)
```

```r
options(repr.plot.height=8, repr.plot.width=8)
pheatmap(df2, 
         color = cols(10),
         show_rownames = FALSE, 
         show_colnames = FALSE)
```

## Heatmap (Score)

```r
set.seed(123)

df1 = dat_astarr_score
df1 = df1 %>% select(starts_with("Mtf_")) %>% sample_n(1000)
print(dim(df1))

df2 = dat_tmpra_score
df2 = df2 %>% select(starts_with("Mtf_")) %>% sample_n(1000)
print(dim(df2))
```

```r
options(repr.plot.height=8, repr.plot.width=8)
pheatmap(df1, 
         color = cols(20),
         show_rownames = FALSE, 
         show_colnames = FALSE)
```

```r
options(repr.plot.height=8, repr.plot.width=8)
pheatmap(df2, 
         color = cols(20),
         show_rownames = FALSE, 
         show_colnames = FALSE)
```

**Score (Log10 scale)**

```r
options(repr.plot.height=8, repr.plot.width=8)
pheatmap(log10(df1+1), 
         color = cols(20),
         show_rownames = FALSE, 
         show_colnames = FALSE)
```

```r
options(repr.plot.height=8, repr.plot.width=8)
pheatmap(log10(df2+1), 
         color = cols(20),
         show_rownames = FALSE, 
         show_colnames = FALSE)
```
