## Execute

**Run**

```bash
### loop through each region set
while read FOLDER_REG FNAME_REG LABEL_REG; do
    
    ### Set directory
    FD_INP_A=${FD_RES}/region/${FOLDER_REG}
    FN_INP_A=${FNAME_REG}
    FP_INP_A=${FD_INP_A}/${FN_INP_A}
    
    ### loop through each assay
    for ASSAY_NAME in ${ASSAY_NAMES[@]}; do

        ### set directory
        FDIRY=${FD_RES}/${ASSAY_TYPE}/${ASSAY_NAME}/coverage/${FOLDER_REG}/${LABEL_REG}
        
        ### set input file
        FD_INP_B=${FDIRY}/overlap_fragment_one2one
        FP_INPS=($(ls ${FD_INP_B}/*hg38*Output*rep*bed.gz))

        ### loop through each sample
        for FP_INP_B in ${FP_INPS[@]}; do
            
            ### set output file
            FN_INP_B=$(basename ${FP_INP_B})
            FD_OUT=${FDIRY}/overlap_count_one2one
            FN_OUT=${FN_INP_B}
            FP_OUT=${FD_OUT}/${FN_OUT}
            
            ### setup log file
            FN_LOG=region.coverage.overlap.one2one.count.${LABEL_REG}.${FN_OUT}.txt
            FP_LOG=${FD_LOG}/${FN_LOG}
            
            ### show progress
            echo ==============================
            echo "Region folder:" ${FOLDER_REG}
            echo "Region label: " ${LABEL_REG}
            echo "Assay  name:  " ${ASSAY_NAME}
            echo
            echo "Input file:" 
            echo ${FP_INP_B}
            echo
            echo "Output file:"
            echo ${FP_OUT}
            echo
            echo "Log file:" 
            echo '${FD_LOG}/'${FN_LOG}
            echo
            
            ### execute
            mkdir -p ${FD_OUT}
            sbatch -p ${NODE} \
                --exclude=dl-01   \
                --cpus-per-task 4 \
                --mem 30G \
                --output ${FP_LOG} \
                --export FD_PRJ=${FD_PRJ},FP_INP=${FP_INP_B},FP_OUT=${FP_OUT} \
            <<'EOF'
#!/bin/bash
### set environment
FD_EXE=${FD_PRJ}/scripts
source ${FD_EXE}/config_func.sh
source ${FD_EXE}/config_load_module_bedtools.sh

### print start message
timer_start=`date +%s`
echo "Hostname:          " $(hostname)
echo "Slurm Array Index: " ${SLURM_ARRAY_TASK_ID}
echo "Time Stamp:        " $(date +"%m-%d-%y+%T")
echo

### show I/O file
echo "Input file: ${FP_INP}"
echo
echo "show lines of file"
fun_cat ${FP_INP} | head
echo

### execute: 
### count total/unique fragments for each region from one2one mapping
zcat ${FP_INP} \
| sort -k7,7 -k8,8n -k9,9n \
| bedtools groupby \
    -g 7-9 \
    -c 4,5 \
    -o count,sum \
| gzip -c \
> ${FP_OUT}

### show I/O file
echo "Output file: ${FP_OUT}"
echo
echo "show lines of file"
fun_cat ${FP_OUT} | head
echo

### print end message
timer=`date +%s`
runtime=$(echo "${timer} - ${timer_start}" | bc -l)
echo
echo 'Done!'
echo "Run Time: $(displaytime ${runtime})"
EOF
            echo
        done
    done
done < <(cat ${FP_REGION_LABEL} | awk 'NR >=2 {print}' | grep fcc_astarr_macs)
```
