**Test loop: Region + Assay**

```bash
### loop through each region set
while read FOLDER_REG FNAME_REG LABEL_REG; do
    
    ### Set directory
    FD_INP_A=${FD_RES}/region/${FOLDER_REG}
    FN_INP_A=${FNAME_REG}
    FP_INP_A=${FD_INP_A}/${FN_INP_A}
    
    ### loop through each assay
    for ASSAY_NAME in ${ASSAY_NAMES[@]}; do

        ### Set directory
        FD_INP_B=${FD_RES}/${ASSAY_TYPE}/${ASSAY_NAME}/fragment_counts
        FP_INPS=($(ls ${FD_INP_B}/*hg38*rep*bed.gz))

        ### loop through each sample
        for FP_INP_B in ${FP_INPS[@]}; do
            ### setup input file
            FD_INP_B=$(dirname  ${FP_INP_B})
            FN_INP_B=$(basename ${FP_INP_B})
            
            ### setup output file
            FD_OUT=$(dirname ${FD_INP_B})
            FD_OUT=${FD_OUT}/coverage/${FOLDER_REG}/${LABEL_REG}/overlap_fragment
            FN_OUT=${FN_INP_B}
            FP_OUT=${FD_OUT}/${FN_OUT}
            
            ### setup log file
            FN_LOG=region.coverage.overlap.${LABEL_REG}.${FN_OUT}.txt
            FP_LOG=${FD_LOG}/${FN_LOG}
            
            ### show progress
            echo ==============================
            echo "Region folder:" ${FOLDER_REG}
            echo "Region label: " ${LABEL_REG}
            echo "Assay  name:  " ${ASSAY_NAME}
            echo
            echo "Output fdiry: " ${FD_OUT}
            echo "Output fname: " ${FN_OUT}
            echo
            echo "Log file:" ${FN_LOG}
            echo
        done
    done
    
done < <(cat ${FP_REGION_LABEL} | awk 'NR >=2 {print}' | grep fcc_astarr_macs)
```

## Execute

**RUN**

```bash
### loop through each region set
while read FOLDER_REG FNAME_REG LABEL_REG; do
    
    ### Set directory
    FD_INP_A=${FD_RES}/region/${FOLDER_REG}
    FN_INP_A=${FNAME_REG}
    FP_INP_A=${FD_INP_A}/${FN_INP_A}
    
    ### loop through each assay
    for ASSAY_NAME in ${ASSAY_NAMES[@]}; do

        ### Set directory
        FD_INP_B=${FD_RES}/${ASSAY_TYPE}/${ASSAY_NAME}/fragment_counts
        FP_INPS=($(ls ${FD_INP_B}/*hg38*rep*bed.gz))

        ### loop through each sample
        for FP_INP_B in ${FP_INPS[@]}; do
            ### setup input file
            FD_INP_B=$(dirname  ${FP_INP_B})
            FN_INP_B=$(basename ${FP_INP_B})
            
            ### setup output file
            FD_OUT=$(dirname ${FD_INP_B})
            FD_OUT=${FD_OUT}/coverage/${FOLDER_REG}/${LABEL_REG}/overlap_fragment
            FN_OUT=${FN_INP_B}
            FP_OUT=${FD_OUT}/${FN_OUT}
            
            ### setup log file
            FN_LOG=region.coverage.overlap.${LABEL_REG}.${FN_OUT}.txt
            FP_LOG=${FD_LOG}/${FN_LOG}
            
            ### show progress
            echo ==============================
            echo "Region folder:" ${FOLDER_REG}
            echo "Region label: " ${LABEL_REG}
            echo "Assay  name:  " ${ASSAY_NAME}
            echo
            echo "Output fdiry: " ${FD_OUT}
            echo "Output fname: " ${FN_OUT}
            echo
            echo "Log file:" ${FN_LOG}
            echo
            
            ### execute
            mkdir -p ${FD_OUT}
            sbatch -p ${NODE} \
                --exclude=dl-01 \
                --cpus-per-task 4 \
                --mem 4G \
                --output ${FP_LOG} \
                ${FD_EXE}/run_bedtools_intersect.sh ${FD_PRJ} ${FP_INP_B} ${FP_INP_A} ${FP_OUT}
        done
    done
    
done < <(cat ${FP_REGION_LABEL} | awk 'NR >=2 {print}' | grep fcc_astarr_macs)
```
