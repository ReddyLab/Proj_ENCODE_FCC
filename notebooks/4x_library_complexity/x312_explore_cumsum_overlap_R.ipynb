```r
vec_txt_assay = c(
    "STARR_ATAC_K562_Reddy_KS274",
    "STARR_ATAC_K562_Reddy_KS91",
    "STARR_WHG_K562_Reddy_A001"
)
vec_txt_assay
```

```r
for (txt_assay in vec_txt_assay){
    cat(txt_assay, "\n")
    txt_fdiry = file.path(
        FD_RES, 
        "assay_fcc", 
        txt_assay,
        "coverage", 
        "fcc_astarr_macs", 
        "astarr_macs_input_overlap",
        "overlap_count_one2one"
    )
    vec = dir(txt_fdiry)
    print(vec)
}
```

```r
lst = lapply(vec_txt_assay, function(txt_assay){
    txt_fdiry = file.path(
        FD_RES, 
        "assay_fcc", 
        txt_assay,
        "coverage", 
        "fcc_astarr_macs", 
        "astarr_macs_input_overlap",
        "overlap_count_one2one"
    )

    vec_txt_fname = dir(txt_fdiry)
    vec_txt_cname = c("Chrom", "ChromStart", "ChromEnd", "Count_Unique", "Count_Total")

    lst = lapply(vec_txt_fname, function(txt_fname){
        txt_fpath = file.path(txt_fdiry, txt_fname)
        dat = read_tsv(txt_fpath, col_names = vec_txt_cname, show_col_types = FALSE)
        dat = dat %>% dplyr::mutate(Region = paste0(Chrom, ":", ChromStart, "-", ChromEnd))
        #print(txt_fname)
        #print(head(dat))
        #flush.console()
        dat = dat %>% 
            dplyr::mutate(Region = factor(Region, levels=vec_txt_region)) %>%
            dplyr::arrange(Region) %>%
            dplyr::mutate(
                CumSum_Index  = 1:nrow(dat),
                CumSum_Total  = cumsum(Count_Total),
                CumSum_Unique = cumsum(Count_Unique)
            ) %>%
            dplyr::mutate(FName = txt_fname)
        return(dat)
    })
    dat = bind_rows(lst)
    return(dat)
})
dat = bind_rows(lst)

### assign and show
dat_count_cumsum = dat
print(dim(dat))
head(dat)
```

```r
dat = dplyr::left_join(
    dat_count_cumsum, 
    dat_lib_size, by="FName")
dat = dat %>%
    dplyr::mutate(
        CumFreq_Unique = CumSum_Unique / Size_Unique,
        CumFreq_Total  = CumSum_Total  / Size_Total,
        
    )
dat_count_cumsum_full = dat
head(dat)
```

```r
dat = dat_count_cumsum_full
gpt = ggplot(dat, aes(x=CumSum_Total, y=CumSum_Unique, group=FName, color=FName)) +
    geom_line() +
    labs(x = "Cumulative Fragments", y = "Cumulative Unique Fragments") +
    theme_cowplot() +
    background_grid() 
options(repr.plot.width=13, repr.plot.height=7)
print(gpt)
```
