## Arrange table

```r
fun_txt2json = function(txt){
    txt = stringr::str_replace_all(txt, pattern = "'", replacement = '"')
    lst = jsonlite::fromJSON(txt)
    return(lst)
}
```

```r
### init
dat = dat_meta_import
fun = function(vec){
    lst = lapply(vec, fun_txt2json)
    lst = lapply(lst, function(tmp){tmp$label})
    res = unlist(lst)
    return(res)
}

###
dat = dat %>% 
    dplyr::filter(`File format` == "bed narrowPeak") %>%
    dplyr::mutate(
        Target = fun(`Biosample genetic modifications targets`)
    ) %>%
    dplyr::select(Target)

### assing and show
dat_label = dat
print(dim(dat))
fun_display_table(head(dat))
```

```r
### init
dat = dat_meta_import
fun = function(vec){
    lst = lapply(vec, fun_txt2json)
    lst = lapply(lst, function(tmp){tmp$label})
    res = unlist(lst)
    return(res)
}

###
dat = dat %>% 
    dplyr::mutate(
        Biosample        = `Biosample term name`,
        Index_Experiment = `Experiment accession`,
        Index_File       = `File accession`,
        File_Type        = `File format`,
        Output_Type      = `Output type`,
        Genome           = `File assembly`,
        Target           = fun(`Biosample genetic modifications targets`),
        FName            = basename(`File download URL`),
        FUrl             = `File download URL`
    ) %>%
    dplyr::select(
        Assay, 
        Index_Experiment, 
        Index_File,
        File_Type,
        Output_Type,
        Genome,
        Target,
        #Lab,
        FName,
        FUrl)

### assing and show
dat_meta_total = dat
print(dim(dat))
fun_display_table(head(dat))
```

## Define column description
The peak file is in narrowPeak format, which is a standard six field bed with four additional fields (BED6+4 format)

```r
### set directory
txt_folder = TXT_FOLDER_REGION
txt_fdiry  = file.path(FD_RES, "region", txt_folder)
txt_fglob  = file.path(txt_fdiry, "*bed*")

### get file names
vec_txt_fpath = Sys.glob(txt_fglob)
vec_txt_fname = basename(vec_txt_fpath)

###
vec = vec_txt_fname
print(length(vec))
print(head(vec))
```

```r
txt_fpath = vec_txt_fpath[1]
dat = read_tsv(txt_fpath, col_names = FALSE, show_col_types = FALSE)
head(dat)
```

```r
### ENCODE narrowPeak: Narrow (or Point-Source) Peaks format
dat = tribble(
    ~Name,        ~Note,
    "Chrom",      "Name of the chromosome",
    "ChromStart", "The starting position of the feature in the chromosome",
    "ChromEnd",   "The ending position of the feature in the chromosome",
    "Name",       "Name given to a region; Use '.' if no name is assigned.",
    "Score",      "Indicates how dark the peak will be displayed in the browser (0-1000).",
    "Strand",     "+/- to denote strand or orientation. Use '.' if no orientation is assigned.",
    "SignalValue","Measurement of overall (usually, average) enrichment for the region.",
    "PValue",     "Measurement of statistical significance (-log10). Use -1 if no pValue is assigned.",
    "QValue",     "Measurement of statistical significance using false discovery rate (-log10). Use -1 if no qValue is assigned.",
    "Peak",       "Point-source called for this peak; 0-based offset from chromStart. Use -1 if no point-source called."
)

### assign and show
dat_cname = dat
fun_display_table(dat)
```
