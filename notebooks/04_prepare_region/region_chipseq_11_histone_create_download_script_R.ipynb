## Import metadata from reference file

````r
### set file path
txt_fdiry = file.path(FD_REF, "encode_chipseq_histone")
txt_fname = "metadata.tsv"
txt_fpath = file.path(txt_fdiry, txt_fname)

### read table
dat = read_tsv(txt_fpath, show_col_types = FALSE)

### show and assign
dat_metadata_import = dat
print(dim(dat))
fun_display_table(head(dat, 3))
````

## Arrange columns

````r
dat = dat_metadata_import
dat = dat %>% 
    dplyr::filter(`File format` %in% c("bigWig", "bed narrowPeak")) %>% 
    dplyr::mutate(`Experiment target` = str_remove(`Experiment target`, "-human")) %>%
    dplyr::select(
        `Assay`,
        `Biosample term name`,
        `Experiment accession`,
        `File analysis title`,
        `File accession`,
        `File type`,
        `Output type`,
        `File assembly`,
        `Experiment target`,
        `File download URL`
    ) 
colnames(dat) = c(
    'Assay', 'Biosample', 
    'Index_Experiment', 'Index_Process', 'Index_File', 
    'File_Type', 'Output_Type',
    'Genome', 'Target', 
    'File_URL_Download')

### assign and show
dat_metadata = dat
print(dim(dat))
fun_display_table(head(dat, 6))
````

## Generate download commands
```
wget -O FILE URL
```

````r
### init
dat = dat_metadata

### setup download file name and wget command
dat = dat %>%
    dplyr::mutate(
        File_Ext1 = ifelse(str_detect(File_Type, "bed"), "bed.gz", "bw"),
        File_Ext2 = ifelse(str_detect(File_Type, "bed"), "bed.gz", "bigWig")
    ) %>%
    dplyr::mutate(
        File_Name = paste(
            Biosample, 
            "hg38", 
            Index_Experiment, 
            Index_File,
            "ChIPseq",
            Target,
            File_Ext1, 
            sep=".")
    ) %>%
    dplyr::mutate(
        CMD = paste("wget", "--append-output=run_download.log.txt", "-O", File_Name, File_URL_Download)
    )

### add Shebang and initial commands
#dat = dat %>% dplyr::select(Assay, Biosample, Index_Experiment, Index_File, File_Name, CMD)
dat = dat %>% dplyr::select(CMD)
dat = rbind('echo -n "" > run_download.log.txt', dat)
colnames(dat) = "#!/bin/bash"

### assign and show
dat_cmd = dat
fun_display_table(dat)
````

## Save to script
Save the command lines for each row into a bash script

````r
### set output path
txt_fdiry = file.path(FD_DAT, "external", "encode_chipseq_subset")
txt_fname = "run_download.sh"
txt_fpath = file.path(txt_fdiry, txt_fname)

### save table
dir.create(txt_fdiry, showWarnings = FALSE)
dat = dat_cmd
write_tsv(dat, txt_fpath)  
````
