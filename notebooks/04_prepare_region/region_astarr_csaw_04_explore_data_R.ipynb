## Explore data

### Compare

```r
lst = list(
    "KSMerge (6Dna7Rna)" = dat_region_ksmerge,
    "KS91 (6Dna4Rna)"    = dat_region_ks91
)
dat = bind_rows(lst, .id = "Label")

###
tmp = dat %>% dplyr::mutate(Sign = "Total")
dat = dat %>% dplyr::mutate(
    Sign = ifelse(
        Log2FC > 0, 
        "Positive",
        "Negative"
    )
)
dat = bind_rows(dat, tmp)

###
dat = dat %>% dplyr::mutate(
    Signif0_3 = MinusLog10QValue >= 0.3,
    Signif0_4 = MinusLog10QValue >= 0.4,
    Signif0_5 = MinusLog10QValue >= 0.5,
    Signif1   = MinusLog10QValue >= 1,
    Signif2   = MinusLog10QValue >= 2,
    Signif3   = MinusLog10QValue >= 3
)

### assign and show
dat_region_arrange = dat
head(dat)
```

```r
dat = dat_region_arrange
dat = dat %>%
    dplyr::group_by(Label, Sign) %>%
    dplyr::summarise(
        Count          = n(),
        `-log10Q >= 2` = sum(Signif2),
        `-log10Q >= 3` = sum(Signif3),
        .groups = "drop"
    )

fun_display_table(dat)
```

### KS Merge
