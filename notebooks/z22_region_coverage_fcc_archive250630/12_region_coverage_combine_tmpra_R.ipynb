## Loop through regions and assays to create count matrix

**Test loop**

```r
### init: region folders
dat = dat_meta_region
vec = dat$Label
vec_txt_region = vec

### init: assay
txt_assay      = "MPRA_Tiling_K562_Tewhey_Hannah"
vec_txt_prefix = c("OL13", "OL43", "OL45")
vec_txt_type   = c("raw", "norm")

### execute
for (txt_region_label in vec_txt_region[1]) {

    ### set directory for region description
    txt_region_folder = lst_txt_region[[txt_region_label]]
    txt_fdiry = file.path(FD_RES, "region", txt_region_folder, "summary")
    txt_fname = "description.tsv"
    txt_fpath = file.path(txt_fdiry, txt_fname)

    ### get region description and set column names/types
    dat = read_tsv(txt_fpath, show_col_types = FALSE)
    vec = dat$Name
    vec_txt_cname = c(vec, "Count")
    vec_col_ctype = cols(Count = col_character())
    
    ### loop through each assay under each region
    for (txt_prefix in vec_txt_prefix) {
        for (txt_type in vec_txt_type) {
            ### set directory for assay metadata
            txt_fdiry = file.path(FD_RES, "assay_fcc", txt_assay, "fragment_counts", "summary")
            txt_fname = "metadata.tsv"
            txt_fpath = file.path(txt_fdiry, txt_fname)
        
            ### read metadata and setup sample names
            dat  = read_tsv(txt_fpath, show_col_types = FALSE)
            vec1 = dat$FName
            vec2 = dat$Sample
            dat_meta_assay = dat
            vec_txt_sample = setNames(vec1, vec2)
            
            ### set directory for assay counts
            txt_fdiry = file.path(
                FD_RES, "region_coverage_fcc",
                txt_region_label,
                txt_assay,
                "overlap_count"
            )
            txt_fname = paste("", txt_prefix, "hg38", txt_type, "bed.gz", sep = "*")
            txt_fglob = file.path(txt_fdiry, txt_fname)

            ### get count files
            vec_txt_fpath = Sys.glob(txt_fglob)
            vec_txt_fname = basename(vec_txt_fpath)

            ### show progress:
            cat("\n====================\n")
            cat("Region Folder:", txt_region_folder, "\n")
            cat("Region Label: ", txt_region_label,  "\n")
            cat("Assay  Folder:", txt_assay,         "\n")
            cat("Import Counts:\n")
            print(vec_txt_fname)
            cat("\n")
            flush.console()
        }
    }
}
```

**Define helper function**

```r
fun = function(dat1, dat2){
    dat = dplyr::full_join(dat1, dat2, by = c("Chrom", "ChromStart", "ChromEnd", "Region"))
    return(dat)
}
```

**Generate count matrix for each region set for each assay**

```r
### init: region folders
dat = dat_meta_region
vec = dat$Label
vec_txt_region = vec

### init: assay
txt_assay      = "MPRA_Tiling_K562_Tewhey_Hannah"
vec_txt_prefix = c("OL13", "OL43", "OL45")
vec_txt_type   = c("raw", "norm")

### loop
for (txt_region_label in vec_txt_region) {
    
    ### set directory for region description
    txt_region_folder = lst_txt_region[[txt_region_label]]
    txt_fdiry = file.path(FD_RES, "region", txt_region_folder, "summary")
    txt_fname = "description.tsv"
    txt_fpath = file.path(txt_fdiry, txt_fname)

    ### get region description and set column names/types
    dat = read_tsv(txt_fpath, show_col_types = FALSE)
    vec = dat$Name
    vec_txt_cname = c(vec, "Count")
    vec_col_ctype = cols(Count = col_character())
        
    ### loop through each assay under each region
    for (txt_prefix in vec_txt_prefix) {
        for (txt_type in vec_txt_type) {
            
            ### set directory for assay metadata
            txt_fdiry = file.path(FD_RES, "assay_fcc", txt_assay, "fragment_counts", "summary")
            txt_fname = "metadata.tsv"
            txt_fpath = file.path(txt_fdiry, txt_fname)
        
            ### read metadata and setup sample names
            dat  = read_tsv(txt_fpath, show_col_types = FALSE)
            vec1 = dat$FName
            vec2 = dat$Sample
            dat_meta_assay = dat
            vec_txt_sample = setNames(vec1, vec2)
            
            ### set directory for assay counts
            txt_fdiry = file.path(
                FD_RES, "region_coverage_fcc",
                txt_region_label,
                txt_assay,
                "overlap_count"
            )
            txt_fname = paste("", txt_prefix, "hg38", txt_type, "bed.gz", sep = "*")
            txt_fglob = file.path(txt_fdiry, txt_fname)

            ### get count files
            vec_txt_fpath = Sys.glob(txt_fglob)
            vec_txt_fname = basename(vec_txt_fpath)

            ### show progress:
            cat("\n====================\n")
            cat("Region Folder:", txt_region_folder, "\n")
            cat("Region Label: ", txt_region_label,  "\n")
            cat("Assay  Folder:", txt_assay,         "\n")
            cat("Import Counts:\n")
            print(vec_txt_fname)
            cat("\n")
            flush.console()

            ### loop through count files and import them
            lst_dat = lapply(vec_txt_fpath, function(txt_fpath){

                ### read table
                dat = read_tsv(
                    txt_fpath, 
                    col_names = vec_txt_cname, 
                    col_types = vec_col_ctype, 
                    show_col_types = FALSE)
    
                ### arrange the columns
                dat = dat %>% 
                    dplyr::select(Chrom, ChromStart, ChromEnd, Count) %>%
                    dplyr::mutate(Region = paste0(Chrom, ":", ChromStart, "-", ChromEnd)) %>%
                    dplyr::mutate(Count  = replace(Count, Count == ".", "0")) %>%
                    dplyr::mutate(Count  = as.integer(Count)) %>%
                    dplyr::select(Chrom, ChromStart, ChromEnd, Region, Count) %>%
                    dplyr::distinct()
    
                ### rename column to include the file name
                txt_fname = basename(txt_fpath)
                vec = c("Chrom", "ChromStart", "ChromEnd", "Region", txt_fname)
                colnames(dat) = vec
                
                return(dat)
            })
            
            ### combine imported counts into a matrix
            dat_cnt = Reduce(fun, lst_dat)
            dat_cnt = dplyr::rename(dat_cnt, any_of(vec_txt_sample))
            
            ### filter out unaligned reigons
            mat_cnt = dat_cnt %>%
                dplyr::select(-Chrom, -ChromStart, -ChromEnd) %>%
                column_to_rownames(var = "Region")
            vec  = apply(mat_cnt, 1, sum)
            idx1 = (vec != 0)
            
            ### re-order columns: Input ---> Output
            vec  = colnames(mat_cnt)
            vec  = sort(vec)
            idx2 = c("Chrom", "ChromStart", "ChromEnd", "Region", vec)
            
            ### get the final count matrix
            dat_cnt = dat_cnt[idx1,idx2]
    
            ### combine count matrix and column data
            dat_col = dat_meta_assay %>% dplyr::filter(FName %in% vec_txt_fname)
            lst_dat = list(
                "data_cnt" = dat_cnt,
                "data_col" = dat_col
            )

            ### show progress
            cat("Removing zero:", "\n")
            cat("Before removing zero:", nrow(mat_cnt), "\n")
            cat("After  removing zero:", nrow(dat_cnt), "\n")
            cat("\n")
            flush.console()

            ### show progress
            cat("Save count matrix and data list.\n")
            flush.console()
            
            ### set directory for saving results
            txt_fdiry = file.path(
                FD_RES, "region_coverage_fcc",
                txt_region_label,
                txt_assay,
                "overlap_count",
                "summary"
            )
            txt_cmd = paste("mkdir -p", txt_fdiry)
            system(txt_cmd)
    
            ### save count matrix
            ###     matrix.count.raw.OL13.tsv
            ###     matrix.count.raw.OL43.tsv
            ###     matrix.count.raw.OL45.tsv
            #txt_fname = "matrix.raw.count.WGS.tsv"
            txt_fname = paste("matrix", "count", txt_type, txt_prefix, "tsv", sep = ".")
            txt_fpath = file.path(txt_fdiry, txt_fname)
            cat(txt_fname, "\n")
            flush.console()
            
            dat = dat_cnt
            write_tsv(dat, txt_fpath)
    
            ### save data list
            ###     data_list.count_column.raw.OL13.tsv
            ###     data_list.count_column.raw.OL43.tsv
            #txt_fname = "data_list_count_column.rds"
            txt_fname = paste("data", "count_column", txt_type, txt_prefix, "rds", sep = ".")
            txt_fpath = file.path(txt_fdiry, txt_fname)
            cat(txt_fname, "\n")
            flush.console()
            
            obj = lst_dat
            saveRDS(obj, txt_fpath)
    
            ### show progress: matrix
            cat("\n")
            cat("First few rows of the matrix:", "\n")
            flush.console()
            fun_display_table(head(dat_cnt, 3))
            
            cat("Last few rows of the matrix:", "\n")
            flush.console()
            fun_display_table(tail(dat_cnt, 3))
            cat("\n")
            flush.console()
        }
    }
}
```
