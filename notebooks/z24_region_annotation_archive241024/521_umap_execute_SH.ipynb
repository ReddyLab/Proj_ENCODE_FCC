```bash
VEC_NUM_NEIGHBORS=(15 50 60 70 90)
VEC_TXT_METRIC=("manhattan" "cosine" "hamming" "jaccard")

FD_INP=${FD_RES}/region_annotation
FD_INP=${FD_INP}/fcc_astarr_macs_input_overlap
FD_INP=${FD_INP}/summary
FP_INPS=($(ls ${FD_INP}/matrix*prepare*tsv))

for TXT_METRIC in ${VEC_TXT_METRIC[@]}; do
    for NUM_NEIGHBORS in ${VEC_NUM_NEIGHBORS[@]}; do
        for FP_INP in ${FP_INPS[@]}; do
            PREFIX="umap.metric_${TXT_METRIC}.n_neighbors_${NUM_NEIGHBORS}"
            FN_INP=$(basename ${FP_INP})

            FD_OUT=${FD_INP}
            FN_OUT=${FN_INP/matrix/${PREFIX}}
            FP_OUT=${FD_OUT}/${FN_OUT}

            FN_LOG=run.${FN_OUT}.txt
            FP_LOG=${FD_LOG}/${FN_LOG}

            echo "Input File:"
            echo ${FN_INP}
            echo "Output File:"
            echo ${FN_OUT}
            echo "Log File:"
            echo '${FD_LOG}/'${FN_LOG}
            echo
        done
    done
done
```

## Execute

```bash
FP_EXE=${FD_EXE}/run_region_annot_umap.py
FP_APP=${FD_APP}/sing_script.sh

VEC_NUM_NEIGHBORS=(15 50 60 70 90)
VEC_TXT_METRIC=("manhattan" "cosine" "hamming" "jaccard")

FD_INP=${FD_RES}/region_annotation
FD_INP=${FD_INP}/fcc_astarr_macs_input_overlap
FD_INP=${FD_INP}/summary
FP_INPS=($(ls ${FD_INP}/matrix*prepare*chipseq_full.fcc.tsv))

for TXT_METRIC in ${VEC_TXT_METRIC[@]}; do
    for NUM_NEIGHBORS in ${VEC_NUM_NEIGHBORS[@]}; do
        for FP_INP in ${FP_INPS[@]}; do
            PREFIX="umap.metric_${TXT_METRIC}.n_neighbors_${NUM_NEIGHBORS}"
            FN_INP=$(basename ${FP_INP})

            FD_OUT=${FD_INP}
            FN_OUT=${FN_INP/matrix/${PREFIX}}
            FP_OUT=${FD_OUT}/${FN_OUT}

            FN_LOG=run.${FN_OUT}.txt
            FP_LOG=${FD_LOG}/${FN_LOG}

            echo "Input File:"
            echo ${FN_INP}
            echo "Output File:"
            echo ${FN_OUT}
            echo "Log File:"
            echo '${FD_LOG}/'${FN_LOG}
            echo
            
            sbatch -p ${NODE} \
                --exclude=dl-01 \
                --cpus-per-task 20 \
                --mem 30G \
                --output ${FP_LOG} \
                ${FP_APP} python ${FP_EXE} ${FP_INP} ${FP_OUT} ${TXT_METRIC} ${NUM_NEIGHBORS}
            echo
        done
    done
done
```

## Execute

```bash
VEC_NUM_NEIGHBORS=(15 50 60 70 90)

FD_INP=${FD_RES}/region_annotation
FD_INP=${FD_INP}/fcc_astarr_macs_input_overlap
FD_INP=${FD_INP}/summary
FP_INPS=($(ls ${FD_INP}/matrix*prepare*cre*tsv) $(ls ${FD_INP}/matrix*prepare*enh*tsv))

FP_EXE=${FD_EXE}/run_region_annot_umap.py
FP_APP=${FD_APP}/sing_script.sh

for NUM_NEIGHBORS in ${VEC_NUM_NEIGHBORS[@]}; do
    for FP_INP in ${FP_INPS[@]}; do
        PREFIX="umap.n_neighbors_${NUM_NEIGHBORS}"
        FN_INP=$(basename ${FP_INP})
        
        FD_OUT=${FD_INP}
        FN_OUT=${FN_INP/matrix/${PREFIX}}
        FP_OUT=${FD_OUT}/${FN_OUT}
        
        FN_LOG=run.${FN_OUT}.txt
        FP_LOG=${FD_LOG}/${FN_LOG}
        
        echo "Input File:"
        echo ${FN_INP}
        echo "Output File:"
        echo ${FN_OUT}
        echo "Log File:"
        echo '${FD_LOG}/'${FN_LOG}
        echo
        
        sbatch -p ${NODE} \
            --exclude=dl-01 \
            --cpus-per-task 10 \
            --mem 20G \
            --output ${FP_LOG} \
            ${FP_APP} python ${FP_EXE} ${FP_INP} ${FP_OUT} ${NUM_NEIGHBORS}
        echo
    done
done
```
