```r
dat1 = mat_annot_fcc
dat2 = mat_annot_chipseq
dat  = dplyr::left_join(dat1, dat2, by = "Region") %>% replace(is.na(.), 0)

mat_annot_full = dat
head(dat)
```

```r
dat = mat_annot_full
#dat = dat %>% dplyr::filter(str_detect(Region, "chrX"))
dat = dat %>% column_to_rownames("Region")

mat_test = dat
print(dim(dat))
head(dat)
```

```r
mat = mat_test
mat = dist(mat,  method = "manhattan")
res = hclust(mat, method = "ward.D2")

res_hcl_row = res
```

```r
#options(repr.plot.height = 7, repr.plot.width=10)
#plot(res, hang = -1, cex = 0.6)
```

```r
mat = mat_test
res = res_hcl_row

vec = rownames(mat)
idx = res$order
vec = vec[idx]

vec_txt_region = vec
head(vec)
```

```r
vec_txt_assay = c("ASTARR", "WSTARR", "LMPRA", "CRISPRi-Growth")
vec_txt_color = c(
    "Active (A)"        = "#fb8072", 
    "Repress (R)"       = "#80b1d3", 
    "Contain Both (AR)" = "#984ea3",
    "Significant"       = "black"
)
theme_fill = scale_fill_manual(values = vec_txt_color)
```

```r
dat = dat_annot_fcc
dat = dat %>% 
    #dplyr::filter(Chrom == "chrX") %>%
    dplyr::mutate(
        Group  = factor(Group,  levels = vec_txt_assay),
        Region = factor(Region, levels = vec_txt_region)
    )

gpt = ggplot(dat, aes(x=Group, y=Region, fill=Peak_Call)) +
    geom_tile() + 
    labs(x="Assay", y="") +
    theme_fill +
    theme_cowplot() +
    theme(
        axis.text.y = element_blank(),
        axis.ticks.y = element_blank()
    )
options(repr.plot.height = 8, repr.plot.width=8)
print(gpt)
```

```r
mat = mat_annot_chipseq
mat = mat %>% column_to_rownames("Region")

options(repr.plot.height = 13, repr.plot.width=13)
pheatmap(
    mat,
    cluster_rows = res_hcl_row,
    #cluster_cols = res_hcl_col,
    fontsize                 = 15,
    treeheight_row           = 70,
    treeheight_col           = 70,
    show_rownames            = FALSE,
    show_colnames            = FALSE,
    legend                   = FALSE,
    color = c("white", "black")
)
```
