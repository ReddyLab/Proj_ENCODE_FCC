## Import data

```r
### loop to import data
lst = lapply(vec_txt_folder, function(txt_folder){
    ### set file directory
    txt_fdiry = file.path(FD_RES, "region_coverage_chipseq", txt_folder, "summary")
    txt_fname = "matrix.coverage.encode_chipseq_tf_full.by_index.mean0.tsv"
    txt_fpath = file.path(txt_fdiry, txt_fname)

    ### read table
    dat = read_tsv(txt_fpath, show_col_types = FALSE)
    return(dat)
})
names(lst) = vec_txt_folder

### assign and show
lst_mat_region_score_by_index = lst

res = lapply(lst, dim)
print(res)

dat = lst[[1]]
fun_display_table(head(dat, 3))
```

## Average score by TF

```r
lst = lst_mat_region_score_by_index
lst = lapply(lst, function(dat){
    
    ### Prepare: prepare matrix
    dat = dat %>% column_to_rownames(var = "Region")
    
    ### Prepare: get TF
    ### Ex: ADNP_ENCFF852FRV -> ADNP
    txt = colnames(dat)
    lst = str_split(txt, "_")
    lst = lapply(lst, function(vec){return(vec[1])})
    vec = unlist(lst)
    vec_txt_target = vec
    
    ### Execute: average scores by TF
    dat = apply(dat, 1, function(vec_num_row){
        vec = tapply(vec_num_row, vec_txt_target, mean)
        return(vec)
    })
    dat = t(dat)
    dat = dat %>% as.data.frame %>% rownames_to_column("Region")
    
    ### setup location columns
    dat = dat %>% 
        tidyr::separate(
            col  = Region, 
            into = c("Chrom", "ChromStart", "ChromEnd"), 
            remove = FALSE)
    
    ### reorder columns
    vec = c("Chrom", "ChromStart", "ChromEnd", "Region", vec_txt_target)
    dat = dat %>% dplyr::select(!!!vec)
    
    ### assign
    return(dat)
})

### assign and show
lst_mat_region_score_by_tfs = lst

res = lapply(lst, dim)
print(res)

dat = lst[[1]]
fun_display_table(head(dat, 3))
```

## Save results

```r
### init
lst = lst_mat_region_score_by_tfs

### loop and write table
for (idx in names(lst)){
    ### set file directory
    txt_folder = idx
    txt_fdiry  = file.path(FD_RES, "region_coverage_chipseq", txt_folder, "summary")
    txt_fname  = "matrix.coverage.encode_chipseq_tf_full.by_target.mean0.tsv"
    txt_fpath  = file.path(txt_fdiry, txt_fname)

    ### write table
    dat = lst[[idx]]
    dat = dat %>% dplyr::arrange(Chrom, ChromStart, ChromEnd)
    write_tsv(dat, txt_fpath)
    cat(txt_fpath, "\n")
}
```
