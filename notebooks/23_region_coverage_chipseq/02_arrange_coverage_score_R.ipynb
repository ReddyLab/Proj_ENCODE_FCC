[Document of bigWigAverageOverBed](https://bioweb.pasteur.fr/docs/modules/BigWig_tools/v4/bigWigAverageOverBed.txt)

```
bigWigAverageOverBed - Compute average score of big wig over each bed, which may have introns.
usage:
   bigWigAverageOverBed in.bw in.bed out.tab
The output columns are:
   name - name field from bed, which should be unique
   size - size of bed (sum of exon sizes
   covered - # bases within exons covered by bigWig
   sum - sum of values over all bases covered
   mean0 - average over bases with non-covered bases counting as zeroes
   mean - average over just covered bases
Options:
   -bedOut=out.bed - Make output bed that is echo of input bed but with mean column appended
   -sampleAroundCenter=N - Take sample at region N bases wide centered around bed item, rather
                     than the usual sample in the bed item.
```

````r
### loop through region A
for (idx in seq_along(lst_region_label_A)){
    
    ### get region A
    lst = lst_region_label_A[[idx]]
    txt_region_folder_A = lst$Folder
    txt_region_label_A  = lst$Label

    ### loop through region B
    for (txt_region_folder_B in vec_txt_region_folder_B){
        
        ### get directory of annotation table
        txt_fdiry = file.path(
            FD_RES, 
            "region_coverage_chipseq",
            txt_region_label_A, 
            txt_region_folder_B
        )
        txt_fname = "*tsv"
        txt_fglob = file.path(txt_fdiry, txt_fname)

        ### get annotation table files
        vec_txt_fpath = Sys.glob(txt_fglob)
        vec_txt_fname = basename(vec_txt_fpath)

        ### show progress
        cat("===========================================\n")
        cat("Read file...", "\n")
        cat("Region A:", txt_region_folder_A, "|", txt_region_label_A, "\n")
        cat("Region B:", txt_region_folder_B, "\n")
        cat("FDiry:   ", "\n")
        cat(txt_fdiry, "\n")
        cat("FName:   ", "\n")
        print(vec_txt_fname)
        cat("\n")
        flush.console()

        ### import annotation tables
        lst = lapply(vec_txt_fname, function(txt_fname){
            ### set directory
            txt_fpath = file.path(txt_fdiry, txt_fname)
            
            ### get annotation of region A and region B
            vec = str_split(txt_fname, "\\.")[[1]]
            txt_annot_A = vec[1]
            txt_annot_B = vec[2]

            ### read table and add annotation labels
            dat = read_tsv(txt_fpath, col_names = vec_txt_cname, show_col_types = FALSE)
            dat = dat %>% 
                dplyr::mutate(
                    Annotation_A = txt_annot_A,
                    Annotation_B = txt_annot_B
                )
            return(dat)
        }) 

        ###
        lst = lapply(lst, function(dat){
            ### setup location columns
            dat = dat %>% 
                tidyr::separate(
                    col  = Region, 
                    into = c("Chrom", "ChromStart", "ChromEnd"), 
                    remove = FALSE)
            
            ### setup target and label columns
            dat = dat %>% 
                tidyr::separate(
                    col  = Annotation_B, 
                    into = c("X1", "X2", "Target", "Index_File"), 
                    remove = FALSE) %>%
                dplyr::mutate(Index_Label = paste(Target, Index_File, sep = "_"))

            ### clean up
            dat = dat %>%
                dplyr::select(-X1, -X2) %>%
                dplyr::distinct()
            return(dat)
        })
        lst_region_coverage = lst

        ### get the value I need; here I am taking mean0
        lst = lst_region_coverage
        lst = lapply(lst, function(dat){
            dat = dat %>% 
                dplyr::mutate(Value = Mean0) %>%
                dplyr::select(Chrom, ChromStart, ChromEnd, Region, Value, Target, Index_Label) %>% 
                dplyr::distinct()
            return(dat)
        })
        dat = bind_rows(lst)
        dat_region_coverage_mean0_by_index = dat

        ### set file directory
        txt_fdiry = file.path(
            FD_RES, 
            "region_coverage_chipseq",
            txt_region_label_A,
            "summary"
        )
        txt_fname = paste("region", "coverage",  txt_region_folder_B, "by_index", "mean0", "tsv", sep = ".")
        txt_fpath = file.path(txt_fdiry, txt_fname)

        ### show progress
        cat("Save file...",  "\n")
        cat("FDiry:", txt_fdiry, "\n")
        cat("FName:", txt_fname, "\n")
        cat("\n")
        flush.console()
        
        ### write tables
        dir.create(txt_fdiry, showWarnings = FALSE)
        dat = dat_region_coverage_mean0_by_index
        write_tsv(dat, txt_fpath)
    } ### end loop region B
} ### end loop region A
````
