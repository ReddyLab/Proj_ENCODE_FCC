```r
set.seed(123)
fun = function(x){as.numeric(scale(x))}
x = rnorm(100)
y = scale(x)
z = fun(x)
print(head(x))
print(head(y))
print(head(z))
```

```r
fun = function(x){as.numeric(scale(x))}
dat = mat_coverage_chipseq_full_log2p
dat = dat %>% dplyr::mutate(across(where(is.numeric), fun))
mat_coverage_chipseq_full_zscore = dat
head(dat, 3)
```

```r
fun = function(x){
    med = median(x)
    mad = median(abs(x-med))
    z   = (0.6745 * (x - med)) / mad
    return(z)
}
dat = mat_coverage_chipseq_full_log2p
dat = dat %>% dplyr::mutate(across(where(is.numeric), fun))
mat_coverage_chipseq_full_zmod = dat
```
