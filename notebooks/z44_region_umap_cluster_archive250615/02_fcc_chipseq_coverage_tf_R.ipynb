### Merge the matrix

```r
dat = dplyr::inner_join(
    dat_region_score_chipseq_histone, 
    dat_region_score_chipseq_tf_subset,
    by = "Region"
)
mat = dat %>% column_to_rownames(var = "Region")

dat_region_score_chipseq_merge = dat
mat_region_score_chipseq_merge = mat
print(dim(dat))
head(dat)
```

## Sample regions

```r
fun = function(x){log2(x+1)}
mat = mat_region_score_chipseq_merge
mat = fun(mat)

set.seed(123)
idx = sample(1:nrow(mat), size=10000)
mat = mat[idx,]

mat_region_score_chipseq_merge_subset = mat
print(dim(mat))
head(mat, 3)
```

```r
mat = mat_region_score_chipseq_merge_subset
tmp = dist(mat)
obj = hclust(tmp)
obj_tree_row = obj
```

```r
mat = mat_region_score_chipseq_merge_subset
tmp = t(mat)
tmp = dist(tmp)
obj = hclust(tmp)
obj_tree_col = obj
```

```r
obj = obj_tree_col
names(obj)
```

```r
obj = obj_tree_col
idx = obj$order
vec = obj$labels
vec = vec[idx]

vec_txt_label_ordered = vec
for (txt in vec){
    cat(txt, "\n")
}
```

## Arrange region annotation label

```r
obj = obj_tree_row
vec = cutree(obj, k = 20)
dat = data.frame(Region = names(vec), Cluster = vec)
dat = dat %>% dplyr::mutate(Cluster = paste("Cluster", Cluster))

vec_txt_region_sample = unique(dat$Region)
dat_region_cluster    = dat
print(dim(dat))
head(dat)
```

```r
dat = dat_region_label
dat = dat %>% 
    dplyr::mutate(Value = "1") %>%
    tidyr::spread(Label, Value) 

dat = dat %>% 
    dplyr::full_join(dat_region_cluster, by = "Region") %>%
    dplyr::filter(Region %in% vec_txt_region_sample)

dat = dat %>% column_to_rownames(var = "Region")
dat_annot_row = dat
head(dat)
```

## Visualize the heatmap

```r
save_pheatmap_pdf = function(x, filename, width=7, height=7) {
   stopifnot(!missing(x))
   stopifnot(!missing(filename))
   pdf(filename, width=width, height=height)
   grid::grid.newpage()
   grid::grid.draw(x$gtable)
   dev.off()
}
```

```r
lst_annot_colors = list(
    Promoter = c("1"="black"),
    Enhancer = c("1"="black"),
    Silencer = c("1"="black")
)
```

```r
mat = mat_region_score_chipseq_merge_subset

options(repr.plot.width = 20, repr.plot.height = 10)
plt = pheatmap(
    mat, 
    annotation_row = dat_annot_row,
    annotation_colors = lst_annot_colors,
    color  = myColor,
    breaks = myBreaks,
    cluster_rows = obj_tree_row,
    cluster_cols = obj_tree_col,
    treeheight_row = 50, 
    treeheight_col = 0,
    show_rownames = FALSE,
    fontsize_col  = 7
)
```

```r
save_pheatmap_pdf(
    plt, 
    "heatmap_fcc_chipseq_region_10k.pdf",
    width=25, 
    height=10
)
```
