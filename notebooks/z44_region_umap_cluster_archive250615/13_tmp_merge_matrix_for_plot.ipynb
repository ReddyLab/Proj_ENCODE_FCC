## Import other variables

```r
txt_fdiry = file.path(
    FD_RES, 
    "region_annotation", 
    "fcc_astarr_macs_input_overlap",
    "summary"
)
vec = dir(txt_fdiry)
for (txt in vec){cat(txt, "\n")}
```

```r
txt_fdiry = file.path(
    FD_RES, 
    "region_annotation", 
    "fcc_astarr_macs_input_overlap",
    "summary"
)
txt_fname = "matrix.annotation.encode_chromatin_states.full.tsv"
txt_fpath = file.path(txt_fdiry, txt_fname)

dat = read_tsv(txt_fpath, show_col_type = "FALSE")

mat_region_annot_chromatin_states = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

```r
txt_fdiry = file.path(
    FD_RES, 
    "region_nuc", 
    "fcc_astarr_macs",
    "summary"
)
txt_fname = "K562.hg38.ASTARR.macs.KS91.input.rep_all.max_overlaps.q5.tsv"
txt_fpath = file.path(txt_fdiry, txt_fname)

dat = read_tsv(txt_fpath, show_col_types = FALSE)

dat_region_nuc = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

```r
dat = dat_region_nuc
dat = dat %>% 
    dplyr::left_join(mat_region_annot_chromatin_states, by = c("Chrom", "ChromStart", "ChromEnd", "Region")) %>%
    dplyr::left_join(mat_coverage_chipseq_full_log2p_fcc_zscore_tss, by = "Region")

dat_region_merge_chipseq_log2p = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

```r
txt_fdiry = file.path(
    FD_RES, 
    "analysis_umap_cluster_by_coverage_score", 
    "fcc_astarr_macs_input_overlap",
    "summary"
)
txt_fname = "matrix.merge.nuc.encode_chromatin_states.encode_chipseq_full_log2p.fcc_zscore.tss.tsv"
txt_fpath = file.path(txt_fdiry, txt_fname)

dat = dat_region_merge_chipseq_log2p

write_tsv(dat, txt_fpath)
```

```r
txt_fdiry = file.path(
    FD_RES, 
    "analysis_umap_cluster_by_coverage_score", 
    "fcc_astarr_macs_input_overlap",
    "summary"
)
txt_fname = "matrix.merge.encode_chipseq_full_log2p.fcc_zscore.tss.encode_chromatin_states.tsv"
txt_fpath = file.path(txt_fdiry, txt_fname)

dat = mat_coverage_chipseq_full_log2p_fcc_zscore_tss
dat = dat %>% 
    dplyr::left_join(mat_region_annot_chromatin_states, by = "Region") %>%
    replace(is.na(.), 0)
print(dim(dat))

write_tsv(dat, txt_fpath)
```
