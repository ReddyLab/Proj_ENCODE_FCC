```r
set.seed(123)
n_perm = 10000

# Sizes
n1 = length(vec1)  # Distal
n2 = length(vec2)  # Proximal:Non-Essential
n3 = length(vec3)  # Proximal:Essential

# Combine all values
vec_comb = c(vec1, vec2, vec3)

# Observed value
obs_diff = (median(vec3) - median(vec1)) - (median(vec2) - median(vec1))

# Permutation test: shuffle all and partition into 3 groups of same sizes
perm_diffs = replicate(n_perm, {
  permuted = sample(vec_comb)
  p1 = permuted[1:n1]
  p2 = permuted[(n1 + 1):(n1 + n2)]
  p3 = permuted[(n1 + n2 + 1):(n1 + n2 + n3)]
  
  (median(p3) - median(p1)) - (median(p2) - median(p1))
})

# Empirical p-value
p_val = mean(abs(perm_diffs) >= abs(obs_diff))
cat("Observed contrast diff:", obs_diff, "\nEmpirical p-value:", p_val, "\n")
```

```r
set.seed(123)
n_boot = 10000

# Vectors
x = vec3  # Proximal:Essential
y = vec2  # Proximal:Non-Essential

# Bootstrap resampling
boot_diffs = replicate(n_boot, {
  x_boot = sample(x, replace = TRUE)
  y_boot = sample(y, replace = TRUE)
  median(x_boot) - median(y_boot)
})

# Bootstrap CI
ci = quantile(boot_diffs, probs = c(0.025, 0.5, 0.975))

# Summary
cat("Bootstrap 95% CI for median difference:\n")
print(ci)
```

```r
library(ggplot2)

df_boot = data.frame(diff = boot_diffs)

ggplot(df_boot, aes(x = diff)) +
  geom_histogram(bins = 60, fill = "skyblue", color = "black") +
  geom_vline(xintercept = ci[2], color = "red", size = 1.2) +
  geom_vline(xintercept = ci[1], color = "darkred", linetype = "dashed") +
  geom_vline(xintercept = ci[3], color = "darkred", linetype = "dashed") +
  labs(
    title = "Bootstrap Distribution of Median Difference",
    x = "Median(Proximal:Essential) − Median(Proximal:Non-Essential)",
    y = "Frequency"
  ) +
  theme_minimal()
```

```r
set.seed(123)
n_perm = 1000
delta_obs = fun_cliffs_delta(vec1, vec3) - fun_cliffs_delta(vec1, vec2)

perm_deltas = replicate(n_perm, {
  # Shuffle group labels (between vec2 and vec3 only)
  vec_comb = c(vec2, vec3)
  grp_labels = sample(rep(c("B", "C"), times = c(length(vec2), length(vec3))))
  
  vecB = vec_comb[grp_labels == "B"]
  vecC = vec_comb[grp_labels == "C"]
  
  delta_perm = fun_cliffs_delta(vec1, vecC) - fun_cliffs_delta(vec1, vecB)
  return(delta_perm)
})

# Compute p-value
p_val = mean(abs(perm_deltas) >= abs(delta_obs))

# Plot
library(ggplot2)
df_perm = data.frame(delta_diff = perm_deltas)
ggplot(df_perm, aes(x = delta_diff)) +
  geom_histogram(bins = 50, fill = "gray70", color = "black") +
  geom_vline(xintercept = delta_obs, color = "red", size = 1.2) +
  labs(title = "Permutation Test: Difference in Effect Sizes",
       x = "Delta(Cliff’s): Essential - Non-Essential",
       y = "Count") +
  theme_minimal()

# Print result
cat("Observed diff:", delta_obs, "\nEmpirical p-value:", p_val, "\n")
```

```r
# Manual Cliff's delta
fun_cliffs_delta = function(x, y) {
  n_x = length(x)
  n_y = length(y)
  
  # All pairwise comparisons
  cmp = outer(x, y, FUN = function(a, b) sign(a - b))
  
  # Count comparisons
  num_greater = sum(cmp == 1)
  num_less    = sum(cmp == -1)
  
  delta = (num_greater - num_less) / (n_x * n_y)
  return(delta)
}

# Calculate
delta = fun_cliffs_delta(vec1, vec2)
print(delta)

```

```r
###
txt_assay = "CRISPRi-Growth"

###
dat = dat_region_merge_crispri
dat = dat %>% 
    dplyr::filter(Assay_Label == txt_assay) %>%
    dplyr::filter(Label == "Screen:Signif") 

###
gpt = ggplot(dat, aes(x=Zcore, y=Assay_Label, fill=Group)) +
    geom_vline(xintercept = 0) +
    geom_boxplot(coef = 2) +
    labs(x = "Z Score", y = "", fill = "STARR/MPRA CRE Group") +
    theme_cowplot() +
    background_grid() +
    theme_color +
    theme_text +
    xlim(-5, 10)

gpt_export_crispri_growth = gpt
options(repr.plot.height=3, repr.plot.width=10)
print(gpt)
```
