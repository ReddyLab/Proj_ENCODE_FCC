## Clustering

```r
mat = mat_region_tfmodule
mat = dist(mat, method = "manhattan")
res = hclust(mat, method = "ward.D2")

res_hclust_module = res
print(res)
```

```r
mat = mat_region_tfmodule
mat = t(mat)
mat = dist(mat, method = "manhattan")
res = hclust(mat, method = "ward.D2")

res_hclust_tf = res
print(res)
```

```r
options(repr.plot.height=20, repr.plot.width=20)

dat = mat_region_tfmodule
plt = pheatmap(
    dat,
    cluster_rows = res_hclust_module,
    cluster_cols = res_hclust_tf,
    color = c("white", "black"),
    breaks = c(0, 0.5, 1),
    legend = FALSE,
    fontsize_col = 5
)

plt_module_tf = plt
```

```r
dat = dplyr::left_join(
    dat_region_score,
    dat_region_annot, 
    by = c("Chrom", "ChromStart", "ChromEnd", "Region"),
    relationship = "many-to-many"
)

dat = dat %>% na.omit

dat_region_merge = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

```r
dat = dat_region_merge
dat = dat %>% 
    dplyr::group_by(Assay_Type, Assay_Group, Assay_Label, Label) %>%
    dplyr::summarise(
        Score   = mean(Score),
        .groups = "drop"
    )
dat = dat %>% dplyr::mutate(Assay = Assay_Label)

dat_region_summary = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

```r
dat = dat_region_summary
dat = dat %>%
    dplyr::select(Assay, Score, Label) %>%
    tidyr::spread(Assay, Score)

mat_region_summary = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

```r
mat = mat_region_summary
mat = mat %>% tibble::column_to_rownames("Label")
mat = t(mat)
mat = dist(mat, method = "euclidean")
res = hclust(mat, method = "ward.D2")

res_hclust_assay = res
print(res)
```

```r
options(repr.plot.height=7, repr.plot.width=10)
res = res_hclust_assay
plot(res)
```

```r
mat = mat_region_summary
mat = mat %>% tibble::column_to_rownames("Label")
#mat = t(mat)
mat = cor(mat, use = "pairwise.complete.obs", method = "pearson")
mat = as.dist(1-mat)
res = hclust(mat, method = "ward.D2")

#res_hclust_assay = res
plot(res)
```

```r
res = res_hclust_assay
idx = res$order
vec = res$label[idx]

vec_txt_assay = vec
```

```r
res = res_hclust_module
idx = res$order
vec = res$label[idx]

vec_txt_module = vec
```

```r
vec_txt_color = colorRampPalette(rev(brewer.pal(10, "RdBu")))
theme_color   = scale_fill_gradientn(colours = vec_txt_color(50), limits=c(-1.1, 1.1), oob=scales::squish)
theme_text    = theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1, size=20))
```

```r
dat = dat_region_summary
dat = dat %>% 
    dplyr::mutate(Assay = factor(Assay, levels=rev(vec_txt_assay))) %>%
    dplyr::mutate(Label = factor(Label, levels=rev(vec_txt_module)))
                
gpt = ggplot(dat, aes(x=Assay, y=Label, fill=Score)) + 
    geom_tile() + 
    labs(x="", y="") +
    theme_cowplot() + 
    background_grid() +
    theme_color +
    theme_text

gpt_module_score = gpt
```

```r
options(repr.plot.height=20, repr.plot.width=10)
gpt = gpt_module_score
print(gpt)
```
