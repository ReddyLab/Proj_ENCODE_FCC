```r
dat = dat_region_annot_fcc
dat = dat %>% 
    dplyr::mutate(
        Assay_Type = Group,
        Peak_Call  = Label3
    ) %>%
    dplyr::select(Chrom, ChromStart, ChromEnd, Region, Assay_Type, Peak_Call)

dat = dat %>% 
    dplyr::inner_join(
    dat_region_score,
    by = c("Chrom", "ChromStart", "ChromEnd", "Region", "Assay_Type"),
    relationship = "many-to-many"
)

dat = dat %>% 
    dplyr::inner_join(
    dat_region_annot_module,
    by = c("Chrom", "ChromStart", "ChromEnd", "Region"),
    relationship = "many-to-many"
)

dat_region_merge = dat
print(dim(dat))
head(dat, 3)
```

```r
dat = dat_region_merge
dat = dat %>% 
    dplyr::group_by(Assay_Type, Peak_Call, Assay_Group, Assay_Label, Label) %>%
    dplyr::summarise(
        Score   = mean(Score),
        .groups = "drop"
    )
dat = dat %>% dplyr::mutate(Assay = Assay_Label)

dat_region_summary = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

```r
dat = dat_region_summary
lst = split(dat, dat$Peak_Call)
lst_region_summary = lst
names(lst)
```

```r
vec_txt_assay2= c(
    "Tiling-MPRA", "WHG-STARR", "ATAC-STARR", "Lenti-MPRA", 
    "CRISPRi-Growth", 
    "CRISPRi-HCR FlowFISH"
)
vec_txt_color = colorRampPalette(rev(brewer.pal(10, "RdYlBu")))
theme_color   = scale_fill_gradientn(colours = vec_txt_color(50), limits=c(-2.0, 2.0), oob=scales::squish)
theme_text    = theme(axis.text.x = element_text(angle = 90, vjust = 0.5, hjust=1, size=20))
```

```r
dat = dat_region_summary_sub
dat = dat %>% 
    dplyr::mutate(Assay = factor(Assay, levels=vec_txt_assay2)) %>%
    dplyr::mutate(Label = factor(Label, levels=rev(vec_txt_module)))
                
gpt = ggplot(dat, aes(x=Assay, y=Label, fill=Score)) + 
    geom_tile() + 
    labs(x="", y="") +
    theme_cowplot() + 
    background_grid() +
    theme_color +
    theme_text

gpt_module_score = gpt
```

```r
options(repr.plot.height=20, repr.plot.width=10)
gpt = gpt_module_score
print(gpt)
```
