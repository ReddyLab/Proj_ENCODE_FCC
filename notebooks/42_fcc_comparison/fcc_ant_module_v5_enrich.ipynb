```r
library(pheatmap)
```

```r
mat = mat_region_tfmodule
mat = dist(mat, method = "manhattan")
res = hclust(mat, method = "ward.D2")

res_hclust_module = res
print(res)
```

```r
mat = mat_region_tfmodule
mat = t(mat)
mat = dist(mat, method = "manhattan")
res = hclust(mat, method = "ward.D2")

res_hclust_tf = res
print(res)
```

```r
res = res_hclust_tf
idx = res$order
vec = res$label[idx]

vec_txt_tf = vec
head(vec)
```
