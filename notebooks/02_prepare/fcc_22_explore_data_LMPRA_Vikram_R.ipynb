## Import data

**Import library design**

```r
### set directory
fdiry = file.path(FD_DAT, "processed", "MPRA_Lenti_K562_Nadav_Vikram_230621")
fname = "LentiMPRA.K562.Library_Design.csv"
fpath = file.path(fdiry, fname)

### read table
dat = read_csv(fpath, show_col_types = FALSE)

### assign and show
dat_lmpra_library = dat
print(dim(dat))
fun_display_table(head(dat))
```

**Import effect size**

```r
### set directory
fdiry = file.path(FD_DAT, "processed", "MPRA_Lenti_K562_Nadav_Vikram_230621")
fname = "LentiMPRA.K562.Log2FC.matrix.csv"
fpath = file.path(fdiry, fname)

### read table
dat = read_csv(fpath, show_col_types = FALSE)

### assign and show
dat_lmpra_score_matrix = dat
print(dim(dat))
fun_display_table(head(dat))
```

## Merge and arrange data

```r
### merge score by oligo's ID
dat = dplyr::left_join(
    dat_lmpra_score_matrix,
    dat_lmpra_library,
    by="name"
)

### arrange table
dat = dat %>% 
    dplyr::select(
        chr.hg38, 
        start.hg38, 
        stop.hg38, 
        name, 
        mean, 
        str.hg38, 
        category)
colnames(dat) = c("Chrom", "ChromStart", "ChromEnd", "Name", "Score", "Strand", "Category")

### assign and show
dat_lmpra_arrange = dat
print(dim(dat))
fun_display_table(head(dat))
```

## Explore distributions of effect sizes

**Distribution of scores (Total)**

```r
dat = dat_lmpra_score_matrix
gpt = ggplot(dat, aes(x=mean)) + 
    geom_histogram(binwidth = 0.1) + 
    labs(
        x = "Log2FC", 
        y = "Count", 
        title="Lenti-MPRA Log2FC (Total)") +
    theme_cowplot() +
    background_grid()

options(repr.plot.height=5, repr.plot.width=7)
print(gpt)
```

**Distribution of scores (Negative control; random shuffled seqeunces)**

```r
dat = dat_lmpra_score_matrix
dat = dat %>% dplyr::filter(str_detect(name, "shuffle"))

gpt = ggplot(dat, aes(x=mean)) + 
    geom_histogram(binwidth = 0.1) + 
    labs(
        x = "Log2FC", 
        y = "Count", 
        title="Lenti-MPRA Log2FC (Negative Control, shuffled sequences)") +
    theme_cowplot() +
    background_grid()

options(repr.plot.height=5, repr.plot.width=7)
print(gpt)
```

**Distribution of scores for each category**

```r
dat = dat_lmpra_arrange
table(dat$Category)
```

```r
dat = dat_lmpra_arrange

dat = dat %>% 
    dplyr::select(Name, Score, Strand, Category) %>%
    dplyr::mutate(
        Category = recode(
            Category, 
            "negative control, shuffled" = "negative control\n(shuffled)",
            "negative control (Ernst et al 2016)" = "negative control\n(Ernst et al 2016)",
            "positive control (Ernst et al 2016)" = "positive control\n(Ernst et al 2016)"
        ))

head(dat)
```

```r
gpt = ggplot(dat, aes(x=Score, group=Category, fill=Category)) +
    geom_density(alpha=0.5) +
    labs(
        title="Lenti-MPRA Log2FC", 
        x="Log2FC", 
        y="Density") +
    theme_cowplot() +
    theme(
        panel.grid.major.y = element_line(color = "grey70", linewidth = 0.5, linetype = 1),
        panel.grid.major.x = element_line(color = "grey70", linewidth = 0.5, linetype = 1),
        panel.grid.minor.x = element_line(color = "grey70", linewidth = 0.5, linetype = 1)
    ) +
    facet_grid(Category~.)

options(repr.plot.height=10, repr.plot.width=8)
print(gpt)
```

## Explore distributions of z scores

**Extract Log2FC of negative control (Shuffled sequences)**

```r
dat = dat_lmpra_arrange
lst = split(dat, dat$Category)
dat = lst[["negative control, shuffled"]]

dat_lmpra_score_negative_shuffled = dat
print(dim(dat))
head(dat)
```

**Z Score transformation**

```r
dat = dat_lmpra_score_negative_shuffled
vec = dat$Score
val_mu = mean(vec)
val_sd = sd(vec)

dat = dat_lmpra_arrange
dat = dat %>% dplyr::mutate(ZScore = (Score - val_mu) / val_sd)

dat_lmpra_zscore = dat
print(val_mu)
print(val_sd)
head(dat)
```

**Visualization**

```r
dat = dat_lmpra_zscore

vec = c("potential enhancer", "promoter", "intergenic (7 loci)")
dat = dat %>% dplyr::filter(Category %in% vec)

dat = dat %>% dplyr::select(Score, ZScore)
colnames(dat) = c("Log2FC", "ZScore")
dat = dat %>% tidyr::gather(Group, Score)

gpt = ggplot(dat, aes(x=Score, fill=Group)) + 
    geom_density(alpha=0.5) + 
    geom_vline(xintercept = 0, color = "grey30", linewidth=1) +
    labs(
        y="Density",
        title="Lenti-MPRA (Promoters, Potential Enhancers, Intergenic regions)\nBefore and after z score transformation\nusing shuffled sequences control") +
    theme_cowplot() +
    background_grid()

options(repr.plot.height=6, repr.plot.width=10)
print(gpt)
```
