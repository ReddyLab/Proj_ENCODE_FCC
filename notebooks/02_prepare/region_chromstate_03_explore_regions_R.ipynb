**Read table: ENCODE cCREs v4**

```r
### set file path
txt_fdiry = file.path(FD_DAT, "external", "encode_chromatin_states")
txt_fname = "K562.hg38.ENCSR913HQX.ENCFF286VQG.cCREs.bed.gz"
txt_fpath = file.path(txt_fdiry, txt_fname)

### read table
vec_txt_cname = c(
    "Chrom", "ChromStart", "ChromEnd", "Name", "Score", "Strand", 
    "ThickStart", "ThickEnd", "ItemRgb", 
    "Label", "Note")
dat = read_tsv(txt_fpath, col_names = vec_txt_cname, show_col_types = FALSE)

### show and assign
dat_region_ccres = dat
fun_display_table(head(dat))
```

**Read table: ChromHMM**

```r
### set file path
txt_fdiry = file.path(FD_DAT, "external", "encode_chromatin_states")
txt_fname = "K562.hg38.ENCSR365YNI.ENCFF106BGJ.ChromHMM.bed.gz"
txt_fpath = file.path(txt_fdiry, txt_fname)

### read table
vec_txt_cname = c(
    "Chrom", "ChromStart", "ChromEnd", "Name", "Score", "Strand", 
    "ThickStart", "ThickEnd", "ItemRgb")
dat = read_tsv(txt_fpath, col_names = vec_txt_cname, show_col_types = FALSE)

### show and assign
dat_region_chromhmm = dat
fun_display_table(head(dat))
```

## Prepare regions: cCREs and ChromHMM

**cCREs**

```r
dat = dat_region_ccres
dat = dat %>%
    dplyr::mutate(
        Length = ChromEnd - ChromStart,
        Group  = "cCREs"
    ) %>%
    dplyr::select(Chrom, ChromStart, ChromEnd, Length, Name, Group, Label)

dat_region_ccres_arrange = dat
fun_display_table(head(dat))
```

**ChromHMM**

```r
dat = dat_region_chromhmm
dat = dat %>%
    dplyr::mutate(
        Length = ChromEnd - ChromStart,
        Group  = "ChromHMM",
        Label  = Name
    ) %>%
    dplyr::select(Chrom, ChromStart, ChromEnd, Length, Name, Group, Label)

dat_region_chromhmm_arrange = dat
fun_display_table(head(dat))
```

## Prepare regions: cCREs silencers

```r
dat = dat_region_ccres_silencer_rest
dat = dat %>%
    dplyr::mutate(
        Chrom      = Chr, 
        ChromStart = Start, 
        ChromEnd   = End,
        Name       = `cCRE accession`
    ) %>%
    dplyr::left_join(
        dat_region_ccres, 
        by = c("Chrom", "ChromStart", "ChromEnd", "Name")
    ) %>%
    dplyr::mutate(
        Length = ChromEnd - ChromStart,
        Group  = "cCREs:REST+ silencer"
    ) %>%
    dplyr::select(Chrom, ChromStart, ChromEnd, Length, Name, Group, Label)

dat_region_ccres_silencer_rest_arrange = dat
fun_display_table(head(dat))
```

```r
dat = dat_region_ccres_silencer_starr
dat = dat %>%
    dplyr::mutate(
        Chrom      = Chr, 
        ChromStart = Start, 
        ChromEnd   = End,
        Name       = `cCRE accession`
    ) %>%
    dplyr::left_join(
        dat_region_ccres, 
        by = c("Chrom", "ChromStart", "ChromEnd", "Name")
    ) %>%
    dplyr::mutate(
        Length = ChromEnd - ChromStart,
        Group  = "cCREs:STARR silencer") %>%
    dplyr::select(Chrom, ChromStart, ChromEnd, Length, Name, Group, Label)


dat_region_ccres_silencer_starr_arrange = dat
fun_display_table(head(dat))
```

## Explore: Active vs Inactive

```r
dat = dat_region_ccres_arrange
txt1 = "Low-DNase"
txt2 = "Low-DNase"
txt3 = "Not Low-DNase"
dat1 = dat %>% dplyr::mutate(Label = ifelse(Label == txt1, txt2, txt3))

dat = dat_region_chromhmm_arrange
txt1 = "Quies"
txt2 = "Quiescent State"
txt3 = "Not Quiescent State"
dat2 = dat %>% dplyr::mutate(Label = ifelse(Label == txt1, txt2, txt3))

dat = bind_rows(dat1, dat2)
dat = dat %>% dplyr::mutate(Length = ChromEnd - ChromStart)

dat = dat %>%
    dplyr::group_by(Group, Label) %>%
    dplyr::summarise(
        Count  = n(), 
        Length = sum(Length),
        .groups="drop")
dat = dat %>%
    dplyr::group_by(Group) %>%
    dplyr::mutate(
        Total_Cnt = sum(Count),
        Total_Len = sum(Length)
    ) %>%
    dplyr::mutate(
        Freq_Cnt = Count  / Total_Cnt,
        Freq_Len = Length / Total_Len
    )

vec = c("Low-DNase", "Quiescent State", "Not Low-DNase", "Not Quiescent State")
dat = dat %>% dplyr::mutate(Label = factor(Label, levels = vec))

dat_region_freq = dat
fun_display_table(dat)
```

```r
###
vec_txt_color = c(
    "Not Quiescent State" = "#252525",
    "Quiescent State"     = "#d9d9d9",
    "Not Low-DNase"       = "#252525",
    "Low-DNase"           = "#d9d9d9")
theme_fill = scale_fill_manual(values = vec_txt_color)

###
dat = dat_region_freq
gpt = ggplot(dat, aes(x=Freq_Cnt, y=Group, fill=Label)) + 
    geom_col() + 
    labs(x = "Proportion (Counts)", y = "") +
    theme_cowplot() + 
    background_grid() +
    theme_fill

options(repr.plot.height=4, repr.plot.width=10)
print(gpt)
```

```r
###
vec_txt_color = c(
    "Not Quiescent State" = "#252525",
    "Quiescent State"     = "#d9d9d9",
    "Not Low-DNase"       = "#252525",
    "Low-DNase"           = "#d9d9d9")
theme_fill = scale_fill_manual(values = vec_txt_color)

###
dat = dat_region_freq
gpt = ggplot(dat, aes(x=Freq_Len, y=Group, fill=Label)) + 
    geom_col() + 
    labs(x = "Proportion (base pairs)", y = "") +
    theme_cowplot() + 
    background_grid() +
    theme_fill

options(repr.plot.height=4, repr.plot.width=10)
print(gpt)
```
