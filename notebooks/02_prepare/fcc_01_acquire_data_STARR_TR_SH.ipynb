## Set data directories and copy the final processed data of STARR-seq

**Create data directories**
```
PROJECT/data/processed
├── ASTARRseq_K562_hg38_KS91_210401
│   ├── fragments
│   └── peaks
│
├── ASTARRseq_K562_hg38_KS274_240311
│   └── fragments
│
└── WSTARRseq_K562_hg38_A001
    ├── fragments
    └── peaks
```

````bash
mkdir -p ${FD_DAT}/processed/WSTARRseq_K562_hg38_A001/fragments
mkdir -p ${FD_DAT}/processed/ASTARRseq_K562_hg38_KS91_210401/fragments
mkdir -p ${FD_DAT}/processed/ASTARRseq_K562_hg38_KS91_210401/peaks
mkdir -p ${FD_DAT}/processed/ASTARRseq_K562_hg38_KS274_240311/fragments
````

**Copy data files**

Copy WSTARR fragment counts

````bash
FD_OUT=${FD_DAT}/processed/WSTARRseq_K562_hg38_A001/fragments

for FPATH in ${FP_WGS_WSTARR[@]}; do
    FNAME=$(basename ${FPATH})
    cp ${FPATH} ${FD_OUT}/${FNAME}
    echo ${FPATH}
    echo ${FNAME}
    echo
done
````

Copy ASTARR fragment counts

````bash
FD_OUT=${FD_DAT}/processed/ASTARRseq_K562_hg38_KS91_210401/fragments

for FPATH in ${FP_WGS_ASTARR_KS91[@]}; do
    FNAME=$(basename ${FPATH})
    cp ${FPATH} ${FD_OUT}/${FNAME}
    echo ${FPATH}
    echo ${FNAME}
    echo
done
````

Copy ASTARR peak calls

````bash
FD_OUT=${FD_DAT}/processed/ASTARRseq_K562_hg38_KS91_210401/peaks

for FPATH in ${FP_WGS_ASTARR_KS91_INP_PEAKS[@]}; do
    FNAME=$(basename ${FPATH})
    cp ${FPATH} ${FD_OUT}/${FNAME}
    echo ${FPATH}
    echo ${FNAME}
    echo
done
````

````bash
FD_OUT=${FD_DAT}/processed/ASTARRseq_K562_hg38_KS274_240311/fragments

for FPATH in ${FP_WGS_ASTARR_KS274[@]}; do
    FNAME=$(basename ${FPATH})
    cp ${FPATH} ${FD_OUT}/${FNAME}
    echo ${FPATH}
    echo ${FNAME}
    echo
done
````
