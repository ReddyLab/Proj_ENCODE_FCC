## Execute

**Run**

```bash
### loop through each region set
while read FOLDER_REG FNAME_REG LABEL_REG; do
    
    ### Set directory
    FD_INP_A=${FD_RES}/region/${FOLDER_REG}
    FN_INP_A=${FNAME_REG}
    FP_INP_A=${FD_INP_A}/${FN_INP_A}
    
    ### loop through each assay
    for ASSAY_NAME in ${ASSAY_NAMES[@]}; do

        ### Set directory
        FD_INP_B=${FD_RES}/${ASSAY_TYPE}/${ASSAY_NAME}/coverage/${FOLDER_REG}/${LABEL_REG}/overlap_fragment
        FP_INPS=($(ls ${FD_INP_B}/*hg38*rep*bed.gz))

        ### loop through each sample
        for FP_INP_B in ${FP_INPS[@]}; do
            ### setup input file
            FD_INP_B=$(dirname  ${FP_INP_B})
            FN_INP_B=$(basename ${FP_INP_B})
            
            ### setup output file
            FD_OUT=$(dirname ${FD_INP_B})
            FD_OUT=${FD_RES}/${ASSAY_TYPE}/${ASSAY_NAME}/coverage/${FOLDER_REG}/${LABEL_REG}/overlap_fragment_one2one
            FN_OUT=${FN_INP_B}
            FP_OUT=${FD_OUT}/${FN_OUT}
            
            ### setup log file
            FN_LOG=region.coverage.overlap.one2one.${LABEL_REG}.${FN_OUT}.txt
            FP_LOG=${FD_LOG}/${FN_LOG}
            
            ### show progress
            echo ==============================
            echo "Region folder:" ${FOLDER_REG}
            echo "Region label: " ${LABEL_REG}
            echo "Assay  name:  " ${ASSAY_NAME}
            echo
            echo "Output fdiry: " ${FD_OUT}
            echo "Output fname: " ${FN_OUT}
            echo
            echo "Log file:" ${FN_LOG}
            echo
            
            ### execute
            mkdir -p ${FD_OUT}
            sbatch -p ${NODE} \
                --exclude=dl-01   \
                --cpus-per-task 4 \
                --mem 4G \
                --output ${FP_LOG} \
                --export FD_PRJ=${FD_PRJ},FP_INP=${FP_INP_B},FP_OUT=${FP_OUT} \
            <<'EOF'
#!/bin/bash
### set environment
FD_EXE=${FD_PRJ}/scripts
source ${FD_EXE}/config_func.sh
source ${FD_EXE}/config_load_module_bedtools.sh

### print start message
timer_start=`date +%s`
echo "Hostname:          " $(hostname)
echo "Slurm Array Index: " ${SLURM_ARRAY_TASK_ID}
echo "Time Stamp:        " $(date +"%m-%d-%y+%T")
echo

### show I/O file
echo "Input file: ${FP_INP}"
echo
echo "show lines of file"
fun_cat ${FP_INP} | head
echo

### init
NCOL=$(zcat ${FP_INP} | head -n 1 | awk '{print NF}')
echo ${NCOL}

### execute
###   one2one mapping/filtering
zcat ${FP_INP} \
| sort -k1,1 -k2,2n -k${NCOL},${NCOL}nr \
| bedtools groupby \
    -g 1-6 \
    -c ${NCOL} \
    -o first \
    -full \
| cut -f 1-${NCOL} \
| gzip -c \
> ${FP_OUT}

### show I/O file
echo "Output file: ${FP_OUT}"
echo
echo "show lines of file"
fun_cat ${FP_OUT} | head
echo

### print end message
timer=`date +%s`
runtime=$(echo "${timer} - ${timer_start}" | bc -l)
echo
echo 'Done!'
echo "Run Time: $(displaytime ${runtime})"
EOF
        done
    done
    
done < <(cat ${FP_REGION_LABEL} | awk 'NR >=2 {print}' | grep fcc_astarr_macs)
```
