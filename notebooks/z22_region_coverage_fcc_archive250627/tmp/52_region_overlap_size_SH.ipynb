```bash
for ASSAY_NAME in ${ASSAY_NAMES[@]}; do
    
    ### set I/O directories
    FD_INP=${FD_RES}/assay_fcc/${ASSAY_NAME}/coverage/fcc_astarr_macs/*/overlap_unique
    FD_INPS=($(ls -d ${FD_INP}))

    ### execute
    ###     loop through each file and get total fragment counts
    for FD_INP in ${FD_INPS[@]}; do
    
        ### set I/O files
        FP_INPS=($(ls ${FD_INP}/*bed.gz))
        
        ### show progress
        echo ======================================
        echo ${ASSAY_NAME}
        echo ${FD_INP}
        echo "#{Input  files} =" ${#FP_INPS[@]}
        
        for IDX in ${!FP_INPS[@]}; do
            
            ### set I/O files
            FD_OUT=${FD_INP}/summary
            FN_OUT=result.size.${IDX}.csv
            FP_OUT=${FD_OUT}/${FN_OUT}
            FP_INP=${FP_INPS[IDX]}
            
            ### set log file
            FN_LOG=coverage.size.${ASSAY_NAME}.${IDX}.txt
            FP_LOG=${FD_LOG}/${FN_LOG}
            
            ### show progress
            echo --------------------------------------
            echo "Log file: " '${FD_LOG}/'${FN_LOG}
            echo "Input:    "  ${FP_INP}
            
            ### init
            mkdir -p   ${FD_OUT}
            echo  -n > ${FP_OUT}
            
            ### execute
            sbatch -p ${NODE} \
                --exclude=dl-01 \
                --cpus-per-task 4 \
                --mem 4G \
                --output ${FP_LOG} \
                ${FD_EXE}/run_calulate_lib_size.sh ${FD_PRJ} ${FP_INP} ${FP_OUT}
            echo
        done
    done
done    
```

```bash
for ASSAY_NAME in ${ASSAY_NAMES[@]}; do
    
    ### set I/O directories
    FD_INP=${FD_RES}/assay_fcc/${ASSAY_NAME}/coverage/fcc_astarr_macs/*/overlap_unique
    FD_INPS=($(ls -d ${FD_INP}))

    ### execute
    ###     loop through each file and get total fragment counts
    for FD_INP in ${FD_INPS[@]}; do
    
        ### set I/O files
        FD_OUT=${FD_INP}/summary
        
        ### show progress
        echo ======================================
        echo ${ASSAY_NAME}
        echo ${FD_INP}
        echo
        
        ls   ${FD_OUT}/result.size.*.csv | xargs -n 1 basename
        echo
        
        ### concat files
        cat  ${FD_OUT}/result.size.*.csv > ${FD_OUT}/result.size.csv
        cat  ${FD_OUT}/result.size.csv
        echo
    done
done    
```

```bash
for ASSAY_NAME in ${ASSAY_NAMES[@]}; do
    
    ### set log file
    FN_LOG=coverage.library_size.${ASSAY_NAME}.txt
    FP_LOG=${FD_LOG}/${FN_LOG}
    
    ### show progress
    echo ======================================
    echo ${ASSAY_NAME}
    echo "Total #{files} =" ${#FP_INPS[@]}
    echo "Log file: " '${FD_LOG}/'${FN_LOG}
    
    sbatch -p ${NODE} \
        --exclude=dl-01   \
        --cpus-per-task 4 \
        --mem 4G \
        --output ${FP_LOG} \
        --export=FD_PRJ=${FD_PRJ},ASSAY_NAME=${ASSAY_NAME} \
        <<'EOF'
#!/bin/bash
### set environment
source ${FD_PRJ}/scripts/config_project.sh

### print start message
timer_start=`date +%s`
echo "Hostname:          " $(hostname)
echo "Slurm Array Index: " ${SLURM_ARRAY_TASK_ID}
echo "Time Stamp:        " $(date +"%m-%d-%y+%T")
echo

### set input and output files
FD_INP=${FD_RES}/assay_fcc/${ASSAY_NAME}/fragment_counts
FP_INPS=($(ls ${FD_INP}/*WGS*bed.gz))

FD_OUT=${FD_RES}/assay_fcc/${ASSAY_NAME}/fragment_counts/summary
FN_OUT=result.library_size.csv
FP_OUT=${FD_OUT}/${FN_OUT}

### init
mkdir -p   ${FD_OUT}
echo  -n > ${FP_OUT}

### execute
###     loop through each file and get total fragment counts
for FP_INP in ${FP_INPS[@]}; do

    ### show I/O file
    echo --------------------------------------
    echo "Input: " ${FP_INP}
    echo
    echo "show first few lines of file"
    fun_cat ${FP_INP} | head -n 3
    echo
    
    ### count the total counts and append the results to the output file
    FN_INP=$(basename ${FP_INP})
    zcat ${FP_INP} |\
        awk -v OFS=, -v FNAME=${FN_INP} '{sum += $5} END {print FNAME, sum}' \
        >> ${FP_OUT}
done

### show I/O file
echo "Output file: ${FP_OUT}"
echo
echo "show lines of file"
fun_cat ${FP_OUT}
echo

### print end message
timer=`date +%s`
runtime=$(echo "${timer} - ${timer_start}" | bc -l)
echo
echo 'Done!'
echo "Run Time: $(displaytime ${runtime})"
EOF
    echo
done
```
