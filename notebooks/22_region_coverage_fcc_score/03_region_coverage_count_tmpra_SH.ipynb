**Test loop: Region + Assay**

```bash
while read FOLDER_A FNAME_A LABEL_A FPATH_A; do
    
    ### set input file
    FD_INP_A=$(dirname ${FPATH_A})
    FN_INP_A=${FNAME_A}
    FP_INP_A=${FPATH_A}
    
    ### loop through each assay
    for ASSAY_NAME in ${ASSAY_NAMES[@]}; do

        ### Set directory
        FD_INP_B=${FD_RES}/${ASSAY_TYPE}/${ASSAY_NAME}/fragment_counts
        #FP_INPS=($(ls ${FD_INP_B}/*hg38*rep*bed.gz))
        FP_INPS=($(ls ${FD_INP_B}/*hg38*rep1*bed.gz))

        ### loop through each sample
        for FP_INP_B in ${FP_INPS[@]}; do
        
            ### set input file
            FD_INP_B=$(dirname  ${FP_INP_B})
            FN_INP_B=$(basename ${FP_INP_B})
            
            ### setup output file
            FD_OUT=${FD_RES}/region_coverage_fcc_score/${LABEL_A}/${ASSAY_NAME}
            FN_OUT=${FN_INP_B}
            FP_OUT=${FD_OUT}/${FN_OUT}
            
            ### setup log file
            FN_LOG=region.coverage.${LABEL_A}.${ASSAY_NAME}.${FN_OUT}.txt
            FP_LOG=${FD_LOG}/${FN_LOG}
            
            ### show progress
            echo ==============================
            echo "Region folder:" ${FOLDER_A}
            echo "Region label: " ${LABEL_A}
            echo "Assay  name:  " ${ASSAY_NAME}
            echo
            echo "Input fdiry:  " ${FD_INP_A/${FD_RES}/'${FD_RES}'}
            echo "Input fname:  " ${FN_INP_A}
            echo "Input fdiry:  " ${FD_INP_B/${FD_RES}/'${FD_RES}'}
            echo "Input fname:  " ${FN_INP_B}
            echo
            echo "Output fdiry: " ${FD_OUT/${FD_RES}/'${FD_RES}'}
            echo "Output fname: " ${FN_OUT}
            echo
            echo "Log file:" '${FD_LOG}/'${FN_LOG}
            echo
        done
    done
done < <(cat ${FP_REGION_LABEL} | awk 'NR >=2 {print}' | head -n 1)    
#done < <(cat ${FP_REGION_LABEL} | awk 'NR >=2 {print}')
```

## Execute

**Check**

```bash
echo ${FP_CNF}
```

**RUN**

```bash
### init: set executable
FN_EXE="run_bedtools_coverage.sh"
FP_EXE=${FD_EXE}/${FN_EXE}

### Execute
while read FOLDER_A FNAME_A LABEL_A FPATH_A; do
    
    ### set input file
    FD_INP_A=$(dirname ${FPATH_A})
    FN_INP_A=${FNAME_A}
    FP_INP_A=${FPATH_A}
    
    ### loop through each assay
    for ASSAY_NAME in ${ASSAY_NAMES[@]}; do

        ### Set directory
        FD_INP_B=${FD_RES}/${ASSAY_TYPE}/${ASSAY_NAME}/fragment_counts
        FP_INPS=($(ls ${FD_INP_B}/*hg38*rep*bed.gz))
        #FP_INPS=($(ls ${FD_INP_B}/*hg38*rep1*bed.gz))

        ### loop through each sample
        for FP_INP_B in ${FP_INPS[@]}; do
        
            ### set input file
            FD_INP_B=$(dirname  ${FP_INP_B})
            FN_INP_B=$(basename ${FP_INP_B})
            
            ### setup output file
            FD_OUT=${FD_RES}/region_coverage_fcc_score/${LABEL_A}/${ASSAY_NAME}
            FN_OUT=${FN_INP_B}
            FP_OUT=${FD_OUT}/${FN_OUT}
            
            ### setup log file
            FN_LOG=region.coverage.${LABEL_A}.${ASSAY_NAME}.${FN_OUT}.txt
            FP_LOG=${FD_LOG}/${FN_LOG}
            
            ### show progress
            echo ==============================
            echo "Region folder:" ${FOLDER_A}
            echo "Region label: " ${LABEL_A}
            echo "Assay  name:  " ${ASSAY_NAME}
            echo
            echo "Input fdiry:  " ${FD_INP_A/${FD_RES}/'${FD_RES}'}
            echo "Input fname:  " ${FN_INP_A}
            echo "Input fdiry:  " ${FD_INP_B/${FD_RES}/'${FD_RES}'}
            echo "Input fname:  " ${FN_INP_B}
            echo
            echo "Output fdiry: " ${FD_OUT/${FD_RES}/'${FD_RES}'}
            echo "Output fname: " ${FN_OUT}
            echo
            echo "Log file:" '${FD_LOG}/'${FN_LOG}
            echo

            ### execute
            mkdir -p ${FD_OUT}
            sbatch \
                --cpus-per-task 4 \
                --mem 10G \
                --output ${FP_LOG} \
                ${FP_EXE} ${FP_CNF} ${FP_INP_A} ${FP_INP_B} ${FP_OUT}
            echo
        done
    done
#done < <(cat ${FP_REGION_LABEL} | awk 'NR >=2 {print}' | head -n 1)    
done < <(cat ${FP_REGION_LABEL} | awk 'NR >=2 {print}')
```
