## Import data

```r
txt_fdiry = file.path(
    FD_RES, 
    "region_annotation", 
    "fcc_astarr_macs_input_overlap",
    "summary"
)
txt_fname = "matrix.annotation.prepare.chipseq_full.fcc.tss.tsv"
txt_fpath = file.path(txt_fdiry, txt_fname)

dat = read_tsv(txt_fpath, show_col_types = FALSE)
dat_region_annot_v1 = dat


txt_fname = "matrix.annotation.prepare.chipseq_full.only.tsv"
txt_fpath = file.path(txt_fdiry, txt_fname)

dat = read_tsv(txt_fpath, show_col_types = FALSE)
dat_region_annot_v2 = dat
```

```r
txt_fdiry = file.path(
    FD_RES, 
    "region_annotation", 
    "fcc_astarr_macs_input_overlap",
    "summary"
)
txt_fname = "umap.n_neighbors_70.annotation.prepare.chipseq_full.fcc.tss.tsv"
txt_fpath = file.path(txt_fdiry, txt_fname)

dat = read_tsv(txt_fpath, show_col_types = FALSE)
dat_region_umap_v1 = dat

txt_fname = "umap.n_neighbors_70.annotation.prepare.chipseq_full.only.tsv"
txt_fpath = file.path(txt_fdiry, txt_fname)

dat = read_tsv(txt_fpath, show_col_types = FALSE)
dat_region_umap_v2 = dat
```

## Arrange columns

```r
dat = bind_cols(
    dat_region_umap_v1,
    dat_region_annot_v1
)

dat_region_merge_v1 = dat
fun_display_table(head(dat, 3)) 
```

```r
dat = bind_cols(
    dat_region_umap_v2,
    dat_region_annot_v2
)

dat_region_merge_v2 = dat
fun_display_table(head(dat, 3)) 
```

```r
dat = dat_region_merge_v1
dat = dat %>% dplyr::select(Region, UMAP1, UMAP2)

dat_region_export_v1 = dat
fun_display_table(head(dat, 3)) 
```

```r
dat = dat_region_merge_v2
dat = dat %>% dplyr::select(Region, UMAP1, UMAP2)

dat_region_export_v2 = dat
fun_display_table(head(dat, 3)) 
```

## Export results

```r
txt_fdiry = file.path(
    FD_RES, 
    "region_annotation", 
    "fcc_astarr_macs_input_overlap",
    "summary"
)
txt_fname = "result.umap.n_neighbors_70.annotation.prepare.chipseq_full.fcc.tss.tsv"
txt_fpath = file.path(txt_fdiry, txt_fname)

dat = dat_region_export_v1
write_tsv(dat, txt_fpath)

txt_fname = "result.umap.n_neighbors_70.annotation.prepare.chipseq_full.only.tsv"
txt_fpath = file.path(txt_fdiry, txt_fname)

dat = dat_region_export_v2
write_tsv(dat, txt_fpath)
```
