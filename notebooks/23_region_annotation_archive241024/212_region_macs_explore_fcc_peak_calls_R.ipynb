## Arrange table

```r
vec = c("ASTARR", "WSTARR", "LMPRA")
vec = c(
    paste(vec, "A", sep="_"),
    paste(vec, "R", sep="_")
)

vec_txt_label_fcc_starrmpra = vec
```

```r
vec = c("Signif")

vec_txt_label_fcc_cgrowth = vec
```

```r
vec_txt_label_fcc = c(
    vec_txt_label_fcc_starrmpra,
    vec_txt_label_fcc_cgrowth
)
print(vec_txt_label_fcc)
```

```r
###
lst = list(
    dat_region_annot_fcc_starrmpra,
    dat_region_annot_fcc_cgrowth
)

###
lst = lapply(lst, function(dat){
    ###
    dat = dat %>%
        dplyr::mutate(
            Chrom      = Chrom_A,
            ChromStart = ChromStart_A, 
            ChromEnd   = ChromEnd_A,
            Region     = Region_A,
            Group      = Group_B,
            Label      = Label_B
        ) %>%
        dplyr::select(Chrom, ChromStart, ChromEnd, Region, Group, Label)
    return(dat)
})

###
dat = bind_rows(lst)
dat = dat %>% 
    dplyr::filter(Label %in% vec_txt_label_fcc) %>% 
    dplyr::distinct()
dat1 = dat

###
dat = dat %>%
    dplyr::group_by(Chrom, ChromStart, ChromEnd, Region, Group) %>%
    dplyr::summarize(
        Label = paste(sort(unique(Label)), collapse = ","),
        .groups = "drop"
    )
dat2 = dat

### assign and show
dat_region_annot_full    = dat1
dat_region_annot_summary = dat2
```

```r
dat = dat_region_annot_full
print(dim(dat))
print(table(dat$Label))
fun_display_table(head(dat, 3))
```

```r
dat = dat_region_annot_summary
print(dim(dat))
print(table(dat$Label))
fun_display_table(head(dat, 3))
```

## Plot

```r
vec  = c("ASTARR", "WSTARR", "LMPRA")

vec1 = paste(vec, "A", sep="_")
vec2 = paste(vec, "R", sep="_")
vec3 = paste(vec1, vec2, sep = ",")

vec4 = c(vec1, vec2, vec3, "Signif")
vec5 = c(
    rep("Active (A)",  length(vec1)),
    rep("Repress (R)",  length(vec2)),
    rep("Contain Both (AR)", length(vec3)),
    "Significant"
)

fun = function(txt){
    res = fun_str_map_match(txt, vec4, vec5)
    return(res)
}
```

```r
dat = dat_region_annot_summary
dat = dat %>% dplyr::mutate(Peak_Call = fun(Label))

dat_region_annot_final = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

```r
dat = dat_region_annot_final
dat = dat %>% dplyr::select(Chrom, Region) %>% dplyr::distinct()
table(dat$Chrom)
```

```r
dat = dat_region_annot_final
table(dat$Label, dat$Peak_Call)
```

```r
vec_txt_assay = c("ASTARR", "WSTARR", "LMPRA", "CRISPRi-Growth")
vec_txt_color = c(
    "Active (A)"        = "#fb8072", 
    "Repress (R)"       = "#80b1d3", 
    "Contain Both (AR)" = "#984ea3",
    "Significant"       = "black"
)
theme_fill = scale_fill_manual(values = vec_txt_color)
```

```r
dat = dat_region_annot_final
dat = dat %>% 
    dplyr::filter(Chrom == "chrX") %>%
    dplyr::mutate(Group = factor(Group, levels = vec_txt_assay))

gpt = ggplot(dat, aes(x=Group, y=Region, fill=Peak_Call)) +
    geom_tile() + 
    labs(x="Assay", y="ATAC") +
    theme_fill +
    theme_cowplot() +
    theme(
        axis.text.y = element_blank(),
        axis.ticks.y = element_blank()
    )
options(repr.plot.height = 8, repr.plot.width=8)
print(gpt)
```

## Clustering

```r
dat = dat_region_annot_final
dat = dat %>% dplyr::filter(Chrom == "chrX")

dat = dat %>% 
    dplyr::mutate(Value  = 1) %>%    
    dplyr::select(Region, Label, Value) %>%
    tidyr::spread(Label, Value) %>%
    replace(is.na(.), 0) %>%
    column_to_rownames("Region")

mat_region_annot_full = dat
print(dim(dat))
head(dat)
```

```r
mat = mat_region_annot_full
dd  = dist(mat,  method = "manhattan")
hc  = hclust(dd, method = "ward.D2")
plot(hc, hang = -1, cex = 0.6)
```

```r
vec = rownames(mat)
idx = hc$order
vec = vec[idx]

vec_txt_region = vec
head(vec)
```

```r
dat = dat_region_annot_final
dat = dat %>% 
    dplyr::filter(Chrom == "chrX") %>%
    dplyr::mutate(
        Group  = factor(Group,  levels = vec_txt_assay),
        Region = factor(Region, levels = vec_txt_region)
    )

gpt = ggplot(dat, aes(x=Group, y=Region, fill=Peak_Call)) +
    geom_tile() + 
    labs(x="Assay", y="ATAC", title = "FCC peak calls across ATAC (ChrX)") +
    theme_fill +
    theme_cowplot() +
    theme(
        axis.text.y = element_blank(),
        axis.ticks.y = element_blank()
    )
options(repr.plot.height = 10, repr.plot.width=8)
print(gpt)
```
