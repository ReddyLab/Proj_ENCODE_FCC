## Import data

```r
txt_fdiry = file.path(
    FD_RES, 
    "region_annotation", 
    "fcc_astarr_macs_input_overlap",
    "summary"
)
txt_fname = "matrix.annotation.test_chrX.chipseq_full.only.tsv"
txt_fpath = file.path(txt_fdiry, txt_fname)

mat = read_tsv(txt_fpath, show_col_types = FALSE)
mat = mat %>% tibble::column_to_rownames("Region")

mat_region_annot = mat
fun_display_table(head(mat, 3))
```

```r
txt_fname_inp = txt_fname
txt_fname_out = txt_fname_inp %>%
    stringr::str_replace("matrix", "hclust.row") %>%
    stringr::str_replace("tsv",    "rds")

cat("Input  File:", txt_fname_inp, "\n")
cat("Output File:", txt_fname_out, "\n")
```

## Execute: Perform HClust

```r
mat = mat_region_annot
mat = dist(mat,  method = "manhattan")
res = fastcluster::hclust(mat, method = "ward.D2")

res_hcl_row = res
```

```r
mat = mat_region_annot
mat = t(mat)
mat = dist(mat,  method = "manhattan")
res = fastcluster::hclust(mat, method = "ward.D2")

res_hcl_col = res
```
