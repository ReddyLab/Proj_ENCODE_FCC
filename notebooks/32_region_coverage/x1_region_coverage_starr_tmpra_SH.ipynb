````bash
###
while read FOLDER_REG FNAME_REG LABEL_REG; do
    
    ###
    FD_INP_A=${FD_RES}/region/${FOLDER_REG}
    FN_INP_A=${FNAME_REG}
    FP_INP_A=${FD_INP_A}/${FN_INP_A}
    
    ###
    for FP_INP_B in ${FP_INPS_ASTARR[@]}; do
        ### setup input file
        FD_INP_B=$(dirname  ${FP_INP_B})
        FN_INP_B=$(basename ${FP_INP_B})
        
        ### setup output file
        FD_TMP=$(dirname ${FD_INP_B})
        FD_OUT=${FD_TMP}/coverage/${FOLDER_REG}/${LABEL_REG}
        FN_OUT=${FN_INP_B}
        FP_OUT=${FD_OUT}/${FN_OUT}
        
        ### setup log file
        ASSAY_NAME=$(basename ${FD_TMP})
        FN_LOG=region.coverage.${LABEL_REG}.${FN_OUT}.txt
        FP_LOG=${FD_LOG}/${FN_LOG}
        
        ### show progress
        echo ==============================
        echo "Region folder:" ${FOLDER_REG}
        echo "Region label: " ${LABEL_REG}
        echo "Assay  name:  " ${ASSAY_NAME}
        echo
        echo "Output fdiry: " ${FD_OUT}
        echo "Output fname: " ${FN_OUT}
        echo
        echo "Log file:" ${FN_LOG}
        echo
        
        ### execute
        mkdir -p ${FD_OUT}
        sbatch -p ${NODE} \
            --exclude=dl-01 \
            --cpus-per-task 4 \
            --mem 4G \
            --output ${FP_LOG} \
            ${FD_EXE}/run_bedtools_coverage.sh ${FD_PRJ} ${FP_INP_A} ${FP_INP_B} ${FP_OUT}
        echo
    done
done < <(cat ${FP_REGION_LABEL} | awk 'NR >=2 {print}')
````
