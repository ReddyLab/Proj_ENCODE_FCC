```r
###
fun = function(dat1, dat2){
    dat = dplyr::full_join(dat1, dat2, by = "Region")
    return(dat)
}

###
#lst_region = lst_metadata_region["fcc_astarr_macs"]
lst_region = lst_metadata_region

for (txt_folder in names(lst_region)){
    
    ###
    vec_txt_label = lst_region[[txt_folder]]

    ###
    txt_fdiry  = file.path(FD_RES, "region", txt_folder)
    txt_fname  = "description.tsv"
    txt_fpath  = file.path(txt_fdiry, txt_fname)
    
    dat = read_tsv(txt_fpath, show_col_types = FALSE)
    vec = dat$Name
    vec_txt_cname = c(vec, "Count")

    ###
    for (txt_label in vec_txt_label){
        
        ###
        cat("====================", "\n")
        cat("Region Folder:", txt_folder, "\n")
        cat("Region Label: ", txt_label,  "\n")
        flush.console()
        
        ###
        txt_fdiry = file.path(FD_RES, ASSAY_TYPE, ASSAY_NAME, "coverage", txt_folder, txt_label)
        txt_fname = "*WGS*bed.gz"
        txt_fglob = file.path(txt_fdiry, txt_fname)
        
        vec = Sys.glob(txt_fglob)
        vec = basename(vec)
        vec_txt_fname = vec

        ###
        txt_fdiry = file.path(FD_RES, ASSAY_TYPE, ASSAY_NAME, "coverage", txt_folder, txt_label)
        lst_dat   = lapply(vec_txt_fname, function(txt_fname){

            ###
            txt_fpath = file.path(txt_fdiry, txt_fname)

            ###
            dat = read_tsv(txt_fpath, col_names = vec_txt_cname, show_col_types = FALSE)
            dat = dat %>% 
                dplyr::select(Chrom, ChromStart, ChromEnd, Count) %>%
                dplyr::mutate(Region = paste0(Chrom, ":", ChromStart, "-", ChromEnd)) %>%
                dplyr::mutate(Count  = replace(Count, Count == ".", "0")) %>%
                dplyr::mutate(Count  = as.integer(Count)) %>%
                dplyr::select(Region, Count) %>%
                dplyr::distinct()

            ###
            vec = c("Region", txt_fname)
            colnames(dat) = vec
            return(dat)
        })
        
        ###
        dat = Reduce(fun, lst_dat)
        dat = dplyr::rename(dat, any_of(vec_txt_sample))
        
        ###
        fun_display_table(head(dat))
        fun_display_table(tail(dat))
        flush.console()

        ### 
        txt_fdiry = file.path(FD_RES, ASSAY_TYPE, ASSAY_NAME, "coverage", txt_folder, txt_label, "summary")
        txt_fname = "matrix.raw.count.WGS.tsv"
        txt_fpath = file.path(txt_fdiry, txt_fname)
        
        txt_cmd = paste("mkdir -p", txt_fdiry)
        system(txt_cmd)
        write_tsv(dat, txt_fpath)
    }
}
```
