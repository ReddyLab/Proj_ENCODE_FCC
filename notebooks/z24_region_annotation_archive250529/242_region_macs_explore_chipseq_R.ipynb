## Summary

```r
dat = dat_region_score_merge
dat = dat %>% 
    dplyr::filter(Assay_Type == "ATAC") %>%
    dplyr::mutate(ATAC = Score) %>%
    dplyr::select(Chrom, ChromStart, ChromEnd, Region, ATAC)

dat_region_score_atac = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

```r
### count
dat = dat_region_annot_chipseq_full
dat = dat %>%
    dplyr::group_by(Region) %>%
    dplyr::summarise(Count = n()) %>%
    dplyr::arrange(Count)

### assign and show
dat_region_summary_count = dat
print(dim(dat))
print(summary(dat$Count))
fun_display_table(head(dat))
```

```r
### count
dat = dat_region_annot_chipseq_full
dat = dat %>%
    dplyr::group_by(Label) %>%
    dplyr::summarise(Count = n()) %>%
    dplyr::arrange(-Count)

### assign and show
dat_feature_summary_count = dat
print(dim(dat))
print(summary(dat$Count))
fun_display_table(head(dat))
```

```r
dat = dplyr::inner_join(
    dat_region_score_atac,
    dat_region_summary_count,
    by = "Region"
)

dat_region_summary_count_signal = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```

## Visualize

```r
dat = dat_region_summary_count

gpt1 = ggplot(dat, aes(x=Count)) +
    geom_histogram(bins = 50) +
    theme_cowplot() +
    background_grid()

gpt2 = ggplot(dat, aes(x=Count)) +
    geom_histogram(bins = 50) +
    scale_y_continuous(trans='log10') +
    theme_cowplot() +
    background_grid()

lst = list(gpt1, gpt2)
lst = lapply(lst, function(gpt){
    gpt = gpt + labs(x = "#TF", y = "Count")
    return(gpt)
})
options(repr.plot.height=5, repr.plot.width=10)
plot_grid(plotlist = lst, nrow = 1)
```

```r
dat = dat_region_annot_chipseq_full
dat = dat %>%
    dplyr::select(Region, Label) %>%
    dplyr::mutate(Value = "1") %>%
    dplyr::mutate(
        Region = factor(Region, levels=dat_region_summary_count$Region),
        Label  = factor(Label,  levels=dat_feature_summary_count$Label)
    )

gpt = ggplot(dat, aes(x = Label, y = Region, fill = Value)) +
    geom_tile() +
    scale_fill_manual(values=c("1"="black")) +
    labs(x = "TF", y = "Region", title = "TF binding sites at accessbile regions") +
    theme_cowplot() +
    theme(
        legend.position = "none",
        axis.ticks = element_blank(),
        axis.text = element_blank()
    )

options(repr.plot.height=7, repr.plot.width=7)
print(gpt)
```

```r
dat = dat_region_summary_count_signal
gpt = ggplot(dat, aes(x=ATAC, y=Count)) + 
    geom_point() + 
    xlim(0, 150) +
    ylim(0, 450) +
    labs(x = "ATAC (TPM)", y = "Count (#TF)") +
    theme_cowplot() +
    background_grid()

gpt1 = gpt

options(repr.plot.height=7, repr.plot.width=7)
print(gpt)
```

```r
dat = dat_region_summary_count_signal
gpt = ggplot(dat, aes(x=ATAC, y=Count)) + 
    geom_hex(bins = 50) + 
    xlim(0, 150) +
    ylim(0, 450) +
    scale_fill_viridis_c(trans = "log10") +
    labs(x = "ATAC (TPM)", y = "Count (#TF)") +
    theme_cowplot() +
    background_grid() +
    guides(fill = guide_colourbar(barwidth = 1, barheight = 15))

gpt2 = gpt

options(repr.plot.height=7, repr.plot.width=7)
print(gpt)
```

```r
lst = list(gpt1, gpt2)
options(repr.plot.height=5, repr.plot.width=10)
plot_grid(plotlist = lst, nrow = 1, rel_widths = c(1, 1.2))
```
