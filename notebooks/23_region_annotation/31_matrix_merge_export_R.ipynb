## Merge

```r
lst = list(
    dat_region_annot_fcc,
    dat_region_annot_tss,
    dat_region_annot_chipseq
)

fun = function(dat1, dat2){
    vec = c("Chrom", "ChromStart", "ChromEnd", "Region")
    dat = dplyr::full_join(dat1, dat2, by = vec)
    return(dat)
}

### merge
dat = Reduce(fun, lst)
dat = dat %>% 
    dplyr::distinct() %>%
    dplyr::arrange(Chrom, ChromStart, ChromEnd)

### assign and show
dat_region_annot_merge = dat
print(dim(dat))
fun_display_table(head(dat, 3))
```
