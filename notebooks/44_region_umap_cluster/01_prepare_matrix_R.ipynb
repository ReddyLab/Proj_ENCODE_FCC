## Import data

**Import region**

```r
lst = lapply(vec_txt_folder, function(txt_folder){
    ### set file directory
    txt_fdiry = file.path(
        FD_RES, 
        "region_annotation", 
        txt_folder,
        "summary"
    )
    txt_fname = "region.annotation.fcc_starrmpracrispr.assayvote.tsv"
    txt_fpath = file.path(txt_fdiry, txt_fname)

    ### read table
    dat = read_tsv(txt_fpath, show_col_types = FALSE)
    return(dat)
})
names(lst) = vec_txt_folder

### assign and show
lst_dat_region_annot_fcc_assayvote = lst
res = lapply(lst, dim)
print(res)
dat = lst[[1]]
fun_display_table(head(dat, 3))
```

## Prepare matrix

```r
lst = lapply(vec_txt_folder, function(txt_folder){
    ### get table
    dat_region_annot = lst_dat_region_annot_fcc_assayvote[[txt_folder]]
    dat_region_score = lst_dat_region_score_chipseq_mean0[[txt_folder]]

    ### filter
    dat = dat_region_annot
    dat = dat %>%
        dplyr::filter(Num_Assay > 1) %>%
        dplyr::distinct()
    dat_region_annot_subset = dat
    
    ### merge
    vec = c("Chrom", "ChromStart", "ChromEnd", "Region")
    dat = dplyr::left_join(
        dat_region_annot_subset,
        dat_region_score,
        by = vec)
    return(dat)
})

### assign and show
lst_dat_region_fcc_vote2_chipseq_mean0 = lst

res = lapply(lst, dim)
print(res)

dat = lst[[1]]
fun_display_table(head(dat, 3))
```

```r
lst = lapply(vec_txt_folder, function(txt_folder){
    ### get table
    dat_region_annot = lst_dat_region_annot_fcc_assayvote[[txt_folder]]
    dat_region_score = lst_dat_region_score_chipseq_log2p[[txt_folder]]

    ### filter
    dat = dat_region_annot
    dat = dat %>%
        dplyr::filter(Num_Assay > 1) %>%
        dplyr::select(Chrom, ChromStart, ChromEnd, Region) %>%
        dplyr::distinct()
    dat_region_annot_subset = dat
    
    ### merge
    vec = c("Chrom", "ChromStart", "ChromEnd", "Region")
    dat = dplyr::left_join(
        dat_region_annot_subset,
        dat_region_score,
        by = vec)
    return(dat)
})

### assign and show
lst_dat_region_fcc_vote2_chipseq_log2p = lst

res = lapply(lst, dim)
print(res)

dat = lst[[1]]
fun_display_table(head(dat, 3))
```

```r
for (txt_folder in vec_txt_folder){
    ### set file directory
    txt_fdiry = file.path(
        FD_RES, 
        "analysis_umap_cluster_by_coverage",
        txt_folder, 
        "fcc_starrmpracrispri_vote2",
        "prepare_matrix"
    )
    txt_fname = "matrix.coverage.encode_chipseq_full_log2p.tsv"
    txt_fpath = file.path(txt_fdiry, txt_fname)

    ### create folder if not exist
    txt_cmd = paste("mkdir -p", txt_fdiry)
    system(txt_cmd)
    
    ### write data
    lst = lst_dat_region_fcc_vote2_chipseq_log2p
    dat = lst[[txt_folder]]
    dat = dat %>% 
        dplyr::arrange(Chrom,  ChromStart,  ChromEnd) %>%
        dplyr::select(-Chrom, -ChromStart, -ChromEnd)
    
    write_tsv(dat, txt_fpath)

    ### show progress
    cat("Save file", txt_fpath, "\n")
}
```
