```r
lst = list(
    dat_region_annot_tss %>% dplyr::mutate(Label = "TSS"),
    dat_region_annot_ccres,
    dat_region_annot_chromhmm
)
dat = bind_rows(lst)
lst = split(dat$Region, dat$Label)

lst_vec_txt_region = lst
names(lst)
```

```r
lst = lst_vec_txt_region
txt = "TSS"
vec = lst[[txt]]

dat = dat_region_umap
tmp = dat %>% dplyr::filter(Region %in% vec)

gpt = ggplot() + 
    geom_point(data = dat, aes(x=UMAP1, y=UMAP2), color="grey30", size = 0.1, alpha = 0.1) + 
    geom_point(data = tmp, aes(x=UMAP1, y=UMAP2), color="#e41a1c",size = 0.1, alpha = 0.7) +
    theme_cowplot() +
    labs(title = txt) 
options(repr.plot.height = 5, repr.plot.width = 5)
print(gpt)
```
