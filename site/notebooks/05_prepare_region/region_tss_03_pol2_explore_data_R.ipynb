## Visualize: Distribution of POL2 Scores

```r
dat = bind_rows(dat_region_tss_pol2, dat_region_tss_pol2_rnaseq)
gpt = ggplot(dat, aes(x = Score, color = Group)) +
    geom_density() +
    theme_cowplot() +
    background_grid()

options(repr.plot.height=4, repr.plot.width=10)
print(gpt)
```

```r
dat = dat_region_tss_pol2
dat = dat %>% dplyr::filter(Score > 0, Score <= 100)
gpt = ggplot(dat, aes(x = Score)) +
    geom_density() +
    theme_cowplot() +
    background_grid()

options(repr.plot.height=4, repr.plot.width=7)
print(gpt)
```

```r
dat = dat_region_tss_pol2_rnaseq
dat = dat %>% dplyr::filter(Score > 0, Score <= 100)
gpt = ggplot(dat, aes(x = Score)) +
    geom_density() +
    theme_cowplot() +
    background_grid()

options(repr.plot.height=4, repr.plot.width=7)
print(gpt)
```

## Save figures

```r
### set text size
theme_text = theme(
    title      = element_text(size = 14),
    axis.title = element_text(size = 14),
    axis.text  = element_text(size = 14)
)
```

```r
dat = dat_region_tss_pol2
dat = dat %>% dplyr::filter(Score > 0, Score <= 100)
gpt = ggplot(dat, aes(x = Score)) +
    geom_density(fill = "#80b1d3") +
    theme_cowplot() +
    background_grid() +
    labs(x = "POLR2A Score", y = "Density", title = "TSS (Filtered by K562 Pol2)")

gpt_score_tss_pol2 = gpt

options(repr.plot.height=4, repr.plot.width=6)
print(gpt)
```

```r
dat = dat_region_tss_pol2_rnaseq
dat = dat %>% dplyr::filter(Score > 0, Score <= 100)
gpt = ggplot(dat, aes(x = Score)) +
    geom_density(fill = "#fb8072") +
    theme_cowplot() +
    background_grid() +
    labs(x = "POLR2A Score", y = "Density", title = "TSS (Filtered by K562 Pol2 and RNA-seq)") +
    scale_y_continuous(breaks=c(0, 0.01, 0.02))

gpt_score_tss_pol2_rnaseq = gpt

options(repr.plot.height=4, repr.plot.width=6)
print(gpt)
```

```r
lst = list(gpt_score_tss_pol2, gpt_score_tss_pol2_rnaseq)
lst = lapply(lst, function(gpt){return(gpt + theme_text)})

plt = plot_grid(plotlist = lst, ncol = 1, align = "v")

plt_export = plt

options(repr.plot.height=6, repr.plot.width=6)
print(plt)
```

```r
txt_fdiry = "./"
txt_fname = "fig.region.tss.distribution.pol2.below1h.png"
txt_fpath = file.path(txt_fdiry, txt_fname)
ggsave(txt_fpath, plt_export, height = 6, width = 6, units = "in")

txt_fdiry = "./"
txt_fname = "fig.region.tss.distribution.pol2.below1h.svg"
txt_fpath = file.path(txt_fdiry, txt_fname)
ggsave(txt_fpath, plt_export, height = 6, width = 6, units = "in")
```
