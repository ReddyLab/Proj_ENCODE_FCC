## Get MD5 checksums for each file

**Helper function**

````r
fun_get_md5sum = function(txt_accession) {
    ### fetch webpage info from ENCODE
    txt_url = paste0("https://www.encodeproject.org/files/", txt_accession, "/?format=json")
    res = GET(txt_url, user_agent("encode-client/1.0"), accept("application/json"))
    stop_for_status(res)

    ### parse the page to json format
    obj = fromJSON(content(res, "text", encoding="UTF-8"))
    
    ### get the md5sum field
    txt_md5sum = obj$md5sum

    return(txt_md5sum)
}
````

**Add md5 checksum as a column**

````r
### fetch MD5 checksum for each file
dat = dat_metadata_import
dat = dat %>%
    dplyr::rowwise() %>%
    dplyr::mutate(md5sum = fun_get_md5sum(Index_File))

### assign and show
dat_metadata_add_md5sum = dat
print(dim(dat))
fun_display_table(head(dat))
````

## Generate download commands
```
wget -O FILE URL
```

**Helper function**

````r
### define helper function: inner function
fun_str_map_inner = function(txt){
    vec1 = c("peak", "signal_R1", "signal_R2")
    vec2 = c("CASA", "signal_R1", "signal_R2")
    
    idx = str_detect(string  = txt, pattern = vec1)
    vec = vec2[idx]
    return(vec)
}

### define helper function: wrapper function
fun_str_map_label = function(vec_txt_input){
    vec_txt_output = sapply(vec_txt_input, fun_str_map_inner)
    return(vec_txt_output)
}

### test function
vec = c("CAPRIN1_peakwise_file", "CAPRIN1_perturb_signal_R1", "CAPRIN1_perturb_signal_R2")
res = fun_str_map_label(vec)
print(res)
````

**Arrange table**

````r
### init
dat = dat_metadata_add_md5sum

### setup download file name and wget command
dat = dat %>%
    dplyr::mutate(
        File_Ext1 = ifelse(str_detect(File_Type, "bed"), "bed.gz", "bw"),
        File_Ext2 = ifelse(str_detect(File_Type, "bed"), "bed.gz", "bigWig")
    )  %>% 
    dplyr::mutate(File_Label = fun_str_map_label(File_Summary)) %>%
    dplyr::mutate(
        File_Name = paste(
            Biosample, 
            Genome, 
            Index_Experiment, 
            Index_File,
            "CRISPRi_HCRFF",
            File_Label,
            Target,
            File_Ext1, 
            sep=".")
    ) %>%
    dplyr::mutate(
        File_URL = file.path(
            "https://www.encodeproject.org/files",
            Index_File,
            "@@download",
            paste(Index_File, File_Ext2, sep = ".")
        )
    ) %>%
    dplyr::mutate(
        CMD = paste("wget", "--append-output=run_download.log.txt", "-O", File_Name, File_URL)
    )

### assign and show
dat_metadata_arrange = dat
print(dim(dat))
fun_display_table(head(dat))
````

**Prepare download script**

````r
### init
dat = dat_metadata_arrange

### add Shebang and initial commands
#dat = dat %>% dplyr::select(Assay, Biosample, Index_Experiment, Index_File, File_Name, CMD)
dat = dat %>% dplyr::select(CMD)
dat = rbind('echo -n "" > run_download.log.txt', dat)
colnames(dat) = "#!/bin/bash"

### assign and show
dat_download_script = dat
fun_display_table(dat)
````

## Prepare `Checksum` table

````r
### init
dat = dat_metadata_arrange

### md5sum and file name
dat = dat_metadata_arrange
dat = dat %>% dplyr::select(md5sum, File_Name)

### assign and show
dat_download_checksum = dat
print(dim(dat))
fun_display_table(head(dat))
````

## Save to script
Save the command lines for each row into a bash script

````r
### set output path
txt_folder = "encode_crispri_hcrff"
txt_fdiry  = file.path(FD_DAT, "external", txt_folder)

### create directory if not exist
dir.create(txt_fdiry, showWarnings = FALSE)

### write checksum file
txt_fname  = "run_download_files.sh"
txt_fpath  = file.path(txt_fdiry, txt_fname)

dat = dat_download_script
write_tsv(dat, txt_fpath)

### save table
txt_fname  = "checksum_md5sum.txt"
txt_fpath  = file.path(txt_fdiry, txt_fname)

dat = dat_download_checksum
write_tsv(dat, txt_fpath, col_names = FALSE)
````
