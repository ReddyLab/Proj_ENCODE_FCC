## Import data

**Import region pairs**

```r
### loop to import data
lst = lapply(VEC_TXT_FOLDER, function(txt_folder){
    ### set file directory
    txt_fdiry = file.path(FD_RES, "region_annotation", txt_folder, "summary")
    txt_fname = TXT_FNAME_ANNOT
    txt_fpath = file.path(txt_fdiry, txt_fname)

    ### read table
    dat = read_tsv(txt_fpath, show_col_types = FALSE)
    return(dat)
})
names(lst) = VEC_TXT_FOLDER

### assign and show
lst_dat_region_annot_import = lst

res = lapply(lst, dim)
print(res)

dat = lst[[1]]
fun_display_table(head(dat, 3))
```

**Explore: Group**

```r
lst = lst_dat_region_annot_import
lst = lapply(lst, function(dat){
    res = table(dat$Group, dnn = "Group")
    dat = as.data.frame(res)
    return(res)
})

dat = bind_rows(lst, .id = "Region")
fun_display_table(dat)
```

**Explore: Group x Label**

```r
lst = lst_dat_region_annot_import
lst = lapply(lst, function(dat){
    res = table(dat$Group, dat$Label, dnn = c("Group", "Label"))
    dat = as.data.frame(res)
    return(dat)
})

dat = bind_rows(lst, .id = "Region")
dat = dat %>% tidyr::spread(Label, Freq)
fun_display_table(dat)
```

**Explore: Group x Region Count**

```r
lst = lst_dat_region_annot_import
lst = lapply(lst, function(dat){
    res = table(dat$Group, dat$Region_Count, dnn = c("Group", "Region_Count"))
    dat = as.data.frame(res)
    return(dat)
})

dat = bind_rows(lst, .id = "Region")
dat = dat %>% tidyr::spread(Region_Count, Freq)
fun_display_table(dat)
```

## Arrange table

```r
### split chromatin states into different groups
lst = lst_dat_region_annot_import
lst = lapply(lst, function(dat){
    ### add column
    dat = dat %>% tidyr::unite("Note", c(Group, Label), sep = ":", remove = FALSE)
    
    ### split group
    dat_ccres    = dat %>% dplyr::filter(Group == "cCREs")
    dat_chromhmm = dat %>% dplyr::filter(Group == "ChromHMM")
    dat_both     = dat %>% dplyr::filter(Group %in% c("cCREs", "ChromHMM"))
    #dat_silencer = dat %>% dplyr::filter(str_detect(Group, "silencer"))

    ### label each group
    tmp = list(
        "ccres"    = dat_ccres,
        "chromhmm" = dat_chromhmm,
        "ccres.chromhmm" = dat_both
        #"silencer" = dat_silencer
    )
    return(tmp)
})

### assign
lol_dat_region_annot_arrange = lst

### summary
lol = lol_dat_region_annot_arrange
for (idx1 in names(lol)){
    lst = lol[[idx1]]
    cat("Region:", idx1, "\n")
    for (idx2 in names(lst)){
        dat = lst[[idx2]]
        cat("Note: ", idx2, "\n")
        cat("Count:", scales::comma(nrow(dat)), "\n")
    } ### end inner loop
    cat("\n")
} ### end outer loop
```

**Show table**

```r
lol = lol_dat_region_annot_arrange
dat = lol[["fcc_astarr_macs_input_overlap"]][["ccres"]]
print(dim(dat))
fun_display_table(head(dat, 3))
```

## Matrix

```r
### assign essential label
lol = lol_dat_region_annot_arrange
lol = lapply(lol, function(lst){
    lst = lapply(lst, function(dat){
        dat = dat %>%
            dplyr::select(Chrom, ChromStart, ChromEnd, Region, Note) %>%
            dplyr::mutate(Value = 1) %>%
            dplyr::distinct()
        dat = dat %>%
            tidyr::spread(Note, Value) %>%
            replace(is.na(.), 0)
        return(dat)
    })
    return(lst)
})

### assign
lol_dat_region_annot_matrix = lol

### show
lol = lol_dat_region_annot_matrix
for (idx1 in names(lol)){
    lst = lol[[idx1]]
    cat("Region:", idx1, "\n")
    for (idx2 in names(lst)){
        dat = lst[[idx2]]
        cat("Note: ", idx2, "\n")
        cat("Count:", scales::comma(nrow(dat)), "\n")
    } ### end inner loop
    cat("\n")
} ### end outer loop
```

**Show matrix**

```r
lol = lol_dat_region_annot_matrix
dat = lol[["fcc_astarr_macs_input_overlap"]][["ccres"]]
print(dim(dat))
fun_display_table(head(dat, 3))
```

```r
lol = lol_dat_region_annot_matrix
dat = lol[["fcc_astarr_macs_input_union"]][["chromhmm"]]
print(dim(dat))
fun_display_table(head(dat, 3))
```

```r
lol = lol_dat_region_annot_matrix
dat = lol[["fcc_astarr_macs_input_union"]][["ccres.chromhmm"]]
print(dim(dat))
fun_display_table(head(dat, 3))
```

## Export results

```r
lol = lol_dat_region_annot_arrange
for (idx1 in names(lol)){
    ### get list of regions
    txt_folder = idx1
    lst = lol[[idx1]]
    
    for (idx2 in names(lst)){
        ### get and arrange table
        txt_group = idx2
        dat = lst[[idx2]]
        dat = dat %>% dplyr::arrange(Chrom, ChromStart, ChromEnd)
        
        ### set file directory
        txt_fdiry = file.path(FD_RES, "region_annotation", txt_folder, "summary")
        txt_fname = paste("region", "annotation", "encode_chromatin_states", txt_group, "tsv", sep = ".")
        txt_fpath = file.path(txt_fdiry, txt_fname)
        
        ### show progress
        cat("Region:", txt_folder, "\n")
        cat("Group: ", txt_group,  "\n")
        cat("File:  ", txt_fname,  "\n")
        cat("\n")

        ### write table
        write_tsv(dat, txt_fpath)
    } ### end inner loop
} ### end outer loop
```
